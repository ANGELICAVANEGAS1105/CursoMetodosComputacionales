```python
#Factorial de un numero 
def factorial(n):
  
  if n<0: 
    print("el factorial no existe")

  elif n==0:
    return 1

  else:
      i=1
      while n>1:
        i=i*n
        n=n-1
      return i

factorial(3)
```
