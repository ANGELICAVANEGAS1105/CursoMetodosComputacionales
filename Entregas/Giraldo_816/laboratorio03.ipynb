# **Exponencial, serie alternante**

```python
import matplotlib.pyplot as plt
import numpy as np
```

## **1)** Programa que calcula $e^{-x}$
$$e^{-x}= \sum_{n=0}^{N}(-1)^n\frac{x^n}{n!} $$

```python
def exp_m1(x,N=100):    # se define el primer método para calcular el inverso
  Sum=0                 # de la exponencial que usa la definición de esta   
  for i in range(N):    #función en series
    Sum += ((-1)**i)*(x**i)/np.math.factorial(i)
  return(Sum)
```

```python
def exp_m2(x,N=100):    # se define el segundo método para calcular el inverso
  Sum=0                 # de la exponencial, esta vez se usa la definición
  for i in range(N):    # en serie de la exponencial y se toma su inverso
    Sum += (x**i)/np.math.factorial(i)
  return(1/Sum)
```

## **2)**  Comparación con exp(x)

```python
a1 = np.arange(0,1.1,0.01)
a2 = np.zeros(len(a1))
for i in range(len(a1)):
  a2[i] = exp_m1(-a1[i])
a3=np.exp(a1)

plt.figure(figsize=(20,12))
plt.title('Error para $e^{-x} con x<1 $')
plt.plot(a1,a2,'*-',label='exp_m1')
plt.plot(a1,a3,'+-',label='np.exp')
plt.xlabel('$x$')
plt.ylabel('$e^{-x}$')
plt.legend()
plt.show()
```

Se puede observar que los resultados de la función en este intervalo son bastante acertados.

## **3)**  Cancelación substractiva

```python
a=np.zeros(30)
for i in range(30):
  a[i]=exp_m1(10,i)

b=np.zeros(30)
for i in range(1,31):
  b[i-1]=exp_m2(10,i)
```

```python
for i in range(len(a)-1):
  if a[i]+a[i+1] < 0.001:
    print('los términos',i,'y',i+1,'se cancelan')
```

Como puede apreciarse en la gráfica siguiente y se corrobora en el código anterior, varios de los términos de la sucesión se cancelan casi exactamente.

## **4)** Comparación de exp(-x) con 1/exp(x) 

```python
plt.figure(figsize=(20,12))
plt.title('Comparación de ambos métodos')
plt.plot(np.arange(30),a,'o-',label='Método 1')
plt.plot(np.arange(1,31),b,'o-',label='Método 2')
plt.xlabel('$x$')
plt.ylabel('$e^{-x}$')
plt.legend()
plt.show()
```

Según se ve en la gráfica, el segundo método converge de manera mas rápida gracias, en parte, a que se elimina la cancelación substractiva por lo que se necesitan menos iteraciones para aproximarse al valor exacto de la función.

## **5)** Pérdida de presición

```python
x1=np.arange(11)
y1=np.zeros(len(x1))
for i in range(len(x1)):
  y1[i]=exp_m1(x1[i])
z1=np.exp(-x1)

plt.figure(figsize=(20,12))
plt.title('Convergencia para 1 < x < 10')
plt.plot(x1,y1,'*r-',label='valor obtenido con la función')
plt.plot(x1,z1,'+g',label='valor real')
plt.xlabel('$x$')
plt.ylabel('$e^{-x}$')
plt.legend()
plt.show()
```

Para esta primera parte, se puede apreciar una buena precisión para los primeros términos pero estra se pierde cerca de los dos últimos.

```python
x2=np.arange(11,101)
y2=np.zeros(len(x2))
for i in range(len(x2)):
  y2[i]=exp_m1(x2[i])
z2=np.exp(-x2)

plt.figure(figsize=(20,12))
plt.title('Convergencia con 10 < x < 100')
plt.plot(x2,y2,'*r-',label='valor obtenido con la función')
plt.plot(x2,z2,'+g',label='valor real')
plt.xlabel('$x$')
plt.ylabel('$e^{-x}$')
plt.legend()
plt.show()
```

Para esta parte se nota que, aproximadamente a partir del término 40 los errores se vuelven impredecibles y, en algunos casos bastante grandes.

## **6)** Error 

```python
x3=np.arange(100)
y3=np.arange(len(x3))
for i in range(len(x3)):
  y3[i]=exp_m1(x3[i])
z3=abs(np.exp(x3)-y3)

plt.figure(figsize=(20,12))
plt.plot(x3,z3,'.-m',label='Error')
plt.legend()
plt.show()
```
