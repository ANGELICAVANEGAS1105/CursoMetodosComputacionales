```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

```python
def dU(U,t):
  return [σ*(U[1]-U[0]),
          U[0]*(ρ-U[2])-U[1],
          U[0]*U[1]-β*U[2]]
```

```python
σ=10; β=8/3 ; ρ=10

ti=np.linspace(0,50,100000)
Δt=ti[-1]-ti[-2]

U0=[1,1,1]

x=odeint(dU,U0,ti)[:,0]
y=odeint(dU,U0,ti)[:,1]
z=odeint(dU,U0,ti)[:,2]

fig = plt.figure()
ax = fig.gca(projection="3d")
ax.plot(x,y,z)
plt.draw()
plt.show()
```

```python
dx=np.diff(x)/Δt
dy=np.diff(y)/Δt
dz=np.diff(z)/Δt
xp=(x[:-1]+x[1:])/2
yp=(y[:-1]+y[1:])/2
zp=(z[:-1]+z[1:])/2

plt.figure()
plt.title('Error')
plt.plot(np.arange(0,len(dx)),abs(dx-σ*(yp-xp)))
plt.plot(np.arange(0,len(dy)),abs(dy-(xp*(ρ-zp)-yp)))
plt.plot(np.arange(0,len(dz)),abs(dz-(xp*yp-β*zp)))
plt.show()
```
