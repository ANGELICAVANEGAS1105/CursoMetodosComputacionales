```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
temp=pd.read_html('https://github.com/hernansalinas/Metodos_Computacionales_taller/blob/main/Asignaciones/Laboratorios/dataset/Temperature_Bogota_Barcelona.csv')[0]
```

```python
temp=temp.drop(['Day','Month','Unnamed: 0', 'Unnamed: 1'] , axis=1)
```

```python
temp['AvgTempCelcius']=(temp['AvgTemperature']-32)/1.8
```

```python
temp
```

```python
plt.figure(figsize=(15,8))
plt.title('Temperatura vs año',size=15)
plt.xlabel('Año',size=12)
plt.ylabel('T[°C]',size=12)
plt.plot(temp['Year'][::2],temp['AvgTempCelcius'][::2],'o-r',label='Barcelona')
plt.plot(temp['Year'][1::2],temp['AvgTempCelcius'][1::2],'o-b',label='Bogotá')
plt.legend()
plt.show()
```
