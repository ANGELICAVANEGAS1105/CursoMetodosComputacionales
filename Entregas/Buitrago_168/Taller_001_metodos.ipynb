```python
import matplotlib.pyplot as plt
import numpy as np
```

1. Programar la ecuación para el mapa logístico, y realizar una grafica del valor de xn como funcion de n

```python
def func(r, xn):
  return r*xn*(1-xn)

x0 = 0.5
r = 0.5
Np = 10
def logisticmap(x0,r,Np):
  f= np.zeros(Np)
  for i in range(0,Np):
    xn=func(r ,x0)
    f[i] =x0
    x0=xn
  return f 
```

```python
plt.title("Mapa logistico con Xn=0.5 y r=3.5")
plt.ylabel("Mapa logistico")
plt.xlabel("Valores de n")
plt.plot(logisticmap(0.5,3.5,100))
```

Encontrar dos formas adicionales de expresar $$f(x)$$ para estos valores funcionales determinar las mismas gráficas del numeral 3 para cada uno de las formas de escribir la expresión. En este caso se espera que los resultados varien. ¿A que se debe esto?. La respuesta no es inmediata, constulta algunas fuentes para sacar tu propia conclusion

```python
#se distibuye r

def func1(r,xn) : 
  return xn*(r-r*xn) 

def logisticmap1(x0,r,Np):
  f = np.zeros(Np)
  for i in range(0,Np):
    xn = func1(r,x0)
    f[i] = x0
    x0 = xn
  return f
r = np.array([1.9, 2.9, 3.1, 3.5, 3.9, 4.0])
Np = 100 # Numero sugerido de iteraciones
x0 = 0.5
plt.figure(figsize=(16,16))
for i in range(len(r)):
  plt.subplot(3,3,i+1)
  plt.plot(logisticmap1(x0,r[i],Np), label= f'r = {r[i]}', color='r')
  plt.title(f'r = {r[i]} Np = {Np} x0 = {x0}')  
```

```python
#se distribuye Xn

def func2(r,xn) : 
  return r*(xn-xn*xn) 

def logisticmap2(x0,r,Np):
  f = np.zeros(Np)
  for i in range(0,Np):
    xn = func2(r,x0)
    f[i] = x0
    x0 = xn
  return f
r = np.array([1.9, 2.9, 3.1, 3.5, 3.9, 4.0])
Np = 100 # Numero sugerido de iteraciones
x0 = 0.5
plt.figure(figsize=(16,16))
for i in range(len(r)):
  plt.subplot(3,3,i+1)
  plt.plot(logisticmap2(x0,r[i],Np), label= f'r = {r[i]}', color='b')
  plt.title(f'r = {r[i]} Np = {Np} x0 = {x0}')  
```

```python
plt.figure(figsize=(20,10))
for i in range(len(r)):
  plt.subplot(3,3,i+1)
  plt.plot(logisticmap2(x0,r[i],Np), ".", color='b')
  plt.plot(logisticmap1(x0,r[i],Np), "--", color='r')

plt.show()
```

Como podemos ver las gráficas aunque a simple vista son exactamente iguales, cuando se ponen una sobre la otra, podemos ver que las pequeñas variaciones se hacen visaibles, como puntos azules que asoman de la linea roja.

Mi conclusión es que debido a que se cambió la posicion de algunas variables, los calculos aunque para nosotros son los mismos, para un computador no lo son y las aproximaciones que se hacian en cuanto a decimales varian de una a la otra, lo que genera estas pequeñas discrepancias.
