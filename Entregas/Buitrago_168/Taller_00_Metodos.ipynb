3. Determine si un número n entero ingresado por el usuario es un palíndromo, (Retorne True en caso afirmativo y False en caso contrario)

```python
#Se pide el numero
print('Programa para comprobar si un numero es palindromo')
num=int(input("Inserte numero: "))
#variable temporal
temp=num
rev=0
#el ultimo digito se obtiene con el modulo
while(num>0):
    dig=num%10
#se invierte el numero    
    rev=rev*10+dig
    num=num//10
#se compara el numero guardado temporalmente con el obtenido     
if(temp==rev):
    print("True")
else:
    print("False")
```

5. Elabore un método para calcular el n-ésimo término de la sucesión de fibonacci:

```python
num = int(input('inserte el número del término:')) 
def fibo(num):
    if num<0:
      print('el número debe ser mayor o igual a 0')
    elif num == 0:
        return 0
    elif num == 1:
        return 1
    else:
        return fibo(num-1)+fibo(num-2)
print(fibo(num))
```

5. Entrar un arreglo de números y ordenarlos de forma ascendente, Ver algoritmo Quicksort.

```python
def particion(l, r, nums):

	pivote, ptr = nums[r], l
	for i in range(l, r):
		if nums[i] <= pivote:

			nums[i], nums[ptr] = nums[ptr], nums[i]
			ptr += 1

	nums[ptr], nums[r] = nums[r], nums[ptr]
	return ptr

def quicksort(l, r, nums):
	if len(nums) == 1: 
		return nums
	if l < r:
		pi = particion(l, r, nums)
		quicksort(l, pi-1, nums) 
		quicksort(pi+1, r, nums) 
	return nums

array = list()
num = input("Cuantos elementos quiere:")
print ('inserte los elementos uno a uno: ')
for i in range(int(num)):
    n = input("num :")
    array.append(int(n))
print(quicksort(0, len(array)-1, array))
```
