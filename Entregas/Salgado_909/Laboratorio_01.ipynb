```python
def fun(r, Xn):
  return r*Xn*(1-Xn)
```

```python
import numpy as np
import matplotlib.pyplot as plt


Xo = 0.5
r = np.array([1.9, 2.9, 3.1, 3.5, 3.9, 4.0])
N = 100

def logismap(Xo, r, N):
  f = np.zeros(N)

  for i in range(0, N):
   Xn = fun(r , Xo)
   f[i] = Xo
   Xo = Xn
  return f
print(f)
```

```python
plt.plot(logismap(0.5, 3.5, 100))
```

```python
plt.figure()
plt.subplot(221)
plt.plot(logismap(0.5, 3.5, 100))
#plt.show()

plt.subplot(222)
plt.plot(logismap(0.5, 3.5, 100))
#plt.show()

plt.subplot(223)
plt.plot(logismap(0.5, 3.5, 100))
#plt.show()

plt.subplot(224)
plt.plot(logismap(0.5, 3.5, 100))
#plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

Np = 1000
xf = np.zeros(Np)
r  = np.linspace(2.5, 4, Np)

def logist_map(Np, xf, r):
  xf = np.zeros(Np)

for i in range(0, Np): 
  xini = np.random.random()
  x_i = logist_map(r = r[i], xf = xini, Np = 1000) 
  xf[i] = x_i


plt.plot((r-1/r), xf, ".", markersize = 0.5)
```
