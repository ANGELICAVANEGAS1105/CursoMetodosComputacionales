El dataset Temperature_Bogota_Barcelona contiene información de la temperatura (en grados Fahrenheit) promedio de Bogotá y Barcelona desde el año 1995. Para el presente dataset realizar lo siguiente:

a) Leer el dataset desde el link sugerido

```python
import pandas

dataset = 'Temperature_Bogota_Barcelona.csv'
data = pandas.read_csv(dataset)
print(data)
```

b) Eliminar las columnas Month, day y Unnamed

```python
data.drop(['Month', 'Day', 'Unnamed: 0'], axis=1)
```

c) Adicionar una nueva columna AvgTempCelcius con la temperatura en grados celcius

```python
from pandas.core.frame import DataFrame

F = data['AvgTemperature']
C = (F-32)/1.8
new_data = data.assign(AvgTempCelcius = C)

print(new_data)
```

d) Realizar un gráfico de AvgTempCelcius como función del año para la ciudad de Bogotá

```python
import matplotlib.pyplot as plt

data1 = new_data[new_data.City == 'Bogota']

plt.plot(data1['Year'],data1['AvgTempCelcius'])
plt.xlabel('Year')
plt.ylabel('AvgTempCelcius')
```

e) Realizar un grafico de AvgTempCelcius como funcion del año para la ciudad de Barcelona

```python
import matplotlib.pyplot as plt

data2 = new_data[new_data.City == 'Barcelona']

plt.plot(data2['Year'],data2['AvgTempCelcius'])
plt.xlabel('Year')
plt.ylabel('AvgTempCelcius')
```
