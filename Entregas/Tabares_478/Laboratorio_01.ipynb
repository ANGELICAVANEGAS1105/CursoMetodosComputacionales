## Caos en la funcion logistica

### 1. y 3. Funcion del mapa logistico
Programar la ecuación para el mapa logístico, y  realizar una grafica del valor de xn como funcion de n.

```python
import numpy as np
import matplotlib.pylab as plt

def LogMap(xi,r,iterations):
    previous = xi
    L = np.zeros(iterations)
    for i in range(0,iterations):
        L[i] = r*previous*(1-previous)
        previous = L[i]
    return L
```

### 2. Pruebas con xi = 0.5 para varios r

Realizar pruebas para un $x?i=0.5$, con valores de r

```python
xi = 0.5
iterations = 100
rValues = [1.9, 2.9, 3.1, 3.5, 3.9,4.0]

xAxis = range(0,iterations)

for r in rValues:
    print("r:",r," with xi:",xi," At",iterations," iterations")
    Map = LogMap(xi,r,iterations)
    plt.plot(xAxis, Map)
    plt.show()
    plt.draw()
```

### 3. Grafico de r = linspace(2.5, 4, 1000) (Funcion Logistica Mandelbrot)

```python
import numpy as np

Np = 1000
xf = []
r  = np.linspace(2.5, 4, Np)

for i in range(0, Np): 
  xini = np.random.random()
  x_i = LogMap(xini,r[i],1000) 
  xf.append(x_i)

plt.plot((r-1/r), xf,  ".", markersize = 0.5);
```
