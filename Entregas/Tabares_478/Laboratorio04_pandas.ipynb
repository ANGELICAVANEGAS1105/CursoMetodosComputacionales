### a)

```python
import requests 
import pandas as pd
import io
import matplotlib.pyplot as plt

r = requests.get("https://raw.githubusercontent.com/hernansalinas/Metodos_Computacionales_taller/main/Asignaciones/Laboratorios/dataset/Temperature_Bogota_Barcelona.csv")
raw_data = r.content
data = pd.read_csv(io.StringIO(raw_data.decode('utf-8')))

data
```

### b)

```python
del data['Unnamed: 0']
del data['Month']
del data['Day']
data
```

### C)

```python
tempFahrenheit = data['AvgTemperature']
avgCelsius = [(F-32)/1.8 for F in tempFahrenheit]
data['AvgTempCelsius'] = avgCelsius 
data
```

### D)

```python
dataBogota = data[data["City"] == "Bogota"]

plt.plot(dataBogota["Year"],dataBogota["AvgTempCelsius"],"b.")
plt.title("Average Bogota Temperature °C")
plt.ylabel("°C")
plt.xlabel("Year");
```

### E)

```python
dataBarcelona = data[data["City"] == "Barcelona"]

plt.plot(dataBarcelona["Year"],dataBarcelona["AvgTempCelsius"],"r.")
plt.title("Average Barcelona Temperature °C")
plt.ylabel("°C")
plt.xlabel("Year");
```
