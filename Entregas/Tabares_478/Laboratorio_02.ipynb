# Alternating series

## $e^{-x}$ function with alternating series:

```python
import math
import matplotlib.pylab as plt

def eToTheMinusAS(x, N):
    value = 0
    for n in range(0,N):
        value += ((-1)**n)*(x**n/math.factorial(n))
    return value

result1 = eToTheMinusAS(1,100)        
print(result1)
print(math.e**-1)   
```

## $e^{-x}$ function 1 under series

```python
def eToTheMinus1S(x, N):
    value = 0
    for n in range(0,N):
        value += (x**n/math.factorial(n))
    return 1/value

result2 = eToTheMinus1S(1,100)
print(result2)
print(math.e**-1)
```

## Plot of alternating series vs iterations ($e^{-1}$)

```python
def plotsComparison(powerValue, iterations):
    xAxis = range(1,iterations)
    yAxisAS = []
    yAxis1S = []
    yAxisBE = []
    for i in xAxis:
        yAxis1S.append(eToTheMinus1S(powerValue,i))
        yAxisAS.append(eToTheMinusAS(powerValue,i))
        yAxisBE.append(math.exp(-powerValue))

    plt.plot(xAxis, yAxisAS, label = "Alternating")
    plt.plot(xAxis, yAxis1S, label = "1 over Series")
    plt.plot(xAxis, yAxisBE, label = "Builtin exp")
    plt.xlabel("Number of iterations")
    plt.ylabel("Value of e^-x")
    plt.legend()
    #plt.axis([0, 10, 0, 1]) #plt.axis([x_min, x_max, y_min, y_max])
    plt.show()

plotsComparison(1,30)
```

We can clearly see it takes at least 20 iterations for the Alternating Series function to stabilize around the expected value of 0.36, meanwhile with the 1 over series function the function stabilizes from the start and keeps it that way!

## Plot for small x Values ($e^{x} x < 2$)

```python
x_values = [1.5,0.9,0.5,0.1]
for x in x_values:
    print("e to the -", x)
    plotsComparison(x,10)
```

## Plot of alternating series vs iterations ($e^{-10}$)

```python
plotsComparison(10,30)
```

## Plot of alternating series vs iterations ($e^{-20}$ ~ $e^{-30}$)

```python
plotsComparison(20,30)
```

Notice how the Alternating method starts being unreliable once the powervalue gets close to the number of iterations, therefore we would have to do a number of iterations greater to the power we're calculating if we use the Alternathing series method for getting the same results that we get with the 1 over series method

```python
plotsComparison(20,40)
```

## Plots of error vs N for various values of x

```python
n_Values = range(1,50)
x_Values = [0.1,0.5,1,2]

def plotErrorvsN(x,n_values):
    errorValues = []
    for n in n_Values:
        errorValues.append(abs(eToTheMinusAS(x,n) - math.exp(x)))
    plt.plot(n_values, errorValues)

for x in x_Values:
    print("e^-",x)
    plotErrorvsN(x,n_Values)
    plt.xlabel("Number of iterations")
    plt.ylabel("Absolute error of x");
```

- blue: $e^{-0.1}$
- orange: $e^{-0.5}$
- green: $e^{-1}$
- red: $e^{-2}$
