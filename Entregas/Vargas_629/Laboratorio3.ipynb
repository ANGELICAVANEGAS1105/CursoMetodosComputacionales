```python
from math import*
import matplotlib.pyplot as plt
import numpy as np

```

Metodo 1:
Serie alternante.


```python
def mexp(x,n):
    ex=0
    for i in range(0,n):
        s=((-1)**i)*((x**i)/factorial(i))
        ex+=s
    return ex
```

Calculo de la serie para $$x\leq1$$

```python
print(abs(mexp(0.1,11))>=abs(10e-7*mexp(0.1,10)))  #Condición.
print("Valor por sumatoria:",mexp(0.1,10))
print("Valor exacto:",exp(-0.1))
print("Diferencia:",abs((1/exp(0.1))-mexp(0.1,10)))
```

Metodo 2:

$$1/e^x$$

```python
def mexp2(x,n):
    ex=0
    for i in range(0,n):
        s=(x**i)/factorial(i)
        ex+=s
    exp=ex**-1
    return exp
```

```python
print(abs(mexp2(0.1,11))>=abs(10e-7*mexp2(0.1,10)))  #Condición.
print("Valor por sumatoria:",mexp2(0.1,10))
print("Valor exacto:",exp(-0.1))
print("Diferencia:",abs((1/exp(0.1))-mexp2(0.1,10)))
```

Para $x=0.1$ y $n=10$ El método 1 resulta más preciso que el método 2.

Gráfica del método 1 para $$ 0\leq n \leq 30$$

```python
lista1=[]
lista2=[]
lista=[]
for i in range(0,30):
    lista1.append(i)
    lista2.append(mexp(10,i))
    lista.append(exp(-10))


x1=np.array(lista1)
x2=np.array(lista2)
x3=np.array(lista)
plt.plot(x1,x2);
plt.plot(x1,x3);
plt.plot(9,mexp(10,9),color='red',marker='*');     #Cancelación para n=x-1.
```
