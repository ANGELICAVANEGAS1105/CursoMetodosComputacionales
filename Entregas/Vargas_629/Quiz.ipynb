```python
import numpy as np
import matplotlib.pylab as plt
import scipy.integrate as integrate
```

```python
def dU_dt(U,t,s=10,b=8/3,r=10): 
    x,y,z=U
    return [s*(y-x),x*(r-z)-y,x*y-(b*z)]

U0 = [1,1,1]
t = np.linspace(0, 10000, 1000000)
Us = integrate.odeint(dU_dt, U0, t)
x=Us[:,0]
y=Us[:,1]
z=Us[:,2]
fig = plt.figure()
ax = fig.gca(projection="3d")
ax.plot(x,y,z)
plt.draw()
plt.show()
```

```python
def dU_dt(U,t,s,b,r): 
    x,y,z=U
    return [s*(y-x),x*(r-z)-y,x*y-(b*z)]

U0 = [1.001,1.001,1.001]
t = np.linspace(0, 10, 1000000)
Us = integrate.odeint(dU_dt, U0, t, args=(10,8/3,25))
x=Us[:,0]
y=Us[:,1]
z=Us[:,2]
fig = plt.figure()
ax = fig.gca(projection="3d")
ax.plot(x,y,z)
plt.draw()
plt.show()
```
