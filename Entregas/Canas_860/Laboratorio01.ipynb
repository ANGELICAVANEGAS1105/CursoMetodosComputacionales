#  Mapa logistico 
Utilizar la ecuación logistica
$$ x_{n+1} = rx_{n}(x_{n} -1)$$

```python
import numpy as np
import matplotlib.pylab as plt
import pandas as pd
```

```python
r = np.array([1.9,2.9,3.1,3.9,4.0])
def Map_Log(xn,r): 
    xn = r*xn*(1-xn)  #Se define la secuencia logistica   
    return xn

def Mapa_Logistico(xi,r,N):
    z = np.zeros(N)
    for i in range(0,N):  #Se construye la función que toma cada valor de la ecuación y lo almacena en la variable z
        x0 = Map_Log(xi,r)
        xi = x0
        z[i] = x0
    return z
```

```python
fig, ax = plt.subplots(6) 
plt.figure(figsize=(8,8))
ax[0].plot(np.arange(0,100),Mapa_Logistico(0.9,1.9,100))
ax[1].plot(np.arange(0,100),Mapa_Logistico(0.9,2.9,100))
ax[2].plot(np.arange(0,100),Mapa_Logistico(0.9,3.1,100))
ax[3].plot(np.arange(0,100),Mapa_Logistico(0.9,3.5,100))
ax[4].plot(np.arange(0,100),Mapa_Logistico(0.9,3.9,100))
ax[5].plot(np.arange(0,100),Mapa_Logistico(0.9,4,100))
fig.suptitle('Mapa logistico')
plt.show()
```

```python
#
Np = 1000
r  = np.linspace(2.5, 4, Np)
xf = np.zeros(Np)

for i in range(0, Np):
    xini = np.random.random()
    x_i = Mapa_Logistico(xi = xini,r = r[i], N = 1000)
    xf[i] = x_i[i]
plt.plot((r-1/r),xf,".",markersize = 0.5)
plt.grid()
```
