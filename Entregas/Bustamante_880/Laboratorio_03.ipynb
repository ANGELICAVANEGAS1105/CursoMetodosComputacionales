```python
import numpy as np
from math import factorial
import matplotlib.pyplot as plt

```

```python
#### Punto 1 (calcular e**(-x))

def exp_metodo(x,N):
  r=0
  for n in range(0,N):
    r = r + (-1)**n*((x)**n/factorial(n))
  return r
```

```python
###Punto 4

def exp_metodo2(x,N):
  c=0
  for n in range(0,N):
    c = c + x**n/factorial(n)
  return 1/c



print("Método con e**(-x): {}".format(exp_metodo(10,1000)))
print("Método con 1/e**x: {}".format(exp_metodo2(10,1000))) ###Vemos que esta opción es mejor
print("Método con la función de python(valor más preciso): {}".format(np.exp(-10)))
```

```python
def error_(x,N): 
  exp=0
  i=0
  lista=[]

  j=0
  exp_2=0
  lista_2=[]

  N_graf=np.arange(0,100,1)

  while i<N:
    exp = exp + (-1)**i*((x)**i/factorial(i))
    lista.append(exp)

    exp_2 = exp_2 + x**i/factorial(i)
    lista_2.append(1/exp_2)
    i+=1

  plt.plot(N_graf,lista)
  plt.plot(N_graf,lista_2)
  plt.xlabel("N")
  plt.xlabel("valor")
  plt.legend(["gráfico programa", "gráfico ideal"])
  plt.xlim(0,N)
  plt.ylim(1,-1)
  plt.show()

error_(1,100)
error_(5,100)
error_(10,100)
error_(15,100)
error_(30,100)
error_(40,100)
```
