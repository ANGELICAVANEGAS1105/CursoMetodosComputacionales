```python
import matplotlib.pyplot as plt
import numpy as np


def fun(r,xn):
  return r*xn*(1-xn)

#xo=0.5
#r=0.5
#Np=10

def logistmap(xo,r,Np):
  f=np.zeros(Np)

  for i in range(0,Np):

    xn=fun(r,xo)
    f[i]=xo
    xo=xn
  return f
```

```python
plt.figure(figsize=(15,15))
plt.subplot(3,2,1)
plt.plot(logistmap(0.5,1.9,100))
plt.xlabel("Tiempo")
plt.ylabel("Población")


plt.subplot(3,2,2)
plt.plot(logistmap(0.5,2.9,100))
plt.xlabel("Tiempo")
plt.ylabel("Población")

plt.subplot(3,2,3)
plt.plot(logistmap(0.5,3.1,100))
plt.xlabel("Tiempo")
plt.ylabel("Población")

plt.subplot(3,2,4)
plt.plot(logistmap(0.5,3.5,100))
plt.xlabel("Tiempo")
plt.ylabel("Población")


plt.subplot(3,2,5)
plt.plot(logistmap(0.5,3.9,100))
plt.xlabel("Tiempo")
plt.ylabel("Población")

plt.subplot(3,2,6)
plt.plot(logistmap(0.5,4.0,100))
plt.xlabel("Tiempo")
plt.ylabel("Población")


plt.show()
```

```python
def logist_map(r0,xi,N):
  return r0*xi*(1-xi)

Np = 1000
xf = np.zeros(Np)
r  = np.linspace(2.5, 4, Np)

for i in range(0, Np): 
  xini = np.random.random()
  x_i = logist_map(r[i], xini, Np) 
  xf[i] = x_i

plt.plot((r-1/r), xf,  ".", markersize = 0.5)
```
