```python
import numpy as np
import matplotlib.pyplot as plt
```

Dado un Valor inicial $x_0$ y un parámetro $r$,  la expresión para el mapa logístico  


$$x_{n+1} = rx_n(1 - x_n)$$



```python
fun = lambda r,xn : r*xn*(1-xn) 

def logist_map(xo,r,Np):
  f = np.zeros(Np)
  for i in range(0,Np):
    xn = fun(r,xo)
    f[i] = xo
    xo = xn
  return f
```

```python
r = np.array([1.9, 2.9, 3.1, 3.5, 3.9, 4.0])
Np = 100
xo = 0.5
plt.figure(figsize=(20,15))
plt.suptitle('x en funcion de número de iteraciones ($N_p$) para diferentes valores del parámetro (r)')

for i in range(len(r)):
  plt.subplot(3,3,i+1)
  plt.plot(logist_map(xo,r[i],Np), label= f'r = {r[i]}')
  plt.title(f'r = {r[i]}   x0 = {xo}   Np = {Np}')
```

```python
Np = 1000
xf = np.zeros([Np,Np])
r  = np.linspace(2.5, 4, Np)
# xini = np.random.random()

for i in range(0, Np): 
  xini = np.random.random()
  x_i = logist_map(xini,r[i], Np) 
  xf[i] = x_i

plt.figure(figsize = (25,10))
plt.title(f'x en función del número de iteraciones ($N_p$). Valor inicial aleatorio. Número de iteraciones: {Np}')
plt.xlabel('Número de iteraciones $Np$')
plt.ylabel('$x_{Np}$')
plt.plot(xf,'y--', linewidth = .5)
plt.xlim 
plt.show()
```
