```python
def factorial(N):
  n = N
  if(N == 0):
    n = 1
    return n
  elif(N<0):
    return print("El numero debe ser mayor o igual a 0")
  else:
    for k in range(1, N-1):
      n = n*(N-k)
    return n

print(f"Factorial: {factorial(5)}")
```

##Problema 2.

```python
def divisores(num):
  div=[]
  for i in range(1, num):
    if num % i == 0:
      div.append(i)
  return div

def primo(num):
  if num < 1:
    return 'No es primo'
  elif num == 2:
    return 'Es primo'
  else:
    for i in range(2, num):
      if num % i == 0:
        return 'No es primo'
    return 'Es primo'

def defectivo(num):
  if primo(num):
    return 'Es defectivo'
  elif sum(divisores(num)) < num:
    return 'Es defectivo'
  else:
    return 'No es defectivo'

def abundante(num):
  if sum(divisores(num)) > num:
    return 'Es abundante'
  else:
    return 'No es abundante'

def perfecto(num):
  if sum(divisores(num)) == num:
    return 'Es perfecto'
  else:
    return 'No es perfecto'

def semiperfecto(num, div):
  if num == 0:
    return True
  for i in range(0, len(div)):
    temp = div.pop(i)
    if (semiperfecto(num - temp, div)):
      return True
    div.insert(i, temp)
  return False

def amigos(a, b):
  if sum(divisores(a)) == b and sum(divisores(b)) == a:
    return print(f'{a} y {b} son numeros amigos.')
  else:
    return print(f'{a} y {b} no son numeros amigos.')


num_a= int(input('Escibe el primer numero: '))
num_b = int(input('Escribe el segundo numero: '))

print('Que le gustaria encontrar?')
p = int(input('''
0- Numeros primos
1- Numeros defectivos.
2- Numeros abundantes.
3- Numeros perfectos.
4- Numeros semiperfectos.
5- Numeros amigos
'''))
if p == 0:
  print(f'Primer numero: {primo(num_a)}, segundo numero: {primo(num_b)}')#
elif p == 1:
  print(f'Primer numero: {defectivo(num_a)}, segundo numero: {defectivo(num_b)}')
elif p == 2:
  print(f'Primer numero: {abundante(num_a)}, segundo numero: {abundante(num_b)}')
elif p == 3:
  print(f'Primer numero: {perfecto(num_a)}, segundo numero: {perfecto(num_b)}')
elif p == 4:
  print(f'Primer numero: {semiperfecto(num_a, divisores(num_a))}, segundo numero: {semiperfecto(num_b, divisores(num_b))}')
elif p == 5:
  print(amigos(num_a, num_b))
else: 
  print('El numero ingresado es incorrecto.')



```
