```python
import math
import matplotlib.pyplot as plt
import numpy as np
```

##Metodo 1

\begin{equation}
e^{-x} = \sum_{n=0}^N \frac{(-1)^n x^n}{n!}
\end{equation}

```python
def metodo_1 (x, N):
  f = 0
  for n in range (0,N):
      f = f + ((-1)**n) * ((x)**n)/(math.factorial(n))
  return f

for num in range (1, 4):
  num*=5
  print(f"El resultado con el metodo 1, para el x = {num} y N = 10, es: {metodo_1(num,10):.10f}")
  print(f"El resultado con el modulo math, para el valor {num} , es: {math.exp(-num):.10f}")
  print("")
```

##Metodo 2

\begin{equation}
e^{x} = \sum_{n=0}^N \frac{x^n}{n!}
\end{equation}

\begin{equation}
e^{-x} = 1/e^{x}
\end{equation}

```python
def metodo_2 (x, N):
  f = 0
  for n in range (0,N):
    f = f + (x**n)/(math.factorial(n))
  return 1/f

for num in range (1, 4):
  num*=5
  print(f"El resultado con el metodo 2, para el x = {num} y N = 10 es: {metodo_2(num,10):.10f}")
  print(f"El resultado con el modulo math, para el valor {num} , es: {math.exp(-num):.10f}")
  print("")
```

Se puede notar que el metodo 2 es el mas efectivo, pues sus resultados son mas exactos que los hallados con el metodo 1.

##Graficas

```python
x = 10
N = np.arange(1,25,1)
F = np.zeros(len(N))
G = np.zeros(len(N))
E = np.zeros(len(N))
Error_F = np.zeros(len(N))
Error_G = np.zeros(len(N))

for i, n in enumerate(N):
  F[i] = metodo_1(x,n)
  G[i] = metodo_2(x,n)
  E[i] = math.exp(-n)
  Error_F[i] = abs(E[i] - F[i])
  Error_G[i] = abs(E[i] - G[i])
```

###Grafica del metodo 1 vs N.

```python
plt.plot(N, F, color='red', label='Metodo 1')
plt.plot(N, E, color='black', label='Original')

plt.legend()
plt.grid()
plt.xlabel("N")
plt.ylabel("f")
```

###Grafica del metodo 2 vs N.

```python
plt.plot(N, G, color='green', label='Metodo 2')
plt.plot(N, E, color='black', label='Original')

plt.legend()
plt.grid()
plt.xlabel("N")
plt.ylabel("f")
```

En las graficas se puede comprobar que el metodo 2 es mas exacto, pues sigue una trayectoria similar a la del exponencial.
Tambien se puede observar que, para ambos metodos la diferencia de convergencia entre las rectas es de, aproximadamente, 0.00005. Teniendo para el metodo 1, una N de aproximadamente 37; y para el metodo 2, una N de aproximadamente 20.

###Grafica del error del metodo 1 vs N.

```python
x = []
N = np.arange(1,25,1)
F = np.zeros(len(N))
G = np.zeros(len(N))
E = np.zeros(len(N))
Error_F = np.zeros(len(N))
Error_G = np.zeros(len(N))

lenN = len(N) + N[0]
for m in range (N[0], lenN):
  m+=1
  x.append(m)

for i, n in enumerate(N):
  F[i] = metodo_1(x[i],n)
  G[i] = metodo_2(x[i],n)
  E[i] = math.exp(-n)
  Error_F[i] = abs(E[i] - F[i])
  Error_G[i] = abs(E[i] - G[i])
```

```python
plt.plot(N, E, color='black', label='Original')
plt.plot(N, F, color='red', label='Metodo 1')
plt.plot(N, Error_F, color='skyblue', label='Error 1')

plt.legend()
plt.grid()
plt.xlabel("x")
plt.ylabel("error")
```

###Grafica del error del metodo 2 vs N.

```python
plt.plot(N, E, color='black', label='Original')
plt.plot(N, G, color='green', label='Metodo 2')
plt.plot(N, Error_G, color='skyblue', label='Error 2')

plt.legend()
plt.grid()
plt.xlabel("x")
plt.ylabel("error")
```
