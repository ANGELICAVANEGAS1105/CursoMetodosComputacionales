**2.0** El algoritmo de gradiente descendente permite encontrar el minimo de funciones en espacios multidimensionales proponiendo puntos en la direccion en la que el gradiente desciende. Para el caso de una dimensión, podemos encontrar el mínimo realizando iteraciones y multiplicando su derivada en cada nuevo punto por una constante $\alpha$

\begin{equation}
x_{i+1}=x_i - \alpha \frac{\mathrm{d}f}{\mathrm{d}x}(x_i)
\end{equation}




Ejemplo:

Sea $f(\omega) = (\omega-1)^2$ , con $\omega \in \mathbb{R} $. El minimo de esta función esta determinado por
\begin{equation}
\frac{\mathrm{d}f}{\mathrm{d}\omega} = 0
\end{equation}

\begin{equation}
\omega = 1
\end{equation}


Para determinar el minimo,  a través del gradiente descendente puede ser aplicada el siguiente algoritmo:
1. Proponer un número aleatorio inicial $\omega_i$
2. Para descender al mínimo de la función, encontremos un  valor para el cual 
la derivada de la función descenciende, asi:
\begin{equation}
\omega_{i+1} = \omega_{i} - \alpha \frac{\mathrm{d}f(\omega_i)}{\mathrm{d}\omega}
\end{equation}

donde, $\alpha$ es conocido como la tasa de aprendizaje del algoritmo. 

3. Evaluar $f(\omega_{i+1})$

4. Iterar hasta encontrar el minimo de la función 


En el caso de la funcion propuesta, asumiendo un valor de $\alpha=1$, tenemos que:

1. Supongamos que $\omega_{0} = 2$, luego :
2. $\omega_{0+1} = 2 - 1 f'(\omega_{0}) =  = 0$
3. $f(0) = 1$
4. $\omega_{1+1} = 0 - 1 f'(\omega_{1}) = 0-1(2(0-1)) =  2$

El valor del $\omega$  despues de un par de iteraciones, no es el esperado, por que no minimiza la función, se debe proponer un tasa de aprendizaje  **$\alpha$** de tal forma que permita descender suavemente para encontrar el minimo.

````python
import numpy as np
import matplotlib.pyplot as plt

''' Ecuación para el mapa logístico, con gráfica del valor de xn como funcion de n (valores fijos para r y x) '''

def f(x_n):
  return r*x_n*(1-x_n)

r = 3.1
x = 0.5
n = 100

X = np.zeros(n)
X[0] = x

for i in range(1,n):
  x_n = f(x)
  x = x_n
  X[i] = (x_n)

# Grafica de los valores de X
plt.title("Mapa logístico", fontsize = 16)
plt.plot(X)
plt.xlabel('$x$', fontsize = 12)
plt.ylabel('$x_n$', fontsize = 12)

plt.show()
````

````python
import numpy as np
import matplotlib.pyplot as plt

''' Función que realiza el mapa logístico para n iteraciones '''

def logistic_map(r, x, n):
    xn = [x] # Lista para almacenar los valores de xn
    for i in range(n):
        x = r * x * (1 - x) # Ecuación del mapa logístico
        xn.append(x)
    return xn[1:] # Obtiene solo los últimos elementos x de xn

# Parámetros iniciales
r = np.array([1.9, 2.9, 3.1, 3.5, 3.9, 4.0])
x0 = 0.5
numpoints = 100

# Crear un array de valores de x para graficar
x = np.linspace(0, 1, numpoints)

# Crear una matriz de subplots
fig, axs = plt.subplots(2, 3, figsize=(12, 8), sharex=True, sharey=True)

# Graficar el resultado para cada valor de r
colors = ['b', 'g', 'r', 'c', 'm', 'y']
for i, ax in enumerate(axs.flat):
    xn = logistic_map(r[i], x0, numpoints)
    ax.plot(x, xn, color=colors[i])
    ax.set_title("r = {}".format(r[i]), fontsize=14)
    ax.set_xlabel("x", fontsize=12)
    ax.set_ylabel("$x_n$", fontsize=12)
    ax.legend(["r = {}".format(r[i])], loc='best', fontsize=10)

# Gráfica
fig.suptitle("Mapa logístico para diferentes valores de r".format(x0), fontsize=16)
fig.subplots_adjust(left=0.06, bottom=0.06, right=0.95, top=0.9, wspace=0.1, hspace=0.2)

plt.show()
````

````python
import numpy as np
import matplotlib.pyplot as plt

""" Función que realiza el mapa logístico para N iteraciones, dado un valor de r y una condición inicial xi """
def logist_map(r, xi, N):
    xn = xi
    for i in range(N):
        xn = r * xn * (1 - xn) # Ecuación del mapa logístico
    return xn

# Parámetros iniciales
Np = 1000
xf = np.zeros(Np)
r = np.linspace(2.5, 4, Np)

# Calcular los valores finales de x para cada valor de r
for i in range(Np): 
    xini = np.random.random()
    xf[i] = logist_map(r[i], xini, Np)

# Graficar el resultado
plt.plot(r, xf, "o", markersize = 0.5)
plt.title("Diagrama de bifurcación del mapa logístico", fontsize = 16)
plt.xlabel("$r$", fontsize = 12)
plt.ylabel("$x$", fontsize = 12)

plt.show()
````

## Formas adicionales de expresar  $f(x) = rx (1-x)$:


La función $f(x)$ que representa una ecuación de mapa logístico se puede representar de diferentes maneras:

*   $f(x) = x^2 + r$
*   $f(x) = rx(1-tanh(x))$
*   $f(x) = sen(rx)$












````python
import numpy as np
import matplotlib.pyplot as plt

""" 
Función que realiza el mapa logístico para N iteraciones, dado un valor de 
r y una condición inicial xi, forma polinómica de la f(x) 
"""

def logist_map(r, xi, N):
    xn = xi
    for i in range(N):
        xn = r*xn - xn**2 # forma cuadrática de la ecuación de mapa logístico
    return xn

# Parámetros iniciales
Np = 1000
xf = np.zeros(Np)
r = np.linspace(2.5, 4, Np)

# Calcular los valores finales de x para cada valor de r
for i in range(Np): 
    xini = np.random.random()
    xf[i] = logist_map(r[i], xini, 1000)

# Graficar el resultado
plt.plot(r, xf, "o", markersize = 0.5)
plt.title("Diagrama de bifurcación del mapa logístico", fontsize = 16)
plt.xlabel("$r$", fontsize = 12)
plt.ylabel("$x$", fontsize = 12)

plt.show()
````

**2.** El algoritmo de gradiente descendente.

````python
import numpy as np
import matplotlib.pyplot as plt

"""
    Encuentra el mínimo de una función utilizando el método de descenso del gradiente.

    Input:
        func: es la función lambda que representa la función a minimizar.
        grad_func: es la función lambda que representa el gradiente de la función.
        alpha: tasa de aprendizaje o tamaño del paso.
        max_iter: número máximo de iteraciones permitidas.
        wi: valor inicial de la variable de la función.

    Output:
        El valor de x que minimiza la función.
    """

def minimo_descenso_gradiente(func, grad_func, alpha=0.01, max_iter=1000, w_i=0.0):
    for i in range(max_iter):
        wi_1 = w_i - alpha * grad_func(w_i)
        if abs(func(wi_1) - func(w_i)) < 1e-6: # 1e-6 es el valor estándar de tolerancia
            break
        w_i = wi_1
    return wi_1
````

````python
# Definir la función y su derivada
f = lambda x: x**2
grad_f = lambda x: 2*x

minimo = minimo_descenso_gradiente(f, grad_f, alpha=0.1, max_iter=1000, w_i=1.0)
print("El mínimo de la función es:", minimo) 
````

````python
import numpy as np
import matplotlib.pyplot as plt

# Definir la función y su derivada
f = lambda x: x**2
df = lambda x: 2*x

# Crear un array con valores de x
x = np.linspace(-5, 5, 100)

# Evaluar la función y la derivada en cada valor de x
y = f(x)
dy = df(x)

# Crear la figura y los ejes
fig, ax = plt.subplots()

# Graficar la función y la derivada
ax.plot(x, y, label='Función f(x) = $x^2$')
ax.plot(x, dy, label='Derivada f\'(x) = $2x$')

# Agregar leyendas
ax.legend()

# Agregar línea vertical para el mínimo
ax.axvline(x=minimo, linestyle='--', color='r', label='Mínimo encontrado')

# Agregar leyenda de la línea vertical
ax.legend()

plt.title("Mínimo de $X^2$", fontsize = 16)
plt.xlabel("$x$", fontsize = 12)
plt.ylabel("$y$", fontsize = 12)
plt.grid() 

plt.show()
````

````python
import numpy as np
import matplotlib.pyplot as plt

# Definimos la función y su gradiente
def f(x):
    return x**2

def grad_f(x):
    return 2*x

# Definimos los parámetros del algoritmo
alphas = [0.1, 0.01]
max_iter = 50
w_i = 4.0

# Graficamos la función en cada punto a iterar
for alpha in alphas:
    x = [w_i]
    y = [f(w_i)]
    for i in range(max_iter):
        wi_1 = w_i - alpha * grad_f(w_i)
        x.append(wi_1)
        y.append(f(wi_1))
        if abs(y[-1] - y[-2]) < 1e-6:
            break
        w_i = wi_1
    
    plt.plot(range(len(y)), y, label=f'$alpha$={alpha}')
    plt.scatter(range(len(y)), y)
    
plt.title('Función a minimizar', fontsize = 16)
plt.xlabel('Número de iteraciones', fontsize = 12)
plt.ylabel('Valor de la función', fontsize = 12)
plt.legend()
plt.grid()

plt.show()
````
