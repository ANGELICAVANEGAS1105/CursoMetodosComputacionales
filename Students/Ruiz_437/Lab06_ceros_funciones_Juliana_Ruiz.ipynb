### **1. Pasos para aplicar el método de bisección**
El método de bisección se basa en el *teorema del valor medio*, en el que una función que es continua y diferenciable debe tener un cero en un intervalo $[a, b]$, en el que $ f(a) f(b) < 0$, o lo que es lo mismo, existe un valor $p$ que pertenezca al intervalo y evaluado en la función sea igual a cero, $ f(p) = 0$.

- *Pasos para aplicar el método:*


> 1. Se debe definir el intervalo $[a, b]$, no sabemos donde se encuentra exactamente el cero de la función, pero gráficamente podemos ver entre qué valores está. 
2. Luego, se debe calcular el valor medio $p_1$ de la siguiente manera:

\begin{equation}
 \ p_1 = \frac{a + b}{2}
\end{equation}


> 3. Se consideran los subintervalos $[p_1, b]$ y $[a, p_1]$, entonces:

  - a). Si $f(p_1) = 0$ entonces $p_1$ es raíz de la función $f(x)$ y el método termina Asimismo, si $b−a<ε$, como $α$ (la raíz de $f(x)$) está dentro de $(a,b)$, ocurrirá que $|α−m|<ε/2$, por lo que podemos considerar que $α≅p_1$ (es decir, $p_1$ se aproxima suficientemente a la raíz $α$), y el método también termina.

  - b). En un caso contrario, tenemos que si $f(a)⋅f(p_1)<0$ es que el cambio de signo se produce en $[a,p_1]$, Hacer $b=p_1$ y repetir el paso 3. Si no, es que el cambio de signo se produce en $[p_1,b]$ Hacer $a=p_1$ y repetir el paso 3.





### **2. Deducción del método de Newton - Rapshon**
Dada una función $ f(x)$ derivable en un intervalo abierto $(a, b)$ que contiene un cero $α$, o sea, $f(α) = 0$, queremos encontrar una aproximación de $α$, este método se basa en el desarrollo de la serie de Taylor de la función.

Para eso partimos de un valor inicial $x_0$, que es un punto suficientemente cercano a $α$, podemos escribir la función como:

> $f(x) = f(x_0) + (α - x_0) f'(x_0) + \frac{(α - x_0)^2}{2} f''(ζ)$ 

donde $ζ$ es un punto entre $x$ y $x_0$. Si hacemos $x = α$, tenemos que 

> $0 = f(α) = f(x_0) + (α - x_0) f'(x_0) + \frac{(α - x_0)^2}{2} f''(ζ)$ 

Y despejando $α$ tenemos

> $α = x_0 \frac{f(x_0}{f'(x_0)} - \frac{f''(x_0}{2f'(x_0)}(α - x_0)^2$

Ahora, si el valor inicial es sufucientemente cercano a $α$ se puede suponer que los términos $(α - x_0)$ $(α - x_1)$ son pequeños, y de esta manera el último término de la ecuación anterior se puede despreciar y tenemos una aproximación lineal de $α$:

> $α ≈ x_1 = x_0 - \frac{f(x_0}{f'(x_0)}$

Repetiendo este proceso con un nuevo valor $x_1$ se obtiene una aproximación mejorada, y así sucesivamente se puede repetir el proceso, hasta que se genera una sucesión {$x_n$} que converge a la raíz (cero de la función) $α$ siempre que se cumplan ciertas condiciones sobre la función y el valor inicial. 

- Se debe tener en cuenta que para usar el método se debe partir de una aproximación $x_0$ del cero $α$. Si la aproximación inicial está lejos de la solución, al despreciar el valor $(α−x_0)^2$ podemos estar cometiendo un error muy grande y el método de Newton termina por alejarnos de la raíz.

### **3. Diferencias entre el método de Newton-Rapshon y el método de la secante**

El método de la secante es una variante del método de N-R. El método de Newton - Rapshon tiene una debilidad y es que se necesita conocer la derivada de la función, por esta razón el método de la secante puede ser más simple de aplicar, de la siguiente manera:

- Empezamos al igual que con el método de N-R con la expresión de la serie de Taylor de la función alrededor de un punto $x_0$: 

\begin{equation}
\ f(x) = f(x_0) + (α - x_0) f'(x_0) + \frac{(α - x_0)^2}{2!} f''(x_0) + ...
\end{equation}

- Gráficamente podemos ver donde queda aproxidamente la raíz de la función, entonces podemos tomar dos puntos $x_n$ y $x_{n-1}$ que estén cercanos a la raíz. Luego, podemos aproximar la función por una recta secante que pasa por esos dos puntos, empleando una aproximación con la definición de derivada, $f'(x_0)$

\begin{equation}
\ f(x) ≈ f(x_n) + \frac{f(x_n) - f(x_{n-1})}{x_n - x_{n-1}} (x - x_n)
\end{equation}

- Si despejamos esta ecuación a cero y despejamos $x$, obtenemos la formula para el siguiente punto de iteración:

\begin{equation}
\ x_{n+1} = x_n - \frac{f(x_n) (x_n - x_{n-1})}{f(x_n) - f(x_{n-1})}
\end{equation}

### **4. Pasos para aplicar el método de punto fijo.**

El método del punto fijo es un método iterativo que permite resolver
ecuaciones no lineales de la forma $f (x) = 0$, El método consiste en los
siguientes pasos:

1. Se debe definir una función auxiliar $g(x) = x − f (x)$
2. Se debe evaluar el punto $x0$ en la función auxiliar, y luego se debe definir $g(x0) = x0 − f (x0) = x0$
3. Luego, se define el punto $p0$, que es un punto que se aproxima al cero de la función y se debe evaluar en la función auxiliar $g(p0)$
4. Calcular el intercepto con la linea recta (recta identidad, $ y = x$), $p1 = g(p0) ,(g(p0), g(p0))$
5. Ir al punto $p1 = g(p0) (g(p0))$, $g(p0)$, el siguiente punto es $p1$, $(p1, g(p1))$
6. Comparar $p1$ con $g(p1)$, si la condiciones no se satisface, repetir
desde el paso 2, pero no con $p1$

- La raíz de $f(x)$ se encuentra en el mismo valor de $x$ donde ocurre la intersección entre la recta identidad y la función $g(x)$.

````python
def pos_x(t, vo, theta0, u, alpha, b, g  ):
  t1 = u*np.cos(alpha)*t  
  t2 = 1./b*(vo*np.cos(theta0)-u*np.cos(alpha))  
  t3 = (1-np.exp(-b*t))
  return t1 + t2 * t3


def pos_y(t, vo, theta0, u, alpha, b, g  ):
  t1 = (g/b + vo*np.sin(theta0) - u* np.sin(alpha) )
  t2 = (1-np.exp(-b*t)) 
  t3 = (g/b - u*np.sin(alpha))*t 
  return 1./b * t1 * t2 - t3
````

````python
t = np.linspace(0, 8.5, 100)
Params = (v0, theta, u, alpha, b , g)  #Parametros en tupla.
x = pos_x(t, *Params)
y = pos_y(t, *Params)
````

### **6.** Construir un fractal empleando el metodo de Newton-Raphson

````python
import numpy as np
import matplotlib.pylab as plt
from PIL import Image

Npx_X = 80 #Numero de pixeles en x
Npx_Y = 80 #Numero de pixeles en y

image = Image.new("RGB", (Npx_X, Npx_Y)) #Creacion de la imagen inicial

npx = np.arange(0, Npx_X)
npy = np.arange(0, Npx_Y)

xmin = -1
xmax = 1
ymin = -1
ymax = 1
# Transformación al plano complejo en un rectangulo (xmin, xmax), (ymin, ymax)
zx = npx * (xmax - xmin) / (Npx_X - 1)   + xmin
zy = npy * (ymax-ymin) / (Npx_Y - 1) + ymin

# Definicion  la funcion y las derivada 
def f(z): 
  return z**3 - 1

def fp(z): 
  return 3*z**2

h = 1E-4
def fpn(z, h): 
  return (f(z+h)-f(z))/h


a = 0.2
Niteraciones = 120
eps = 1E-8

x = - 1
for zx_ in zx:
  x = x + 1 
  y = -1
  for zy_ in zy: 
      y = y + 1
      Z_0 = complex(zx_, zy_)

      for i in range(0, Niteraciones): 
        Z = Z_0 - a * f(Z_0)/fp(Z_0)
        if( abs(Z-Z_0)<= eps):
          break           
        Z_0 = Z
      #print(i)
      #print(i)
      r, g, b = i % 4 *64 , i % 8 *16, i% 16* 16      
      image.putpixel((x,y), (r,g,b) )

plt.imshow(image)
# plt.xlim(400, 600)
# plt.ylim(400, 523)
````
