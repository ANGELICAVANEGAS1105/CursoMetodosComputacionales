## **Problema 2.**

````python
PS4 = ("https://docs.google.com/spreadsheets/d/e/2PACX-1vT4dd63iONuYN84UHbi1BUBb--kAwa-CcbYe6QnxiICFzH2fBszfd1mU_7ODyev4w/pub?output=xlsx")
````

````python
import pandas as pd
df = pd.read_excel(PS4)
df
````

````python
import pandas as pd
from datetime import datetime

df = pd.read_excel(PS4)
for d in df.Date:
  if(type(d)==str):
    df["Date_"] = datetime.strptime(d, "%d/%m/%Y")
  else:
    df["Date_"]=d

df=df.set_index("Date_") # Establece la columna de Date como índice
df=df.drop(columns=["Date","Unnamed: 1"]) # Elimina columnas no deseadas
df.reset_index(inplace=True) # agrega una columna numérica con el número de índice original de cada fila
df.rename(columns={'index':'# Fila'}, inplace=True) # renombra la columna de índice como '# Fila'
````

````python
# Agrega una columna que muestre el número de registros no NaN de cada columna
df['No. Registros'] = df.apply(lambda x: x.count(), axis=1)
````

````python
# Estadística básica
mean = df.mean()
std = df.std()
max = df.max()

# Agregar estadísticas al dataframe
df.loc['Promedios'] = mean
df.loc['Desviación estándar'] = std
df.loc['Máximo'] = max
````

````python
# Reemplazar los NaN por None
df = df.replace({pd.np.nan: None})

# Cantidad de memoria usada
memoria = df.memory_usage() 
df.loc['Memoria Usada'] = memoria

df
````

````python
# Comparación total de vacunados en Latinoamérica

df = df.loc[:, ['country','total_vaccinations']]
Filtro = df[df['country'].isin(['Argentina', 'Bolivia', 'Brasil', 'Chile', 'Colombia', 'Costa Rica', 'Cuba', 'Ecuador', 'El Salvador', 
                                'Guatemala', 'Haití', 'Honduras', 'México', 'Nicaragua', 'Panamá', 'Paraguay', 'Perú', 'Puerto Rico', 
                                'República Dominicana', 'Uruguay', 'Venezuela'])]

Filtro = Filtro.groupby('country').sum().reset_index()
Filtro
````

````python
import seaborn as sns
import matplotlib.pyplot as plt

x = Filtro['country']
y = Filtro['total_vaccinations']

plt.figure(figsize = (16,12))
ax = sns.barplot(x,y)
plt.title('Vacunados en Latinoamérica en 2021', fontsize = 18)
plt.xticks(rotation = 90, fontsize = 10)
plt.yticks(fontsize = 12)
plt.xlabel('Países de Latinoamérica', fontsize=14)
plt.ylabel('Total de vacunados', fontsize=14)
plt.grid()

plt.show()
````
