#### Método de Euler para ecuaciones diferenciales (Chat GPT)


El método de Euler es un método numérico utilizado para aproximar la solución de una ecuación diferencial ordinaria (EDO). Este método es uno de los más simples y se basa en la aproximación de la solución mediante la construcción de una secuencia de valores aproximados.

Supongamos que tenemos una EDO de primer orden de la forma:

y' = f(x,y)

donde y es la función desconocida de x y y' es su derivada. El método de Euler aproxima la solución y(x) de la siguiente manera:

1. Se elige un valor inicial para y(x0) = y0.

2. Se elige un paso h, que representa la distancia entre los valores de x en los que se calculará la solución aproximada.

3. Se utiliza la fórmula:
    y_{i+1} = y_i + h * f(x_i,y_i)

para calcular la aproximación de y en el siguiente valor x_{i+1} a partir de la aproximación anterior y_i en x_i.

4. Se repite el proceso para obtener una secuencia de aproximaciones de y(x) en los valores de x_i, con i = 0, 1, 2, ..., n.


### Código en Python
En Python, el método de Euler se puede implementar de la siguiente manera:

```python
def euler(f, x0, y0, h, n):
    # Inicializar los arreglos de x e y
    x = [x0]
    y = [y0]
    
    # Calcular las aproximaciones sucesivas de y
    for i in range(n):
        y.append(y[-1] + h * f(x[-1], y[-1]))
        x.append(x[-1] + h)
    
    # Devolver los arreglos de x e y
    return x, y
```

En este código, f es la función f(x,y) que define la EDO, x0 y y0 son los valores iniciales de x e y, h es el paso y n es el número de aproximaciones que se desean calcular.


````python
#Aplicando el metodo de Euler para resolver la ecuacion diferencial dN/dt = - lambda * N

#Definimos la funcion que vamos a resolver
lamb = 5

def f(t, N):
    return -lamb * N

#Definimos el metodo de Euler (Propuesto por Chat Gpt)
def euler(f, x0, y0, h, n):
    # Inicializar los arreglos de x e y
    x = [x0]
    y = [y0]
    
    # Calcular las aproximaciones sucesivas de y
    for i in range(n):
        y.append(y[-1] + h * f(x[-1], y[-1]))
        x.append(x[-1] + h)
    
    # Devolver los arreglos de x e y
    return x, y

#Definimos el metodo de Euler mejorado (Propuesto por Copilot)
def euler_mejorado(f, x0, y0, h, n):
    # Inicializar los arreglos de x e y
    x = [x0]
    y = [y0]
    
    # Calcular las aproximaciones sucesivas de y
    for i in range(n):
        y.append(y[-1] + h * f(x[-1] + h/2, y[-1] + h/2 * f(x[-1], y[-1])))
        x.append(x[-1] + h)
    
    # Devolver los arreglos de x e y
    return x, y

#Lo aplicamos para resolver la ecuacion diferencial

#Definimos los parametros
x0 = 0
y0 = 1
h = 0.1
n = 100

#Calculamos las soluciones
x, y = euler(f, x0, y0, h, n)
x2, y2 = euler_mejorado(f, x0, y0, h, n)

#Graficamos
plt.plot(x, y, label = "Euler")
plt.plot(x2, y2, label = "Euler mejorado")
plt.title(r'Solucion de la ecuacion diferencial $dN/dt = - \lambda N$')
plt.xlabel('t')
plt.ylabel('N')
plt.legend()
plt.grid()
plt.show()
````

m \frac{d^2x}{dt^2}+kx + bv=0

````python
#Resolver la siguiente ecuacion difirencial, empleando el metodo de Euler.

def oscilador(x0,v0,m,b,k,dt,num,tiempo):
  w= np.sqrt(k/m)
  X = np.zeros(num) #Valores en ceros para el numero de puntos
  V = np.zeros(num)
  X[0]= x0
  V[0]= v0
  t = np.linspace(0,tiempo,num)
  
  for i in range(0, len(t)-1):
   X[i+1]= X[i] + V[i]*dt
   V[i+1]= V[i]-(w**2)*X[i+1]*dt-b*V[i]
   
  return X,V

#Definimos los parametros
x0 = 0.01
v0 = 1
m = 2
b = 0.01
k = 2
dt = 0.1
num = 1000
tiempo = 100

#Calculamos las soluciones
X, V = oscilador(x0,v0,m,b,k,dt,num,tiempo)

#Graficamos
figure = plt.figure(figsize=(10, 5))
plt.plot(X, label = "Posicion")
plt.plot(V, label = "Velocidad")
plt.title(r'Solucion de la ecuacion diferencial $d^2x/dt^2 = - \omega^2 x - b dx/dt$')
plt.xlabel('t')
plt.ylabel('x,v')
plt.legend()
plt.grid()
plt.show()

````

````python
#Construir un codigo en python a traves de la IA, realizando preguntas adeacuadoas para para  dar la solución la ecuación diferencial, compare los  resultados.

## Resolver la ecuación diferencial del oscilador amortiguado con el método de Euler
# m d^2x/dt^2 + b*dx/dt + k*x = 0

# Definimos la función que vamos a resolver
m = 2
b = 0.2
k = 2

def f(t, x, v):
    return -b/m * v - k/m * x

# Definimos el método de Euler
def euler_oscilador(f, x0, v0, h, n):
    # Inicializar los arreglos de x, v y t
    x = [x0]
    v = [v0]
    t = [0]

    # Calcular las aproximaciones sucesivas de x, v y t
    for i in range(n):
        v.append(v[-1] + h * f(t[-1], x[-1], v[-1]))
        x.append(x[-1] + h * v[-1])
        t.append(t[-1] + h)

    # Devolver los arreglos de x, v y t
    return x, v, t

# Lo aplicamos para resolver la ecuación diferencial

# Definimos los parámetros
x0 = 0
v0 = 1
h = 0.1
n = 1000

# Calculamos las soluciones
x, v, t = euler_oscilador(f, x0, v0, h, n)

# Graficamos x vs t y v vs t
figure = plt.figure(figsize=(10, 5))
plt.plot(t, x, label="x vs t")
plt.plot(t, v, label="v vs t")
plt.title(r'Solución de la ecuación diferencial $m d^2x/dt^2 + b*dx/dt + k*x = 0$')
plt.xlabel('t')
plt.ylabel('x, v')
plt.legend()
plt.grid()
plt.show()
````
