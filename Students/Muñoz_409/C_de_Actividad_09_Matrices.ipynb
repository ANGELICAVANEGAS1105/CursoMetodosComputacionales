Operaciones matricial con regresión lineal.

Cuando se desea hacer una regresion lineal multivariada, existen diferentes formas de realizar la regresión, una de ellas es encontrar el error cuadrático medio y miniminzar el conjunto de parámetros de la regresión emplando algún metodo de minimización, otra forma es encontrar la expresión analítica en forma matricial para enoncontrar el conjunto de parametros de la regresión. En este último caso, la regresion lineal viene dada por la siguiente expresion generalizada:


- $\Theta = (X^T X)^{-1} X^T y$

donte $\Theta$ es la matriz de parametros asociados a la regresión lineal, X la matriz de características y y los valos esperados, así, si tenemos una conjunto de datos en N dimensiones la recta, plano o hyperplano puede ser calculado como:

$Y=\Theta^T X$


En el caso de dos dimensiones, tenemos lo siguiente, Supongamos que X, y Y presentan un comportamiento lineal, y vienen dados por:


- Sea $\vec{X}=x^1, x^2, x^3, x^4, ..., x^m$ 
- Sea $\vec{Y}=y^1, y^2, y^3, y^4, ..., y^m$ 

Para determinar la ecuacion de la recta, hay que conocer los párametros $\theta_0 $, $\theta_1$, que cumplen que :   $y^{i} = \theta_1 x^{i} + \theta_0$ la anterior expresion puede ser escrita como:

\begin{equation*}
\begin{bmatrix}
y^{1}\\
y^{2}\\
\vdots\\
y^{m}\\
\end{bmatrix}=
\begin{bmatrix}
\theta_0 & \theta_1 \\
\end{bmatrix}
\begin{bmatrix}
1 & 1 & \cdots & 1 \\
x_1^{1} & x_1^{2} & \cdots & x_1^{m} \\
\end{bmatrix}
\end{equation*}


Para determinar la matriz de theta entonces se puede aplicar la ecuación:

- $\Theta = (X^T X)^{-1} X^T y$

donde: 
\begin{equation*}
X=
\begin{bmatrix}
1 & x_1^{1} \\
1 & x_1^{2} \\
\vdots & \vdots \\
1 & x_1^{m} 
\end{bmatrix}
\end{equation*}

y 


\begin{equation*}
y=
\begin{bmatrix}
y^1 \\
y^2 \\
\vdots\\
y^m
\end{bmatrix}
\end{equation*}

````python
N = 10
x1 = np.linspace(0, 10, N)
noise=2*np.random.random(N)-1
y = 0.9*x1 + noise
````

````python
#Puntos 1d a interpolar
plt.plot(x1, y, 'o')
plt.xlabel('x1')
plt.ylabel('y')
plt.grid()
plt.show()
````

````python
ones = np.ones(N)

X1D = np.array([ones, x1]).T
X1D
````

````python
theta1D = np.linalg.inv(X1D.T @ X1D) @ X1D.T @ y.T
theta1D
````

````python
#Graficamos la interpolacion lineal 1D

plt.plot(x1, y, 'o', label='Datos')
plt.plot(x1, theta1D[0] + theta1D[1]*x1, 'r', label='Interpolacion')
plt.xlabel('x1')
plt.ylabel('y')
plt.legend()
plt.grid()
plt.show()
````

````python
#Generalice para el siguiente problema bidimensional:
N = 1000
noise1=2*np.random.random(N)-1
noise2=2*np.random.random(N)-1
x1 = np.linspace(-1, 1, N) + noise1
x2 = np.linspace(-2, 2, N) + noise2
y = 1.5* x1 + 2.5* x2 + 0.0 #+ 4*np.random.random(100)
#y = 2*(x1+noise1) - 3*(x2+noise2) + 0.0 #+ 4*np.random.random(100)
````

````python
ones=np.ones(N)
````

````python
X2D = np.array([ones, x1,x2]).T
X2D[0:15]
````

````python
theta2D = np.linalg.inv(X2D.T @ X2D) @ (X2D.T @ y.T)
theta2D
````

````python
N = 10
X1, X2 = np.meshgrid(x1,x2)
Y = 1.5*X1 + 2.5*X2 + theta2D[0]
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
ax.scatter(x1, x2, y, c='k', marker='o')
surf = ax.plot_surface(X1, X2, Y,alpha=0.09)
ax.view_init(10, 0)
````

````python
#Definimos nuevos intervalos
X_1 = np.linspace(-5,5,100)
X_2 = np.linspace(-5,5,100)

X1, X2 = np.meshgrid(X_1,X_2)
Y = 1.5*X1 + 2.5*X2 + theta2D[0]
````

````python
#La misma grafica anterior pero con diferentes perspectivas
fig,ax = plt.subplots(nrows=1, ncols=3,figsize = (15,10),subplot_kw={"projection": "3d"})

ax[1].scatter(x1, x2, y, c='y', marker='o')
surf = ax[1].plot_surface(X1, X2, Y,alpha=0.5)
ax[1].set_title('XY')
ax[1].view_init(90, -90)

ax[2].scatter(x1, x2, y, c='g', marker='o')
surf = ax[2].plot_surface(X1, X2, Y,alpha=0.5)
ax[2].set_title('XZ')
ax[2].view_init(0, -90)

ax[0].scatter(x1, x2, y, c='r', marker='o')
surf = ax[0].plot_surface(X1, X2, Y,alpha=0.5)
ax[0].set_title('YZ')
ax[0].view_init(10, 0)
````
