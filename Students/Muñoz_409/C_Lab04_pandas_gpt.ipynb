### Problema 4

1. Leer el dataset country_vaccinations.csv  y filtrar los datos para Colombia
2. Crear un csv con los datos para colombia
3. Realizar una comparacion con los paises latinoamericanos de la cantidad de vacunados.
4. A traves de mascaras determinar el numero de vacunados en el intervalo  [1.5E6, 2.0E6]

````python
data4 = 'https://raw.githubusercontent.com/hernansalinas/autogrades/main/Laboratorios_Taller/dataset/country_vaccinations.csv'
````

````python
#Leer el dataset country_vaccinations.csv  
pf4 = pd.read_csv(data4)
pf4.head()
````

````python
pf4.index = pd.to_datetime(pf4.date,format='%Y-%m-%d') #Cambiar el formato de la fecha
del(pf4['date'])                                       #Eliminar columnas Unnamed: 1 y Date
pf4.head()
````

````python
#Filtra los datos para Colombia
pf4[pf4['country'] == 'Colombia'].head()
````

````python
#Crear un csv con los datos para colombia
data_colombia = pf4[pf4['country'] == 'Colombia']
data_colombia.to_csv('data_Colombia.csv')#, index=False)
pd.read_csv('data_Colombia.csv',index_col='date').head() #Leer el csv creado con los datos para Colombia
````

````python
#Conjunto de paises latinoamericanos 
paises_latinoamericanos = {"México", "Brasil", "Argentina", 
                           "Colombia", "Perú", "Chile", "Venezuela", 
                           "Ecuador", "Bolivia", "Paraguay", "Uruguay",
                           "Costa Rica", "Panamá", "Nicaragua", "Honduras", 
                           "El Salvador", "Guatemala", "Cuba", "República Dominicana"}

countrys = set(pf4['country'])
countrys_lati = tuple(countrys & paises_latinoamericanos)

print('')
print('Paises latinoamericanos que estan en el dataset:')
print(countrys_lati)
````

````python
#Flitrar los datos para los paises latinoamericanos
paises_l = pf4[pf4['country'].isin(countrys_lati)]
paises_l.head()
````

````python
#separamos el dataframe en dos partes, una con los datos de los paises y otra con los datos de las vacunas
data_paises = paises_l[['country','total_vaccinations']]
#convertir los tados de los paises en categoricos
data_paises = pd.get_dummies(data_paises, columns=['country'])

#convertimos los NaN en 0
data_paises = data_paises.fillna(0)
data_paises['total_vaccinations'] = data_paises['total_vaccinations'].astype(int)
data_paises
````

````python
#multiplicamos los datos de los paises por la cantidad de vacunados
for i in data_paises.columns:
    if i != 'total_vaccinations':
        data_paises[i] = data_paises[i] * data_paises['total_vaccinations']
data_paises = data_paises.drop('total_vaccinations', axis=1) # Eliminamos la columna total_vaccinations
````

````python
#Construimos en el mismo grafico la cantidad de vacunados por pais con sus respestivas legendas

plt.figure(figsize=(20,10))
for i , j in enumerate(data_paises.columns):
    plt.scatter(data_paises.index, data_paises[j],label=j)
    
plt.legend()
plt.grid()
plt.title('Cantidad de vacunados por pais',fontsize=20)
plt.xlabel('Fecha',fontsize=20)
plt.ylabel('Cantidad de vacunados',fontsize=20)
plt.show()
````

````python
#A traves de mascaras determinar el numero de vacunados en el intervalo  [1.5E6, 2.0E6]
total_vacunados = pf4[(pf4['total_vaccinations'] > 1.5E6) & (pf4['total_vaccinations'] < 2.0E6)]
total_vacunados.head()
````
