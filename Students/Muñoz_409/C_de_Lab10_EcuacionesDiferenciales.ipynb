# Oscilacones Acopladas

Considere dos osciladores acoplados, de masas idénticas y  conectadas mediante tres resortes de constantes elásticas $k_1, k_2, k_3$. Como se muestra en la figura.





```python
import numpy as np
import matplotlib.pylab as plt
from scipy.integrate import odeint
```




Después de realizar un análisis detallado del problema, a través de las leyes de Newton ó empleando las ecuaciones de Euler-Lagrange, se puede mostrar que las ecuaciones de movimiento para el sistema vienen dadas por: 

\begin{equation}
m_1 \frac{ \mathrm{d}^2 x_1 }{\mathrm{d}t^2} = -k_1 x_1 + k_2 (x_2-x_1)
\end{equation}



\begin{equation}
m_2 \frac{ \mathrm{d}^2 x_2 }{\mathrm{d}t^2} = -k_3 x_2 - k_2 (x_2-x_1)
\end{equation}

De forma compacta,  la ecuacion anterior puede ser escrita como: 


\begin{equation}
m_1 \ddot{x_1} = -k_1 x_1 + k_2 (x_2-x_1)
\end{equation}



\begin{equation}
m_2  \ddot{x_2} = -k_3 x_2 - k_2 (x_2-x_1)
\end{equation}

donde $\ddot{x} =\frac{\mathrm{d^2}x}{\mathrm{d}t^2}$.

Las ecuaciones anteriores, son ecuaciones diferenciales lineales de segundo orden y acopladas, que pueden ser escritas de forma matricial como $m\vec{\ddot{x}}= -A\vec{x}$, donde A, es una matriz simétrica: 

\begin{equation}
m
\begin{pmatrix} 
\ddot{x_1}  \\
\ddot{x_2}    
\end{pmatrix}
=-
\begin{pmatrix} 
k_1 + k_2 & -k_2  \\
-k_2       &  k_2+k_3  \\
\end{pmatrix}
\begin{pmatrix} 
x_1\\
x_2
\end{pmatrix}
\end{equation}


Para resolver el sistema de ecuaciones acopladas, se puede realizar una transformación, que permite desacoplar el sistema de ecuaciones diferenciales. Sea
 $\vec{\eta}=S\vec{x}$, donde $\eta$ es una nueva variable en el sistema transformado y S una matriz que cumple que $SAS^{-1}=A_d$, siendo $A_d$ una matriz diagonal. 
 
 Se cumple que $\vec{\ddot {\eta}}=S\vec{\ddot{x}}$ 
 y  $m\vec{\ddot{x}}=A\vec{x}$, multiplicando por una matriz S, a ambos lados de la ecuación $m\vec{\ddot{x}}=-A\vec{x}$ obtenemos: 

\begin{equation}
mS\vec{\ddot{x}}=- S A\vec{x}
\end{equation}

\begin{equation}
m \vec{\ddot{\eta}} = -S A S^{-1}S\vec{x}
\end{equation}


\begin{equation}
m \vec{\ddot{\eta}} = -(S A S^{-1})\eta
\end{equation}


\begin{equation}
m \vec{\ddot{\eta}} = -A_d\eta
\end{equation}

\begin{equation}
m\begin{pmatrix} 
\ddot{\eta_1}\\
\ddot{\eta_2}
\end{pmatrix}
=
-\begin{pmatrix} 
\lambda_1 & 0\\
0          &\lambda_2
\end{pmatrix}
\begin{pmatrix} 
\eta_1\\
\eta_2
\end{pmatrix},
\end{equation}

donde $\lambda=m\omega^2$ son los autovalores de A. Notese que S es la matriz formada por los autovectores de A. La ecuacion anterior, tiene solución en términos de funciones armónicas.


Basado en lo anterior, y resolviendo  el problema más sencillo en el que $k_1 = k_3 = k$, tenemos que:


\begin{equation}
\begin{pmatrix} 
\frac{k + k_2}{m}-\omega^2 & -\frac{k_2}{m}\\
 -\frac{k_2}{m}          &\frac{k_2 + k}{m}-\omega^2
\end{pmatrix}
\begin{pmatrix} 
x_1\\
x_2
\end{pmatrix}
=
-\begin{pmatrix} 
0\\
0
\end{pmatrix}
\end{equation}


\begin{equation}
det(A-\omega^2 I)= \left(\frac{k + k_2}{m}-\omega^2\right) \left(\frac{k_2+k}{m}-\omega^2\right )-\frac{k_2^2}{m^2}=0
\end{equation}

Resolviendo para $\omega^2$, tenemos que : 

\begin{equation}
\omega^2_1=\frac{k+2k_2}{m}
\end{equation}


\begin{equation}
\omega^2_2=\frac{k}{m}, 
\end{equation}


los autovecres vienen dados por: 

\begin{equation}
\vec{v_1}=\frac{1}{\sqrt{2}}
\begin{pmatrix}
1\\
-1
\end{pmatrix}
\end{equation}


\begin{equation}
\vec{v_2}=\frac{1}{\sqrt{2}}
\begin{pmatrix}
1\\
1
\end{pmatrix}
\end{equation}

y la matriz S, viene dada por: 


\begin{equation}
S=\frac{1}{\sqrt{2}}
\begin{pmatrix}
1 & -1\\
1 & 1
\end{pmatrix}
\end{equation}

las variables $x_1,x_2$, en terminos de $\eta$ están dadas por: 

\begin{equation}
x_1=\frac{1}{\sqrt{2}}(\eta_1+\eta_2)
\end{equation}


\begin{equation}
x_2=\frac{1}{\sqrt{2}}(\eta_2+\eta_1)
\end{equation}

En conclusión el problema puede ser resuelto, si solucionamos la 
ecuaciones diferenciables desacopladas: 
\begin{equation}
\ddot{\eta_1}+\omega_1^2\eta_1=0
\end{equation}

\begin{equation}
\ddot{\eta_2}+\omega_2^2\eta_2=0
\end{equation}

 con: 
 
 \begin{equation}
\omega^2_1=\frac{k+2k_2}{m}
\end{equation}


\begin{equation}
\omega^2_2=\frac{k}{m}
\end{equation}

```python
#Funcion que me encuentra los valores propios de la matriz A
def valores_propios(k,k2,m):
    A = np.array([[(k+k2)/m,-k2/m],[-k2/m,(k+k2)/m]])
    w , v = np.linalg.eig(A)
    return np.sqrt(w)


#Funciones de los osciladores con frecuencia angular w1 y w2 asociadas a los valores propios de la matriz A
 
def oscillator_1(y, t, k, k2, m):
    w1 = valores_propios(k,k2,m)[0]
    return [y[1], -w1**2*y[0]]

def oscillator_2(y, t, k, k2, m):
    w2 = valores_propios(k,k2,m)[1]
    return [y[1], -w2**2*y[0]]
```

```python
#Solucion analitica
def x(t,k,k2,m,x1_0,v1_0,x2_0,v2_0):
    
    w1 = np.sqrt((k+k2)/m)
    w2 = np.sqrt(k/m)
    
    parte1 = (x1_0-x2_0)*np.cos(w1*t)/2 + (v1_0-v2_0)*np.sin(w1*t)/2*w1
    parte2 = (x1_0+x2_0)*np.cos(w2*t)/2 + (v1_0+v2_0)*np.sin(w2*t)/2*w2
    
    x1 = parte1 + parte2
    x2 = parte2 - parte1
    
    return [x1, x2]
```

```python
# creamos una funcion que nos permita graficar la solucion analitica y la solucion numerica apartir de los parametros iniciales

def plot_oscillator(parms_x1_0,parms_x2_0,k,k2,m):
    t = np.linspace(0, 10, 1000)
    
    #Los parametros iniciales de eta_1 y eta_2 se calculan a partir de los parametros iniciales de x1 y x2
    parms_eta_1_0 = np.array([(parms_x1_0[0]-parms_x2_0[0])/np.sqrt(2), (parms_x1_0[1]-parms_x2_0[1])/np.sqrt(2)])
    parms_eta_2_0 = np.array([(parms_x2_0[0]+parms_x1_0[0])/np.sqrt(2), (parms_x2_0[1]+parms_x1_0[1])/np.sqrt(2)])
    
    eta_1 = odeint(oscillator_1, parms_eta_1_0, t, args=(k,k2,m))[:,0]
    eta_2 = odeint(oscillator_2, parms_eta_2_0, t, args=(k,k2,m))[:,0]
    
    # Solucion numerica
    x1_ = (eta_1 + eta_2)/np.sqrt(2)
    x2_ = (eta_2 - eta_1)/np.sqrt(2) 
    
    fig, axs = plt.subplots(2, 1, figsize=(10, 8))
    fig.subplots_adjust(hspace=0.4)
    axs[0].plot(t, x1_, 'k', label='x1_(t)')
    axs[0].plot(t, x2_, 'b', label='x2_(t)')
    axs[0].set_xlabel('t')
    axs[0].set_ylabel('x(t)')
    axs[0].set_title('Solucion numerica')
    axs[0].legend(loc='best')
    axs[0].grid()
    
    axs[1].plot(t, x(t,k,k2,m,parms_x1_0[0],parms_x1_0[1],parms_x2_0[0],parms_x2_0[1])[0], 'k', label='x1(t)')
    axs[1].plot(t, x(t,k,k2,m,parms_x1_0[0],parms_x1_0[1],parms_x2_0[0],parms_x2_0[1])[1], 'b', label='x2(t)')
    axs[1].set_xlabel('t')
    axs[1].set_ylabel('x(t)')
    axs[1].set_title('Solucion analitica')
    axs[1].legend(loc='best')
    axs[1].grid()
```

**RETO 1** Considere ahora, un sistema con amortiguamiento proporcional a la velocidad, las ecuaciones de movimiento, 
vienen dadas por:

\begin{equation}
m\ddot{y_1} + \gamma \dot{y_1} + (k +k_2) y_1 - k_2 y_2 = 0
\end{equation}


\begin{equation}
m\ddot{y_2} + \gamma \dot{y_2} + (k +k_2) y_2 - k_2 y_1 = 0
\end{equation}

Estudiar en este caso, la evolucion temporal de cada masa para diferentes condiciones.

```python
# Definimos el sistema de ecuaciones diferenciales a resolver con odeint

def sistema_ecuaciones(y, t, m, gamma, k, k2):
    y1, y2, dy1, dy2 = y
    
    dy1_dt = dy1
    dy2_dt = dy2
    
    ddy1_dt = (-gamma * dy1 - (k + k2) * y1 + k2 * y2) / m
    ddy2_dt = (-gamma * dy2 - (k + k2) * y2 + k2 * y1) / m
    
    return [dy1_dt, dy2_dt, ddy1_dt, ddy2_dt]
```

```python
# Parámetros del sistema modo antisimetrico

A = 1.0         # amplitud
m = 1.0         # masa
gamma = 0.5     # coeficiente de amortiguamiento
k = 2.0         # constante del resorte k
k2 = 1.0        # constante del resorte k2
y1_0 = A        # posición inicial de y1
y2_0 = -A       # posición inicial de y2
dy1_dt_0 = 0.0  # velocidad inicial de y1
dy2_dt_0 = 1.0  # velocidad inicial de y2

# Vector de condiciones iniciales
params = [y1_0, y2_0, dy1_dt_0, dy2_dt_0]

# Tiempo de integración
t = np.linspace(0, 20, 500)  # desde 0 hasta 10 segundos con 100 puntos

# Resolver el sistema de ecuaciones diferenciales
solucion = odeint(sistema_ecuaciones, params, t, args=(m, gamma, k, k2))

# Extraer las soluciones
y1 = solucion[:, 0]     
y2 = solucion[:, 1]     
dy1_dt = solucion[:, 2]  
dy2_dt = solucion[:, 3]  

# Graficar las soluciones
fig , ax = plt.subplots(1,2,figsize=(15, 5))
ax[0].plot(t, y1, label='y1(t)', color='k')
ax[0].plot(t, y2, label='y2(t)', color='r')
ax[0].set_xlabel('t')
ax[0].set_ylabel('y(t)')
ax[0].legend(loc='best')
ax[0].title.set_text('Posiciones')
ax[0].grid()

ax[1].plot(t, dy1_dt, label='dy1/dt(t)', color='k')
ax[1].plot(t, dy2_dt, label='dy2/dt(t)', color='r')
ax[1].set_xlabel('t')
ax[1].set_ylabel('dy/dt(t)')
ax[1].legend(loc='best')
ax[1].title.set_text('Velocidades')
ax[1].grid()

plt.tight_layout()
plt.show()
```
