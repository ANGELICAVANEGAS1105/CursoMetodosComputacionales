### $ 7 = a\sqrt{x^2+1}+be^x\sin x $ 

```python
#Definiendo la función
def f(x:float,a:float,b:float)->float:
  return a*np.sqrt(x**2+1) + b*np.exp(x)*np.sin(x) - 7

#Definimos los argumentos de la función
x = np.linspace(2,10,1000)
arg1 = (1,1)
arg2 = (2,2)
```

```python
#@title visualización de la función para los argumentos dados
fig , axs = plt.subplots(1,2,figsize=(15,5))
ax = axs.flatten()

ax[0].plot(x,f(x,*arg1))
ax[0].set_title('a=1,b=1')
ax[0].set_xlabel('x')
ax[0].set_ylabel('f(x)')
ax[0].set_xlim(2,10)
ax[0].set_ylim(-1000,5000)
ax[0].grid()

ax[1].plot(x,f(x,*arg2))
ax[1].set_title('a=2,b=2')
ax[1].set_xlabel('x')
ax[1].set_ylabel('f(x)')
ax[1].set_xlim(2,10)
ax[1].set_ylim(-1000,10000)
ax[1].grid()


fig.subplots_adjust(wspace=0.3)
```

```python
#metodo de bisección para argumentos (1,1)
root1 = optimize.bisect(f,5,7,args=arg1)
root2 = optimize.bisect(f,9,10,args=arg1)
print(f'las raices de la función con a=1 y b=1 son: {root1:0.3f} y {root2:0.3f}')

#metodo de bisección para argumentos (2,2)
rot1 = optimize.bisect(f,5,7,args=arg2)
rot2 = optimize.bisect(f,9,10,args=arg2)
print(f'las raices de la función con a=2 y b=2 son: {rot1:0.3f} y {rot2:0.3f}')
```

```python
#@title visualización de la función con sus raices
fig , axs = plt.subplots(1,2,figsize=(15,5))
ax = axs.flatten()

ax[0].plot(x,f(x,*arg1))
ax[0].plot([root1,root2],[f(root1,*arg1),f(root2,*arg1)],'ro')
ax[0].set_title('a=1,b=1')
ax[0].set_xlabel('x')
ax[0].set_ylabel('f(x)')
ax[0].set_xlim(2,10)
ax[0].set_ylim(-1000,5000)
ax[0].grid()

ax[1].plot(x,f(x,*arg2))
ax[1].plot([rot1,rot2],[f(rot1,*arg2),f(rot2,*arg2)],'ro')
ax[1].set_title('a=2,b=2')
ax[1].set_xlabel('x')
ax[1].set_ylabel('f(x)')
ax[1].set_xlim(2,10)
ax[1].set_ylim(-1000,10000)
ax[1].grid()


fig.subplots_adjust(wspace=0.3)
```

  
### $a e^{x}-2= b\cos(e^x-2)$

```python
#Definiendo la función
def g(x,a,b):
  return b*np.cos(np.exp(x)-2) + 2 - a*np.exp(x)

#Definimos los argumentos de la función
x = np.linspace(0,2.5,1000)
arg1 = (1,1)
arg2 = (2,2)
```

```python
#@title visualización de la función para los argumentos dados
fig , axs = plt.subplots(1,2,figsize=(15,5))
ax = axs.flatten()

ax[0].plot(x,g(x,*arg1))
ax[0].set_title('a=1,b=1')
ax[0].set_xlabel('x')
ax[0].set_ylabel('g(x)')
ax[0].set_xlim(0,2.5)
ax[0].set_ylim(-10,2)
ax[0].grid()

ax[1].plot(x,g(x,*arg2))
ax[1].set_title('a=2,b=2')
ax[1].set_xlabel('x')
ax[1].set_ylabel('g(x)')
ax[1].set_xlim(0,2.5)
ax[1].set_ylim(-10,2)
ax[1].grid()


fig.subplots_adjust(wspace=0.3)
```

```python
#metodo de bisección para argumentos (1,1)
root1 = optimize.bisect(g,0,1.5,args=arg1)
print(f'las raices de la función con a=1 y b=1 son: {root1:0.3f}')

#metodo de bisección para argumentos (2,2)
rot1 = optimize.bisect(g,0,1.5,args=arg2)
print(f'las raices de la función con a=2 y b=2 son: {rot1:0.3f}')
```

```python
#@title visualización de la función para los argumentos dados
fig , axs = plt.subplots(1,2,figsize=(15,5))
ax = axs.flatten()

ax[0].plot(x,g(x,*arg1))
ax[0].plot(root1,g(root1,*arg1),'ro')
ax[0].set_title('a=1,b=1')
ax[0].set_xlabel('x')
ax[0].set_ylabel('g(x)')
ax[0].set_xlim(0,2.5)
ax[0].set_ylim(-10,2)
ax[0].grid()

ax[1].plot(x,g(x,*arg2))
ax[1].plot(rot1,g(rot1,*arg2),'ro')
ax[1].set_title('a=2,b=2')
ax[1].set_xlabel('x')
ax[1].set_ylabel('g(x)')
ax[1].set_xlim(0,2.5)
ax[1].set_ylim(-10,2)
ax[1].grid()


fig.subplots_adjust(wspace=0.3)
```

3. Fixed point methods. 

When a new planet is discovered, there are different methods to estimate its physical properties. Many times is only possible to estimate either the planet mass or the planet radius and the other property has to be predicted through computer modelling.

If one has the planet mass, a very rough way to estimate its radius is to assume certain composition (mean density) and a homogeneous distribution (a very bad assumption!). For example, for the planet [Gliese 832c](http://es.wikipedia.org/wiki/Gliese_832_c) with a mass $M= 5.40 M_{\oplus}$, if we assume an earth-like composition, i.e. $\bar \rho_{\oplus} = 5520\ kg/m^3$, we obtain:

$$R_{g832c} = \left( \frac{3 M_{g832c}}{ 4 \pi \bar\rho_{\oplus} } \right)^{1/3} \approx 1.75 R_{\oplus}$$

That would be the planet radius if the composition where exactly equal to earth's.

A more realistic approach is assuming an internal one-layer density profile like:

$$\rho(r) = \rho_0 \exp\left( -\frac{r}{L} \right)$$

where $\rho_0$ is the density at planet centre and $L$ is a characteristic lenght depending on the composition. From numerical models of planet interiors, the estimated parameters for a planet of are $M= 5.40 M_{\oplus}$ are approximately $\rho_0 = 18000\ kg/m^3$ and $L = 6500\ km$.

Integrating over the planet volume, we obtain the total mass as

$$M = 4\pi \int_0^R \rho(r)r^2dr$$

This is a function of the mass in terms of the planet radius. 

Solving the equation $M(R) = M_{g832c}$ it would be possible to find a more realistic planet radius. However when using numerical models, it is not possible to approach the solution from the left side as a negative mass makes no sense.

```python
#Datos

Mt = 5.972e24       #Masa de la tierra en kg
Rt = 6.371e3          #radio de la tierra en m
M_g832c = 5.40*Mt   # masa del planeta g832c
R_g832c = 1.75*Rt   # radio del planeta g832c (Aproximación)
rho = 18000         # Kg/m^3
L = 6500e3           #m
rho_ = 5520         #kg/m^3
```

```python
R_g832c,M_g832c
```

```python
#Luego de resolver la integral, se obtiene la siguiente expresión para M(R)
def mass(R:float, L:float = 6500, rho:float = 18000) -> float:
    return 4*np.pi*rho*L*(2*L**2-(2*L**2 + 2*L*R + R**2)*np.exp(-R/L))
```

```python
R = np.linspace(0,16000e3,100000)
def F(R:float) -> float:
    return (5.40*Mt - mass(R,L,rho))/Mt
```

```python
plt.plot(R/1e3,F(R))
plt.xlabel('R (km)')
plt.ylabel('F(R)')
plt.grid()
```

```python
#Con el método de punto fijo, se obtiene la raíz de la función F(R) = R - (5.40*Mt - mass(R,L,rho))

root = optimize.fixed_point(F,1)
print(f'El radio del planeta g832c es {root} veces el radio de la tierra')
```
