Operaciones matricial con regresión lineal.

Cuando se desea hacer una regresion lineal multivariada, existen diferentes formas de realizar la regresión, una de ellas es encontrar el error cuadrático medio y miniminzar el conjunto de parámetros de la regresión emplando algún metodo de minimización, otra forma es encontrar la expresión analítica en forma matricial para enoncontrar el conjunto de parametros de la regresión. En este último caso, la regresion lineal viene dada por la siguiente expresion generalizada:


- $\Theta = (X^T X)^{-1} X^T y$

donte $\Theta$ es la matriz de parametros asociados a la regresión lineal, X la matriz de características y y los valos esperados, así, si tenemos una conjunto de datos en N dimensiones la recta, plano o hyperplano puede ser calculado como:

$Y=\Theta^T X$


En el caso de dos dimensiones, tenemos lo siguiente, Supongamos que X, y Y presentan un comportamiento lineal, y vienen dados por:


- Sea $\vec{X}=x^1, x^2, x^3, x^4, ..., x^m$ 
- Sea $\vec{Y}=y^1, y^2, y^3, y^4, ..., y^m$ 

Para determinar la ecuacion de la recta, hay que conocer los párametros $\theta_0 $, $\theta_1$, que cumplen que :   $y^{i} = \theta_1 x^{i} + \theta_0$ la anterior expresion puede ser escrita como:

\begin{equation}
\begin{bmatrix}
y^{1}\\
y^{2}\\
\vdots\\
y^{m}\\
\end{bmatrix}=
\begin{bmatrix}
\theta_0 & \theta_1 \\
\end{bmatrix}
\begin{bmatrix}
1 & 1 & \cdots & 1 \\
x_1^{1} & x_1^{2} & \cdots & x_1^{m} \\
\end{bmatrix}
\end{equation}


Para determinar la matriz de theta entonces se puede aplicar la ecuación:

- $\Theta = (X^T X)^{-1} X^T y$

donde: 
\begin{equation}
X=
\begin{bmatrix}
1 & x_1^{1} \\
1 & x_1^{2} \\
\vdots & \vdots \\
1 & x_1^{m} 
\end{bmatrix}
\end{equation}

y 


\begin{equation}
y=
\begin{bmatrix}
y^1 \\
y^2 \\
\vdots\\
y^m
\end{bmatrix}
\end{equation}

Su tarea será construir un programa que realice la regresión lineal para los siguientes puntos: 
```python 
N = 10
x1 = np.linspace(0, 10, N)
noise=2*np.random.random(N)-1
y = 0.9*x1 + noise
```

````python
import numpy as np
import matplotlib.pyplot as plt

N = 10
x1 = np.linspace(0, 10, N)
noise=2*np.random.random(N)-1
y = 0.9*x1 + noise
````

````python
plt.plot(x1, y, 'ro')
plt.xlabel('x1')
````

````python
unos = np.ones(10)
````

````python
#contruimos la matriz que contiene los unos y los elementos de las x

mat_x = np.array([unos, x1]).T
mat_x             #se pone de manera vertical desde un comienzo par que funcione la forma de los cálculos
````

````python
np.shape(mat_x)
````

````python
#construimos la matriz de los y

mat_y = y
y, np.shape(mat_y)
````

````python
#ahora, la idea es encontrar la matriz de theta

prod_x =  mat_x.T @ mat_x
prod_as_matr = np.asmatrix(prod_x)


#saco la inversa
inv = prod_as_matr.I

theta = inv @ mat_x.T @ mat_y
````

````python
#@title Obtenemos la matriz de theta que nos da el valor del intercepto y de la pendiente respectivamente

theta[0,1], theta       #--> la matriz y la forma de acceder a los elementos
````

````python
#@title Obtenemos la ecuación correspondiente a la regresion

ec = lambda x: theta[0,1] * x + theta[0,0]
````

````python
#@title Graficamos

plt.plot(x1, y, 'ro', label='datos')
plt.xlabel('x1')
plt.plot(x1, ec(x1), label ='regresión lineal')
plt.title('registro de datos y regresió  lineal')
plt.legend(loc= 'best')
plt.grid();
````

````python
#@title La idea es ahora genealizarlo para obtener el fit en un plano; generamos los datos sintéticos
N = 100

noise1=2*np.random.random(N)-1
noise2=2*np.random.random(N)-1

x1 = np.linspace(-1, 1, N)  + noise1
x2 = np.linspace(-2, 2, N) + noise2

y = 1*(x1) - 1*(x2) + 2.0 #+ 4*np.random.random(100) 
#ecuacion teórica
````

````python
unos = np.ones(N)
````

````python
#contruimos la matriz que contiene los unos y los elementos de las x

mat_x = np.array([unos, x1, x2]).T
#mat_x                                                                          #se pone de manera vertical desde un comienzo par que funcione la forma de los cálculos
````

````python
mat_y = y
````

````python
#ahora, la idea es encontrar la matriz de theta

prod_x =  mat_x.T @ mat_x
prod_as_matr = np.asmatrix(prod_x)

np.linalg.det(prod_as_matr)          
````

````python
#saco la inversa
inv = prod_as_matr.I

theta = inv @ mat_x.T @ mat_y
print(theta)
````

````python
#@title Obtenemos la ecuación correspondiente a la regresion

ec = lambda x1, x2: theta[0,2] * x2 + theta[0,1] * x1 + theta[0,0]
````

````python
#ahora, la idea es hacer el gráfico tridimensional

a= np.linspace(min(x1), max(x1), 100)
b= np.linspace(min(x2), max(x2), 100)

A,B = np.meshgrid(a,b)

fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
surf = ax.plot_surface(A, B, ec(A, B))  
scatter = ax.scatter(x1, x2, y,"-")
````
