El método de Euler es un método numérico utilizado para aproximar la solución de ecuaciones diferenciales ordinarias de primer orden. Este método se basa en la aproximación de la solución mediante la construcción de una aproximación lineal en cada paso del tiempo.

El método de Euler se puede implementar en un programa de computadora de la siguiente manera:

1. Se establece un valor inicial para la solución de la ecuación diferencial en el tiempo inicial, es decir, `t = t0`.

2. Se define el tamaño del incremento de tiempo, denotado por `h`.

3. Se divide el intervalo de tiempo `t0` a `tf` en pequeños incrementos de tiempo de tamaño `h`.

4. En cada paso del tiempo, se utiliza la derivada de la solución en el tiempo actual `t` y el valor actual de la solución `y` para construir una aproximación lineal de la solución en el siguiente paso del tiempo.

5. Calcular el valor de la solución en el siguiente paso del tiempo utilizando la aproximación lineal.

6. Repetir los pasos 4 y 5 para cada incremento de tiempo hasta llegar al tiempo final.

La fórmula para la aproximación lineal se define como:
$$
y(t + h) ≈ y(t) + h * f(y(t), t)
$$

donde `y(t)` es el valor actual de la solución, `f(y(t), t)` es la derivada de la solución en el tiempo actual y `h` es el tamaño del incremento de tiempo.

El método de Euler es fácil de implementar y es adecuado para ecuaciones diferenciales sencillas. Sin embargo, para ecuaciones diferenciales más complejas, el método puede requerir un tamaño de paso muy pequeño para obtener una solución precisa, lo que puede ser computacionalmente costoso.

A continuación, se muestra un ejemplo de implementación en Python del método de Euler para resolver la ecuación diferencial `y' = y - t^2 + 1` con las condiciones iniciales `y(0) = 0.5`, en el intervalo de tiempo `t = [0, 2]` con un tamaño de paso `h = 0.2`:

###Resolvamos el problema

````python
lambda_ = 2
def f(N, t):
  return -lambda_*N

t0 = 0
y0 = 1
tf = 10
h = 0.1

t = [t0]
y = [y0]
while t[-1] < tf:
  t.append(t[-1] + h)
  y.append(y[-1] + h * f(y[-1], t[-1]))

# Graficar la solución aproximada
import matplotlib.pyplot as plt
plt.plot(t, y)
plt.xlabel('t')
plt.ylabel('y(t)')
plt.show()  
````

m\frac{d^2x}{dt^2} + kx + bv = 0

2. Construir un codigo en python a traves de la IA, realizando preguntas adeacuadoas para para dar la solución la ecuación diferencial, compare los resultados.


Para resolver la ecuación diferencial dada mediante el método de Euler en Python, es necesario convertirla en dos ecuaciones diferenciales de primer orden. Para ello, se introduce una nueva variable $v = \frac{dx}{dt}$, lo que nos permite escribir la ecuación como:

$$
\frac{dx}{dt} = v
$$
$$
\frac{dv}{dt} = -\frac{k}{m}x - \frac{b}{m}v
$$

##Resuelto con chatgpt

````python
import numpy as np
import matplotlib.pyplot as plt

#constantes
m = 1.0 
k = 1.0   
b = 0.2   

#condiciones iniciales
t0 = 0.0         
tf = 10.0            
x0 = 1.0             
v0 = 0.0             
h = 0.01             

t = np.arange(t0, tf+h, h)  
x = np.zeros_like(t)         
v = np.zeros_like(t)    

x[0] = x0
v[0] = v0

for i in range(1, len(t)):
    #derivada de la posición y la velocidad en el tiempo actual
    dxdt = v[i-1]
    dvdt = -k/m*x[i-1] - b/m*v[i-1]
    
    #nueva posición y velocidad utilizando el método de Euler
    x[i] = x[i-1] + h*dxdt
    v[i] = v[i-1] + h*dvdt
plt.plot(t, x, label='Posición')
plt.plot(t, v, label='Velocidad')
plt.xlabel('Tiempo')
plt.legend();
````
