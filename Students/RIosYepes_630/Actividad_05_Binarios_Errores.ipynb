### Activity 1.0:

Implement a function that get the binary representation of an integer with name mybin, the algorithm should return
a string with 8 characters, the maximun number that represent the funcition is integer of 8 bit. The name of function is
mybin

Ejexample :

b=mybin(x)

b='00101000'

````python
##!/usr/bin/env python3

# Estudiante: Alejandro Rios Yepes. (CC1036966630)

# Definimos la función
def mybin(numero):
    # Inicializar una cadena vacía para guardar el resultado
    binario = ""
    # Repetir mientras el número sea mayor que 0
    while numero > 0:
        # Obtener el residuo de dividir el número entre 2
        residuo = numero % 2
        # Agregar el residuo al inicio de la cadena binaria
        binario = str(residuo) + binario
        # Actualizar el número dividiéndolo entre 2
        numero = numero // 2
    # Repetir mientras la longitud de la cadena binaria sea menor que 8
    while len(binario) < 8:
        # Agregar un cero al inicio de la cadena binaria
        binario = "0" + binario
    # Devolver la cadena binaria como resultado
    return binario

# Pedir el número al usuario como entrada y convertirlo a entero
numero = int(input("Ingresa un número entero: "))
# Mostrar el resultado llamando a la función mybin con el número ingresado
print(mybin(numero))
````

### Activity 2.0:

With the binary representation please try to implement the formula to recover the number.

Hint: Use as input the binary representation as a string and invert its order

```
number32(BIN):

```

Where BIN is a string with the binary number

Example:


Test your solution for the next number:

number32("00111110001000000000001000010000")

````python
def mydec(binario):
    # Inicializar una variable para guardar el resultado
    decimal = 0
    # Invertir el orden de la cadena binaria
    binario = binario[::-1]
    # Recorrer cada dígito binario desde la derecha
    for i in range(len(binario)):
        # Convertir el dígito binario a entero
        digito = int(binario[i])
        # Multiplicar el dígito por 2 elevado a la posición que ocupa
        producto = digito * (2 ** i)
        # Sumar el producto al resultado
        decimal = decimal + producto
    # Devolver el resultado
    return decimal

# Usar la función mybin para obtener la cadena binaria del número
binario = mybin(numero)
# Usar la función mydec para recuperar el número a partir de la cadena binaria
decimal = mydec(binario)
# Mostrar el resultado
print("El número decimal ingresado en el anterior código es:",decimal) # Debería ser igual al número ingresado
````

### Activity 3.0:
1. Write a python script that calculates the double precision number represented by a 64-bits binary.

the name of function should be:
```
number64(BIN):

```

2. What is the number represented by:

BIN="0 10000000011 1011100100001111111111111111111111111111111111111111"

````python
def number64(BIN):
  # Separar el binario en signo, exponente y mantisa
  signo = BIN[0]
  exponente = BIN[1:12]
  mantisa = BIN[12:]

  # Convertir el signo a -1 o 1
  if signo == "0":
    signo = 1
  else:
    signo = -1

  # Convertir el exponente a un número entero
  exponente = int(exponente, 2) - 1023

  # Convertir la mantisa a un número decimal
  mantisa = int(mantisa, 2) / (2 ** len(mantisa))

  # Calcular el número de doble precisión
  numero = signo * (1 + mantisa) * (2 ** exponente)

  # Devolver el número
  return numero

BIN="010000000011101110010000111111111111111111111111111111111111111111"
print("El número representado es:", number64(BIN))
````
