```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 100
noise=2*np.random.random(N)-1
x1 = np.linspace(0, 10, N)+noise
y = 0.9*(x1 + noise)+2
plt.plot(x1,y,"ro")
```

```python
col_1= np.ones((N,))
X= np.array([col_1,x1])
X_1= X.T
```

```python
theta= np.linalg.inv(X@X_1)@X@y
```
