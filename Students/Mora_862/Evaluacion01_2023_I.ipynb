## 1. (50%). Movimiento de una esfera de radio r en un fluido.

 Una esfera de radio $R$ que asciende o desciende por un fluido puede ser modelada a través de un modelo de partícula, sometida a tres fuerzas: el peso, el empuje y la fuerza rozamiento. Las  ecuaciones de posición (x) y velocidad (v) para un sistema de refencia úbicado en la base del fluido, pueden ser escritas como:

\begin{equation}
x = \frac{G}{\alpha} \left( 1+\frac{\alpha v_0}{G} \right)(1-e^{-\alpha t}) - \frac{G}{\alpha}t
\end{equation}

\begin{equation}
v= \left ( v_0 +\frac{G}{\alpha} \right )e^{-\alpha t} - \frac{G}{\alpha}
\end{equation}



donde $G=\left( 1-\frac{\rho_f}{\rho} \right)g$, con $\rho_f$, $\rho$ la densidad del fluido y la esfera respectivamente y $g=9.8$ la aceleración de la gravedad y $\alpha$  una constante que depende del radio de la esfera, la viscosidad y la densidad.
$\alpha=\frac{9\eta}{2R^2\rho}$.

Suponga que una esferita de radio $R=2\times10^{-5}m$ es lanzada verticalmente hacia arriba en el agua, con una rapidez de $v_0=0.01$m/s.
Tomando la densidad del agua en  $\rho_f =10000 kg/m^3$,  la densidad de la esfera en $\rho=2670 kg/m^3$ y la viscosidad en $\eta=0.001 kg/ms$.  Determine: 

1. El tiempo total en que la esfera asciende  y desciende. 

2. Determine la rapidez final, con la que llega nuevamente de donde fue lanzada. 



Para ello construya un programa en python que:
- Permita calcular la posición como función del tiempo
- Permita calcular la rapidez como función del tiempo
- Cálcule  el tiempo total de subida y bajada empleando el método de la secante.

 

\begin{equation}
x = \frac{G}{\alpha} \left( 1+\frac{\alpha v_0}{G} \right)(1-e^{-\alpha t}) - \frac{G}{\alpha}t
\end{equation}

```python
#la posición como función del tiempo (x vs t)
import numpy as np
import matplotlib.pyplot as plt
R=2e-5
G=(1-(1000/2670))*9.8
alf=9*0.001/(2*2670*R**2)

v0=0.01

def funcionx(t):
  x1= G/(alf)**2*(1+(alf*v0/G))*(1-np.exp(-1*alf*t))
  x2=t*G/alf
  return x1-x2

T = np.linspace(0,0.1,1000) #creo un linspace para un intevalo del tiempo. Con esto resolvemos x para cada instante de tiempo
```

```python
plt.plot(T, funcionx(T))
plt.xlabel('Tiempo')
plt.ylabel('x')
plt.xlim(0,0.002)
plt.ylim(-0.000002,0.000002)
plt.title('posición como función del tiempo')
plt.grid()
```

\begin{equation}
v= \left ( v_0 +\frac{G}{\alpha} \right )e^{-\alpha t} - \frac{G}{\alpha}
\end{equation}

```python
#Permita calcular la rapidez como función del tiempo

def funcionv(t):
  x3=v0+G/alf
  x4=np.exp(-1*alf*t)
  x5=G/alf
  return x3*x4-x5

Funv = [funcionv(t) for t in T]
```

```python
plt.plot(T, Funv)
plt.xlabel('Tiempo')
plt.ylabel('rapidez')
plt.xlim(0,0.002)
plt.title('rapidez como función del tiempo')
plt.grid()
```

```python
#Cálcule el tiempo total de subida y bajada empleando el método de la secante.
from scipy import optimize

#rapidez como función del tiempo

#Determine la rapidez final, con la que llega nuevamente de donde fue lanzada.

puntosecantev=optimize.newton(funcionv,0.0005)
print('la rapidez final, con la que llega nuevamente de donde fue lanzada es:', puntosecantev)
```

```python
#posición como función del tiempo
puntosecantex=optimize.newton(funcionx,0.002) #la grafica x vs t presenta dos ceros, uno de ellos es en la posición inicial (0,0)

#El tiempo total en que la esfera asciende y desciende.
print('El tiempo total en que la esfera asciende y desciende es:', puntosecantex)
```

```python
import pandas as pd
import numpy as np
```

```python
#Leer el dataset desde la url, con el comando pd.read_csv.
data = pd.read_csv('https://raw.githubusercontent.com/hernansalinas/autogrades/main/Evaluaciones/dataset/Temperature_Bogota_Barcelona.csv')
data
```

```python
#Adicionar una nueva columna AvgTempCelcius con la temperatura en grados celcius.  C=(F−32)/1.8
data['AvgTempCelcius'] = data['AvgTemperature'].apply(lambda F: (F-32)/1.8)
data
```

```python
#Realizar un gráfico de AvgTempCelcius como función del año para la ciudad de Bogotá.

#Para poder realizar este gráfico, necesitamos filtrar toda la información de Bogotá de 'data'

#datos Bog
databog = data.loc[data['City']=='Bogota']
databog
```

```python
import matplotlib.pyplot as plt
x=databog['Year']
y=databog['AvgTempCelcius']
```

```python
plt.plot(x,y, color='hotpink')
plt.xlim(1995,2020)
plt.xlabel('$Año$')
plt.ylabel('$Temperatura$ $[°C]$')
plt.title('Temperatura para la ciudad de Bogotá de 1995 hasta 2020')
```

```python
#Realizar un gráfico de AvgTempCelcius como función del año para la ciudad de Barcelona.

#Hacemos el mismo proceso pero para Barcelona:
databar = data.loc[data['City']=='Barcelona']

x2=databar['Year']
y2=databar['AvgTempCelcius']

plt.plot(x2,y2, color='violet')
plt.xlim(1995,2020)
plt.xlabel('$Año$')
plt.ylabel('$Temperatura$ $[°C]$')
plt.title('Temperatura para la ciudad de Barcelona de 1995 hasta 2020')
```

```python
#Aplique el metodos de interpolación de hermite para Bogotá y el metodo de interpolación de splines cubicos para Barcelona.

#BARCELONA
from scipy.interpolate import CubicSpline

cubic=CubicSpline(x2,y2)

plt.plot(x2, cubic(x2), 'o', color='coral')
plt.plot(x2,y2, color='violet')
plt.xlabel('$Año$')
plt.ylabel('$Temperatura$ $[°C]$')
plt.title('Temperatura para la ciudad de Barcelona de 1995 hasta 2020  y el metodo de interpolación de splines cubicos')

```
