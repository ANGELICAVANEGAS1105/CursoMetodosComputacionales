# 1. Movimiento de proyectiles en presencia de viento.

Supongamos que un proyectil es lanzado desde el origen con una rapidez $v=vo( \cos \theta , \sin \theta) $. Supongamos además que el efecto del viento sobre el proyectil es el de una fuerza proporcional a la velocidad relativa entre el viento y el proyectil. Teniendo presente que si $u$ es la velocidad del viento y $\alpha$ el ángulo que hace dirección de esta velocidad con el eje x, se puede entrontrar las siguientes ecuaciones diferenciales, asociada al estado de movimiento del [proyectil](https://iopscience.iop.org/article/10.1088/0143-0807/36/2/025016/meta):


\begin{equation}
m\frac{dv_x}{dt} = -mb(v_x-u\cos \alpha )
\end{equation}

\begin{equation}
m\frac{dv_y}{dt} = -mg-mb(v_x-u\sin \alpha ),
\end{equation}

donde b es el párametro de rozamiento. La soluciones a las ecuaciones anteriores vienen dadas por:

Posición en x
\begin{equation}
x=(u\cos \alpha) t + \frac{1}{b} (v_0\cos \theta_0 -u \cos\alpha)(1-\mathrm{exp}(-bt))
\end{equation}

Posicion en y 
\begin{equation}
y= \frac{1}{b}\left(\frac{g}{b}+v_0 \sin \theta_0 -u\sin\alpha \right)(1-\mathrm{exp}(-bt))-\left( \frac{g}{b}-u\sin\alpha\right)t
\end{equation}

````python
import numpy as np
import matplotlib.pylab as plt
import pandas as pd
from scipy.optimize import newton
from scipy.interpolate import lagrange
````

````python
vo = 60 #Velocidad en m/s
alpha = np.pi #Angulo del viento
theta = np.deg2rad(45) #Angulo de lanzamiento
u = vo/2             #rapidez del viento
b = 0.05             #Parametro de friccion 
g = 9.8                #Aceleracion de la gravedad m/s2

def pos_x(t, vo, theta0, u, alpha, b, g  ):
  t1 = u*np.cos(alpha)*t  
  t2 = 1./b*(vo*np.cos(theta0)-u*np.cos(alpha))  
  t3 = (1-np.exp(-b*t))
  return t1 + t2 * t3


def pos_y(t, vo, theta0, u, alpha, b, g  ):
  t1 = (g/b + vo*np.sin(theta0) - u* np.sin(alpha) )
  t2 = (1-np.exp(-b*t)) 
  t3 = (g/b - u*np.sin(alpha))*t 
  return 1./b * t1 * t2 - t3
````

````python
t = np.linspace(0, 8.5, 100)
Params = (vo, theta, u, alpha, b , g)  #Parametros en tupla.
x = pos_x(t, *Params)
y = pos_y(t, *Params)
````

````python
dicc = {"x":x,"y":y ,"t":t}
df = pd.DataFrame(dicc)
````

````python
plt.plot(df.x, df.y)
plt.xlabel("x")
plt.ylabel("y")
plt.grid()
````

````python
plt.plot(df.t, df.y)
plt.xlabel("t")
plt.ylabel("y")
plt.grid()
````

````python
f_s= lambda x: 1/(1+25*x**2)
x= np.linspace(-1,1,100)

N=4
n=N+1
x_= np.linspace(-1,1,n)
````

````python
P=lagrange(x_, f_s(x_))
xx= np.linspace(np.sin(x_), np.cos(x_))

````

````python
plt.figure()
plt.plot(x,f_s(x), 'o-')
plt.plot(x_,f_s(x_), 'ro')
plt.plot(xx, P(xx), 'k-')
````

````python
def interpolacion(N):
  n=N+1
  x_=np.linspace(-1,1,n)

  #interpolación
  poly=lagrange(x_,f_s(x_))

  #Valores de x con los que se grafica la interpolación
  x_p=np.linspace(np.min(x_),np.max(x_))
  return x_p,poly
````

````python
NN=[4,8,12,16,20,30]

i=0
fig,axs=plt.subplots(figsize=(12,8),nrows=2,ncols=3)

for n_ in NN:
  ax=axs.flatten()
  ax[i].plot(x,f_s(x),'o-')
  ax[i].plot(x_,f_s(x_),'ro')
  N=n_
  xx,P=interpolacion(N)
  ax[i].plot(xx,P(xx),"k")
  ax[i].set_xlabel('x')
  ax[i].set_xlabel('f(x)')
  i=i+1

fig.subplots_adjust(wspace=0.3)
fig.subplots_adjust(hspace=0.2)
````

\begin{equation}

\end{equation}

````python
#La magnetización reducida empleando el metodo de la secante(Muestre su resultado con 4 cifras)
import math
from scipy import optimize

def f(m,T):
  return np.tanh(m/T)-m

root = optimize.bisect(f, -1, 2, args=(0.5), xtol=2e-12)
print('el primer cero se encuentra en:', round(root,4))

````

````python
#d. Defina T = np.linspace(0.1,0.99,100), y emple el metodo de la secante para construir
# un gráfico de magnetización reducida como función de la temperatura para los 100 valores de T.
````

````python
T = np.linspace(0.1,0.99,100)
````

````python
import numpy as np
import scipy.optimize as optimize
import matplotlib.pyplot as plt

def reduced_magnetization(x_initial, T):
    x = optimize.newton(lambda x: x - np.tanh(x/T), x_initial)
    return x

T = np.linspace(0.1, 0.99, 100)
reduced_mag = [reduced_magnetization(0.5, t) for t in T]

plt.plot(T, reduced_mag)
plt.xlabel('Temperatura')
plt.ylabel('Magnetización Reducida')
plt.title('Magnetización Reducida vs Temperatura')
plt.show()
````

````python
#e. Para los 100 puntos anteriores, emplee el metodo de interpolación de lagrange y cubic-spline. Realizar una gráfica mostrando los puntos
#y el polinomio de interpolación para cada unos de los metodos. ¿Que puede concluir de los metodos?

#LAGRANGE
poly = lagrange(T, reduced_mag)
print(poly)
````

````python
# Interpolación de Lagrange
poly = lagrange(T, reduced_mag)


# Spline cúbico
cubic = CubicSpline(T, reduced_mag)


# Graficar
plt.plot(T, reduced_mag, 'o', label='Datos')
plt.plot(T, poly(T), label='Lagrange')
plt.plot(T, cubic(T), color='cyan', label='Cubic Spline')
plt.legend(loc='best')
plt.xlabel('$Temperatura$')
#plt.ylim(0.2,1.2)
plt.ylabel('$Magnetización$ $reducida$')
plt.title('$Comparación$ $de$ $métodos$')
plt.show()
````
