````python
def mybin(x):
  bin = []

  while x!=0:
    cociente = x // 2 # Divide x entre dos y el cociente es numero entero
    modulo = x % 2 # Modulo de x divido entre 2
    bin.append(modulo) # Agrega los resultados de modulo a bin
    x = cociente # Almacena x por los cocientes
    b = "".join([str(i) for i in bin [::-1]]) # Cambia lista bin a cadena de caracteres de la lista invertida bin
    b = b.rjust(8, "0") # Restringe b a 8 caracteres
  return b

# Escribir x
x = 128
print("El número binario de", x, "es", mybin(x))
````

````python
#binario = 11101
def bina_a_decimal(binario):
  bin_lista = [int(digito) for digito in str(binario)]
  
  bin_inverse = bin_lista[::-1]
  w = len(bin_inverse)
  #print(bin_inverse)
  
  lista = []
  operaciones = []
  cont = 0
  
  for j in range(w):
    x = cont
    cont +=1
    operaciones.append(x)

    #print(operaciones)
    
  for i in range(0,w):
    x = bin_inverse[i] * (2**operaciones[i])
    lista.append(x)
    #print(x)
    
  total = sum(lista)
  
  return total
````

````python
#number32("00111110001000000000001000010000")

def number32(BIN):
  numero32 = str(BIN)
  numero32 = list(numero32)

  list1 = []
  list2 = []
  list3 = []

  numero32.reverse()

  list1 = numero32[0]    #sign
  list2 = numero32[1:9]    #8 bits
  list3 = numero32[9:-1]   #23 bits

  w = len(list3)
  h = len(list2)

  list3.reverse()
  list2.reverse()

  #sumatorio de b
  b =[]
  for i in range(0,w):
    x = int(list3[i])
    x1 = x / (2**((w)-i))
    b.append(x1)
    
  suma_b = sum(b)

  #sumatoria e
  e = []
  for j in range(0,h):
    y = int(list2[j])
    b_e = y * (2**j)
    e.append(b_e)

  suma_e = sum(e)

  #calculo s
  
  for k in range(len(list1)):
    
    if int(list1[k]) == 0:
      s = 1
    
    else:
      s = -1

  #Total
   
  T1 = ((-1)**s) / (2**(127-suma_e))
  T2 = 1 + suma_b

  T = T1 * T2

  return(T)
````

1. Write a python script that calculates the double precision number represented by a 64-bits binary.

the name of function should be: 
```
number64(BIN):
```

2. What is the number represented by:

BIN="0 10000000011 1011100100001111111111111111111111111111111111111111"

````python
def number64(BIN):
  numero32 = str("0100000000111011100100001111111111111111111111111111111111111111")
  numero32 = list(numero32)

  list1 = []
  list2 = []
  list3 = []

  numero32.reverse()

  list1 = numero32[0]    #sign
  list2 = numero32[1:12]    #11 bits
  list3 = numero32[12:-1]   #52 bits

  w = len(list3)
  h = len(list2)

  list3.reverse()
  list2.reverse()

  #sumatorio de b
  b =[]
  for i in range(0,w):
    x = int(list3[i])
    x1 = x * (2**(-i))
    b.append(x1)
    
  suma_b = sum(b)

  #sumatoria e
  e = []
  for j in range(0,h-1):
    
    y = int(list2[j])
    b_e = y * (2**j)
    e.append(b_e)

  suma_e = sum(e)

  #calculo s
  
  for k in range(len(list1)):
    
    if int(list1[k]) == 0:
      s = 1
    
    else:
      s = -1

   #Total
   
  T1 = ((-1)**s) * (2 **(suma_e - 1023))
  T2 = 1 + suma_b

  T = (T1 * T2)

  return T
````
