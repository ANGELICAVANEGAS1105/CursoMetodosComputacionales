El método de Euler es un método numérico utilizado para aproximar soluciones a ecuaciones diferenciales ordinarias de primer orden. Se basa en la idea de aproximar la solución de la ecuación diferencial por una línea recta que conecta dos puntos cercanos en la curva.

A medida que el tamaño de los pasos se hace más pequeño, la aproximación de Euler se acerca cada vez más a la solución exacta de la ecuación diferencial.



El método de Euler es un método simple y fácil de implementar, pero puede tener errores significativos si el tamaño de paso es grande o si la función $f(x,y)$ es complicada. Existen otros métodos numéricos más sofisticados que mejoran la precisión de la solución, como el método de Runge-Kutta.


\title{Método de Euler para resolver ecuaciones diferenciales}


\section{Deducción del método de Euler}

Supongamos que tenemos una ecuación diferencial de primer orden de la forma:

\begin{equation}
\frac{dy}{dt} = f(t, y)
\end{equation}

donde $y$ es la función desconocida que queremos resolver, $t$ es la variable independiente y $f$ es una función conocida que define la tasa de cambio de $y$ con respecto al tiempo.

Si queremos encontrar el valor de $y$ en un punto $t$ dado, podemos aproximar su valor mediante el método de Euler. Para ello, dividimos el intervalo de tiempo $t$ en $N$ subintervalos iguales de ancho $h$, de modo que:

\begin{equation}
h = \frac{t}{N}
\end{equation}

Luego, aproximamos la función $y$ en cada uno de los puntos $t_i = ih$ mediante una serie de valores discretos:

\begin{equation}
y_i = y(t_i)
\end{equation}

Para obtener el siguiente valor de la función $y$ en el punto $t_{i+1} = (i+1)h$, utilizamos la fórmula de Euler:

\begin{equation}
y_{i+1} = y_i + hf(t_i, y_i)
\end{equation}

Esta fórmula se obtiene a partir de la aproximación de la tasa de cambio de $y$ mediante una diferencia finita:

\begin{equation}
\frac{dy}{dt} \approx \frac{y_{i+1} - y_i}{h}
\end{equation}

y sustituyendo $\frac{dy}{dt} = f(t_i, y_i)$.

\section{Ejemplo de aplicación}

Supongamos que queremos resolver la ecuación diferencial:

\begin{equation}
\frac{dy}{dt} = -2y + 3
\end{equation}

con condición inicial $y(0) = 1$. Para ello, podemos utilizar el método de Euler con $N = 4$ subintervalos en el intervalo $[0,1]$. Entonces, el ancho de cada subintervalo es $h = 1/4 = 0.25$.

Aplicando la fórmula de Euler, obtenemos los siguientes valores de la función $y$ en cada uno de los puntos:

\begin{align}
t_0 &= 0, &y_0 &= 1 \
t_1 &= 0.25, &y_1 &= 1 + 0.25(-2 \cdot 1 + 3) = 1.25 \
t_2 &= 0.5, &y_2 &= 1.25 + 0.25(-2 \cdot 1.25 + 3) = 1.4375 \
t_3 &= 0.75, &y_3 &= 1.4375 + 0.25(-2 \cdot 1.4375 + 3) = 1.53515625 \
t_4 &= 1, &y_4 &= 1.53515625 + 0
\end{align}

```python
def euler_method(m, k, b, x0, v0, dt, num_steps):
    x = np.zeros(num_steps)
    v = np.zeros(num_steps)
    t = np.zeros(num_steps)
    
    x[0] = x0
    v[0] = v0
    
    for i in range(1, num_steps):
        t[i] = i * dt
        x[i] = x[i-1] + dt * v[i-1]
        v[i] = v[i-1] - (k/m) * x[i-1] * dt - (b/m) * v[i-1] * dt
        
    return t, x

# Parámetros de la ecuación diferencial
m = 1.0  # masa
k = 2.0  # constante del resorte
b = 0.5  # coeficiente de amortiguamiento

# Condiciones iniciales
x0 = 0.0  # posición inicial
v0 = 1.0  # velocidad inicial

# Parámetros de la simulación
dt = 0.01  # tamaño del paso de tiempo
num_steps = 1000  # número de pasos de tiempo

# Resolver la ecuación diferencial utilizando el método de Euler
t, x = euler_method(m, k, b, x0, v0, dt, num_steps)

# Graficar la solución
plt.plot(t, x, 'm')
plt.xlabel('Tiempo')
plt.ylabel('Posición')
plt.title('Solución de la ecuación diferencial por el método de Euler')
plt.grid()
```
