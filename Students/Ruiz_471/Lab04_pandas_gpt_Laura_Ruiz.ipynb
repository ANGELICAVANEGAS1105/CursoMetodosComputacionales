```python
import pandas as pd
import datetime
```

```python
url = "https://docs.google.com/spreadsheets/d/e/2PACX-1vRgLiVkqxr3oML0jwCCYjDSzXo8UMRSmsgrJc7w1qDM-EWGcuOmhPUKcaASReC8rw/pub?output=xlsx"
```

```python
df = pd.read_excel(url)
df
```

```python
cools = ["Date", "Unnamed: 1", "Berril", "Maisonneuve_1", "Maisonneuve_2", "Brebeuf"]
df = df.rename(columns={"Date":"Serie de tiempo"})
df = df.drop('Unnamed: 1', axis=1)
df
```

```python
#Numero de registros NaN de cada columna

col1 = df['Serie de tiempo'].isna().sum()
col2 = df['Berri1'].isna().sum()
col3 = df['Maisonneuve_1'].isna().sum()
col4 = df['Maisonneuve_2'].isna().sum()
col5 = df['Brebeuf'].isna().sum()
```

```python
#Convertir fechas a numeros

df['Serie de tiempo'] = pd.to_datetime(df['Serie de tiempo'])

# Convertir las fechas a números enteros que representan el número de días transcurridos desde el 1 de enero de 1970
df['Serie de tiempo'] = (df['Serie de tiempo'] - pd.Timestamp('1970-01-01')) / pd.Timedelta('1 day')
```

```python
#Estadistica basica

media1 = df["Serie de tiempo"].mean()
media2 = df["Berri1"].mean()
media3 = df["Maisonneuve_1"].mean()
media4 = df["Maisonneuve_2"].mean()
media5 = df["Brebeuf"].mean()


des1 = df["Serie de tiempo"].std()
des2 = df["Berri1"].std()
des3 = df["Maisonneuve_1"].std()
des4 = df["Maisonneuve_2"].std()
des5 = df["Brebeuf"].std()

max1 = df["Serie de tiempo"].max()
max2 = df["Berri1"].max()
max3 = df["Maisonneuve_1"].max()
max4 = df["Maisonneuve_2"].max()
max5 = df["Brebeuf"].max()


desv_est = df.std()
maxi = df.max()
media = df.mean()
```

```python
memo_t = df.memory_usage(index = False)
memo_t
```

```python
#Memoria usada

memo1 = df["Serie de tiempo"].memory_usage(deep = True, index = False)
memo2 = df["Berri1"].memory_usage(deep = True, index = False)
memo3 = df["Maisonneuve_1"].memory_usage(deep = True,  index = False)
memo4 = df["Maisonneuve_2"].memory_usage(deep = True, index = False)
memo5 = df["Brebeuf"].memory_usage(deep = True, index = False)

print(memo1, memo2, memo3, memo4, memo5)
```

# **Solución problema 03:**

```python
import pandas as pd
import numpy as np
```

```python
url = 'https://raw.githubusercontent.com/hernansalinas/autogrades/main/Laboratorios_Taller/dataset/Crimes_-_2019.csv'

crimenes_2019 = pd.read_csv(url)
crimenes_2019
```

```python
crimenes = pd.DataFrame(crimenes_2019)
crimenes
```

```python
#Estadistica basica

#describe nos da toda la información estadistica del dataframe
#estadistic = crimenes.describe().loc[['mean']] asi solo nos dara la media

estadistic = crimenes.describe()
estadistic
```

```python
#Eliminar columnas
 
crimenes1 = crimenes.drop('Primary Type', axis=1)
crimenes1 = crimenes.drop('Date', axis=1)

crimenes1
```

```python
#Transformar la columna date

crimenes['Date'] = pd.to_datetime(crimenes['Date'], format='%m/%d/%Y %H:%M')
crimenes
```

```python
# Total de crimenes por tipo

numero_de_crimenes = crimenes.groupby('Primary Type').Date.count().sort_values()
numero_de_crimenes
```

```python
#Ordenar de forma alfabetica la columna primary type

crimenes = crimenes.sort_values('Primary Type')
crimenes
```

```python
#Mostrar de la fila 100 a la 120

datos_new = crimenes.iloc[101:121]
datos_new
```

```python
#Visualización de datos

import seaborn as sns
import matplotlib.pyplot as plt

casos = pd.DataFrame(numero_de_crimenes)

casos = casos.rename(columns = {'Date':'Numero_de_casos'})

#data frame numero de casos y y la categoria

casos = casos.reset_index()
casos = casos.rename(columns = {'Primary Type':'Tipo de casos'})
casos

#grafico

plt.figure(figsize = (16,12))
ax = sns.barplot(x = 'Tipo de casos', y='Numero_de_casos', data = casos)
ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
ax.set_title('Crimenes 2019')
ax.set_xlabel('Tipo de casos')
ax.set_ylabel('Número de casos')
plt.grid()
```

```python
#Crear DATAFRAME
vacunados = pd.DataFrame(paises_vacu)
vacunados
```

```python
#Filtrar datos para colombia

datos_colombia = vacunados.loc[vacunados['country']=='Colombia']
```

```python
#Comparación vacunados

vacunados = vacunados.loc[:, ['country','total_vaccinations']]
latinos = vacunados[vacunados['country'].isin(['Argentina', 'Bolivia', 'Brasil', 'Chile', 'Colombia', 'Costa Rica', 'Cuba', 'Ecuador', 'El Salvador', 
                                'Guatemala', 'Haití', 'Honduras', 'México', 'Nicaragua', 'Panamá', 'Paraguay', 'Perú', 'Puerto Rico', 
                                'República Dominicana', 'Uruguay', 'Venezuela'])]

latinos = latinos.groupby('country').sum().reset_index()
latinos
```

```python
#Grafico de comparación

plt.figure(figsize = (14,10))
ax = sns.barplot(x = 'country', y='total_vaccinations', data = latinos)
ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
ax.set_title('Vacunados en Latinoamerica')
ax.set_xlabel('Paises')
ax.set_ylabel('Cantidad de vacunados')
plt.grid()
```
