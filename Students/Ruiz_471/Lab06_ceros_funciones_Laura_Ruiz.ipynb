# Metodo de Bisecciónn

Para utilizar este metodo el primer paso es tener identificada nuestra función objetivo en este caso sera la siguiente 


$f(x) = (x-1) - 1 = -x + 2$


Luego de este debemos encontrar las raices de la función es decir igualamos a cero nuestra función y despejamos nuestra variable x, de la siguiente manera



$f(x) = -x + 2$ = 0

$-x + 2 = 0$

$x = 2$

Aca vemos que una raiz de la función la encontramos en e valor de $x = 2$.

Ahora debemos construir un intervalo alrededor del punto de la raiz ya que el emtodo de bisección lo que haces es ir acercandose poco a poco a ese punto mediante la reduccion dle intervalo de la siguiente manera


El valor que queremos encontra es $c$ el cual se haya de la siguiente manera

$c = (a + b) / 2$

````python
#METODO DE BISECCIÓN

#Identificamos la función a utilizar y la definimos

def f(x):

  return -x + 2

#Encontramos las raices de la función

#f(x) = 0

# x = 2

#Con la raiz encontrada creamos un intervalo que se acerque a este valor 

#N =  numero de iteraciones

N = 20

for i in range(N):    #Creamos un ciclo que evalue un intervalo hasta que el error de la raiz sea el minimo

  a = 1.5
  b = 3

  c = (a+b)/2

  b = b + c
  
  c = (a+b)/2

  a = a + c

  c = (a+b)/2

  if abs(f(c) - 0) < 0.001:

    break
  
  else:
    
    print("Faltan mas iteraciones")
````

# **Metodo Newton - Raphson**

El objetivo de este metodo es encontrar un valor cercano a la raiz de una funcióon utilizando la formula recursiva la cual se deriva de la expansion de la serie de Taylor con un valor de $X0$ cercano a la valor de α la cual es la raiz de la función objetivo.


Formula recursiva:

$Xn+1 = Xn - (f(Xn) / f'(Xn))$

Con esta formula podremos encontrar un valor cercano a la raiz α con $Xn$ cada vez mas cercanos a la raiz de la función objetivo



````python
#Deducción

N = 25 #Numero de iteraciones a realizar

def f(x):    #Funcion a utilizar

  return x**3 - x +1

def derif(x):  #Derivada de la funcion

  return 3*(x**2) - 1
````

````python
import matplotlib.pyplot as plt   #Graficamos la funcion

x = np.linspace(-5,5,50)
plt.plot(x,f(x))
plt.axhline(y=0, color='r', linestyle='--')
plt.grid()
````

````python
#Vemos que la grafica nos indica que puntos de partida debems tomar para calcular la raiz y que estos no se alejen mucho de esta
````

````python
x1 = -2   #Tomamos este numero como punto de partida
x = []
x.append(x1)  #Lo agregamos a una lista 

for i in range(N):   #Creamos un cilco el cual tome los diferentes valores de la lista y los vaya evaluando en la fomrula recursiva

  x_ = x[i] - f(x[i]) / derif(x[i])

  x.append(x_)  #Agregamos cada resultado en la lista ya que como es un ciclo este debe evaluar cada resultado obtenido


print(x)  #Resultado de las iteraciones, alguno de estos valores es el mas optimo para ser la raiz o es muy cerca
         
````

````python
#Estos resultados estan cerca del valor de la raiz de la funcion si evaluamos estos y nos dan cero la funcion sera el valor correcto de la raiz
ceros = []
ceros1 = []

for j in range(len(x)):
  
  y = f(x[j])
  
  if y <= 0:
    
    ceros.append(y)


#Nuestros valor de la raiz debe ser el minimo

valores_absolutos = [abs(cero) for cero in ceros]

raiz = min(valores_absolutos, key = abs)

print(ceros)
print(len(ceros))
````

# **Diferencia entre el metodo de Newton-Raphson y la secante**

Para el metodo de newton vimos anteriormente que se utiliza la formula recursiva derivada de la expansion de la serie de taylor por otro lado el metodo de la secante tambien se basa en la expancion de la serie de taylor pero en este caso aproximamos la funcion objetivo a una recta secante que pasa por los puntos especificos de la serie y con ayuda de la derivada podemos obtener dicha recta secante, la aproximacion de la recta secante se da gracias a la aproximacion de la defincion de la derivada y despejando los valores de $x$ podemos obtener las raices de la función.

Para la secante se utiliza la siguiente formula


$Xn+1 = Xn - (f(Xn) * (Xn - Xn-1) / f(Xn) - f(Xn-1))$

# **Punto Fijo**

Este metodo lo que utiliza es crear una nueva funcion donde tenemos como inicio la defincion de un punto $X0$ especifico  el cual le restaremos el valor de la funcion lo que esto hace es ir buscando valores en la recta $y = x$ que correspondan a la funcion y asi hasta que corte en un respectivo punto con la recta y asi encontrar la raiz de dicha función.

Se siguen los sigueintes pasos:

1) Definir $g(x) = x0 -f(x)$

2) Evaluar $g(x0) = x0 - f(x0) = x0$

3) Definir el punto $P0$  y evaluarlo en $g(P0)$

4) Calcular el intercepto con la linea recta en $P1 = (g(P0) , (g(P0),g(P0)))$

5)Luego ir al punto$P1$

6)El siguiente punto seria $P1 = (P1, g(P1))$

7) Comparar el punto$P1$ con $g(P1)$ y deben satisfacer las condiciones de los primeros puntos si no pasa se debe escoger otro punto $x0$

````python
#Punto 2

#Metodo biseccion

import numpy as np
from scipy.optimize import bisect
import matplotlib.pyplot as plt

x = np.linspace(0,3,1000)

def f(x):
  return np.exp(x) - 2 - np.cos(np.exp(x) - 2)

a = -100
b = 100

z = bisect(f,a,b)

print("La raíz de la función es:", z)
````

#3. Partícula en reposo
Una particula parte del reposo sobre un plano inclinado uniforme, cuyo ángulo 
$\theta$ cambia con una rapidez constante de : 

\begin{equation}
\frac{\mathrm{d}\theta}{\mathrm{d} t} = \omega < 0
\end{equation}

Al final de t segundos, la posición del objeto esta dada por: 


\begin{equation}
x(t) = -\frac{g}{2\omega^2} \left( \frac{e^{\omega t}-e^{-\omega t}}{2}-\sin\omega t\right) 
\end{equation}

Suponga que la particula se desplazo 0.51m en 1s. Encuentre, con exactitud de $10^{-5}$ la rapidez $\omega$ con que cambia $\theta$. Supoga $g=9.76$m/s

 

````python
#SOLUCIÓN:
````

````python
import numpy as np
````

````python
def parti(w):

  t = 1

  y1 = -9.76/2*(w**2)
  y2 = (np.exp(w*t) - np.exp(-w*t)) / 2
  y3 = np.sin(w*t)

  y = (y1 * (y2 -y3)) - 0.51

  return y
````

````python
from scipy.optimize import fsolve
````

````python
raiz = fsolve(parti, 2,  xtol=1e-5)
raiz
````

````python
print("La rapidez con la que cambia el angulo respecto a el tiempo es w =", raiz)
````

````python
#El metodo de newton tambien nos da una mayor presición de 10E-6

#Metodo newton

from scipy.optimize import newton

y = newton(parti, x0=-1)

print("La raíz de la función es:", y)
````

Supongamos que un proyectil es lanzado desde el origen con una rapidez $v=vo( \cos \theta , \sin \theta $. Supongamos además que el efecto del viento sobre el proyectil es el de una fuerza proporcional a la velocidad relativa entre el viento y el proyectil. Teniendo presente que si u es la velocidad del viento y $\alpha$ el ángulo que hace dirección de esta velocidad con el eje x, se puede entrontrar las siguientes ecuaciones diferenciales, asociada al estado de movimiento del [proyectil](https://iopscience.iop.org/article/10.1088/0143-0807/36/2/025016/meta):


\begin{equation}
m\frac{dv_x}{dt} = -mb(v_x-u\cos \alpha )
\end{equation}

\begin{equation}
m\frac{dv_y}{dt} = -mg-mb(v_x-u\sin \alpha ),
\end{equation}

donde b es el párametro de rozamiento. La soluciones a las ecuaciones anteriores vienen dadas por:

Posición en x
\begin{equation}
x=(u\cos \alpha) t + \frac{1}{b} (v_0\cos \theta_0 -u \cos\alpha)(1-\mathrm{exp}(-bt))
\end{equation}

Posicion en y 
\begin{equation}
y= \frac{1}{b}\left(\frac{g}{b}+v_0 \sin \theta_0 -u\sin\alpha \right)(1-\mathrm{exp}(-bt))-\left( \frac{g}{b}-u\sin\alpha\right)t
\end{equation}

````python
# Parametros iniciales

v0 = 60 #Velocidad en m/s
alpha = np.pi #Angulo del viento
theta = np.deg2rad(45) #Angulo de lanzamiento
u = v0/2             #rapidez del viento
b = 0.05             #Parametro de friccion 
g = 9.8                #Aceleracion de la gravedad m/s2

parametros = (alpha, v0, b, u,g)  #Parametros en tupla.
````

````python
# Posicion en x

def pos_x(t, v0, u, alpha, theta, b, g):

  x1 = u*np.cos(alpha)* t
  x2 =  (1/b) * (v0*np.cos(theta)) - (u)*np.cos(alpha)
  x3 = (1-np.exp(-b*t))

  x = x1 + x2 * x3

  #x = ((60/2)*np.cos(np.pi)) * t + (1/0.05) * (60*np.cos(np.deg2rad(45)) - (60/2)*np.cos(np.pi)) * (1-np.exp(-0.05*t))

  return x

# Posicion en y

def pos_y(t, v0, u, alpha, theta, b, g):

  y1 = (1/b) * (g/b + v0*np.sin(alpha) - u*np.sin(theta))
  y2 = (1-np.exp(-b*t))
  y3 =  (g/b - u*np.sin(alpha)) * t

  y = y1 * y2 - y3

  #y = (1/0.05) * ((9.8/0.05) + 60*np.sin(np.pi) - (60/2)*np.sin(np.deg2rad(45))) * (1-np.exp(-0.05*t)) - ((9.8/0.05) - (60/2)*np.sin(np.deg2rad(45))) * t

  return y 
````

````python
vo = 60 #Velocidad en m/s
alpha = np.pi #Angulo del viento
theta = np.deg2rad(45) #Angulo de lanzamiento
u = vo/2             #rapidez del viento
b = 0.05             #Parametro de friccion 
g = 9.8                #Aceleracion de la gravedad m/s2

def pos_x(t, vo, theta0, u, alpha, b, g  ):
  t1 = u*np.cos(alpha)*t  
  t2 = 1./b*(vo*np.cos(theta0)-u*np.cos(alpha))  
  t3 = (1-np.exp(-b*t))
  return t1 + t2 * t3


def pos_y(t, vo, theta0, u, alpha, b, g  ):
  t1 = (g/b + vo*np.sin(theta0) - u* np.sin(alpha) )
  t2 = (1-np.exp(-b*t)) 
  t3 = (g/b - u*np.sin(alpha))*t 
  return 1./b * t1 * t2 - t3
````


# 5. Construir un fractal empleando el metodo de Newton-Raphson, 
(Reproducir código y entender)
\begin{equation}
z_{n+1} = z_n -a \frac{f(z)}{f'(z)} 
\end{equation}



\begin{equation}
f'(z) = \frac{ f(z+h)+f(z) }{h}
\end{equation}

````python
import numpy as np
import matplotlib.pylab as plt
from PIL import Image

Npx_X = 80 #Numero de pixeles en x
Npx_Y = 80 #Numero de pixeles en y

image = Image.new("RGB", (Npx_X, Npx_Y)) #Creacion de la imagen inicial

npx = np.arange(0, Npx_X)
npy = np.arange(0, Npx_Y)

xmin = -1
xmax = 1
ymin = -1
ymax = 1
# Transformación al plano complejo en un rectangulo (xmin, xmax), (ymin, ymax)
zx = npx * (xmax - xmin) / (Npx_X - 1)   + xmin
zy = npy * (ymax-ymin) / (Npx_Y - 1) + ymin

# Definicion  la funcion y las derivada 
def f(z): 
  return z**3 - 1

def fp(z): 
  return 3*z**2

h = 1E-4
def fpn(z, h): 
  return (f(z+h)-f(z))/h


a = 0.2
Niteraciones = 120
eps = 1E-8

x = - 1
for zx_ in zx:
  x = x + 1 
  y = -1
  for zy_ in zy: 
      y = y + 1
      Z_0 = complex(zx_, zy_)

      for i in range(0, Niteraciones): 
        Z = Z_0 - a * f(Z_0)/fp(Z_0)
        if( abs(Z-Z_0)<= eps):
          break           
        Z_0 = Z
      #print(i)
      #print(i)
      r, g, b = i % 4 *64 , i % 8 *16, i% 16* 16      
      image.putpixel((x,y), (r,g,b) )

plt.imshow(image)
# plt.xlim(400, 600)
# plt.ylim(400, 523)
````
