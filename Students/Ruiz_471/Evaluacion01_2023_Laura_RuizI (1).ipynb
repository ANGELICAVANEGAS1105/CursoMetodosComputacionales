## 1. (50%). Movimiento de una esfera de radio r en un fluido.

 Una esfera de radio $R$ que asciende o desciende por un fluido puede ser modelada a través de un modelo de partícula, sometida a tres fuerzas: el peso, el empuje y la fuerza rozamiento. Las  ecuaciones de posición (x) y velocidad (v) para un sistema de refencia úbicado en la base del fluido, pueden ser escritas como:

\begin{equation}
x = \frac{G}{\alpha} \left( 1+\frac{\alpha v_0}{G} \right)(1-e^{-\alpha t}) - \frac{G}{\alpha}t
\end{equation}

\begin{equation}
v= \left ( v_0 +\frac{G}{\alpha} \right )e^{-\alpha t} - \frac{G}{\alpha}
\end{equation}



donde $G=\left( 1-\frac{\rho_f}{\rho} \right)g$, con $\rho_f$, $\rho$ la densidad del fluido y la esfera respectivamente y $g=9.8$ la aceleración de la gravedad y $\alpha$  una constante que depende del radio de la esfera, la viscosidad y la densidad.
$\alpha=\frac{9\eta}{2R^2\rho}$.

Suponga que una esferita de radio $R=2\times10^{-5}m$ es lanzada verticalmente hacia arriba en el agua, con una rapidez de $v_0=0.01$m/s.
Tomando la densidad del agua en  $\rho_f =10000 kg/m^3$,  la densidad de la esfera en $\rho=2670 kg/m^3$ y la viscosidad en $\eta=0.001 kg/ms$.  Determine: 

1. El tiempo total en que la esfera asciende  y desciende. 

2. Determine la rapidez final, con la que llega nuevamente de donde fue lanzada. 



Para ello construya un programa en python que:
- Permita calcular la posición como función del tiempo
- Permita calcular la rapidez como función del tiempo

## 1. (50%). Movimiento de una esfera de radio r en un fluido.

 Una esfera de radio $R$ que asciende o desciende por un fluido puede ser modelada a través de un modelo de partícula, sometida a tres fuerzas: el peso, el empuje y la fuerza rozamiento. Las  ecuaciones de posición (x) y velocidad (v) para un sistema de refencia úbicado en la base del fluido, pueden ser escritas como:

\begin{equation}
x = \frac{G}{\alpha} \left( 1+\frac{\alpha v_0}{G} \right)(1-e^{-\alpha t}) - \frac{G}{\alpha}t
\end{equation}

\begin{equation}
v= \left ( v_0 +\frac{G}{\alpha} \right )e^{-\alpha t} - \frac{G}{\alpha}
\end{equation}



donde $G=\left( 1-\frac{\rho_f}{\rho} \right)g$, con $\rho_f$, $\rho$ la densidad del fluido y la esfera respectivamente y $g=9.8$ la aceleración de la gravedad y $\alpha$  una constante que depende del radio de la esfera, la viscosidad y la densidad.
$\alpha=\frac{9\eta}{2R^2\rho}$.

Suponga que una esferita de radio $R=2\times10^{-5}m$ es lanzada verticalmente hacia arriba en el agua, con una rapidez de $v_0=0.01$m/s.
Tomando la densidad del agua en  $\rho_f =1000 kg/m^3$,  la densidad de la esfera en $\rho=2670 kg/m^3$ y la viscosidad en $\eta=0.001 kg/ms$.  Determine: 

1. El tiempo total en que la esfera asciende  y desciende. 

2. Determine la rapidez final, con la que llega nuevamente de donde fue lanzada. 



Para ello construya un programa en python que:
- Permita calcular la posición como función del tiempo
- Permita calcular la rapidez como función del tiempo
- Cálcule  el tiempo total de subida y bajada empleando el método de la secante.


```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
```

```python
#DEFINIMOS LA FUNCION DE LA POSICION EN X
```

```python
R = 2*(10**-5)
v0 = 0.01
den_agua = 1000
den_esfe = 2670
vis = 0.001
G = (1 - (den_agua/den_esfe))
alpha = vis / (2* (R**2) * den_esfe)

param = (R,v0,den_agua,den_esfe,vis,G,alpha)
param1 = (v0,G,alpha)
```

```python
def x(t,R,v0,den_agua,den_esfe,vis,G,alpha):

  t = sp.symbols('t')

  y = ((G/((alpha)**2)) *(1 + (alpha*v0 / G)) * (1 - sp.exp(-alpha*t))) - ((G/alpha) * t)

  return y
```

```python
t = sp.symbols('t')
x(t,*param)
```

```python
def v(t,v0,G,alpha):

  t = sp.symbols('t')

  y = (v0 + (G/alpha)) * sp.exp(-alpha*t)  -  (G/alpha)

  return y
```

```python
t = sp.symbols('t')
v(t,*param1)
```

```python
from sympy.utilities.lambdify import lambdify

x_ = lambdify(t,x(t,*param),'numpy')
v_ = lambdify(t,v(t,*param1),'numpy')
```

```python
t1 = np.linspace(0,0.1,50)

plt.plot(t1,x_(t1), color='red')
plt.grid()
```

```python
t1 = np.linspace(0,0.1,50)

plt.plot(t1,v_(t1))
plt.grid()
```

```python
#Metodo de la secante para la funcion x_ tiempo de bajada

from scipy.optimize import root_scalar

r = root_scalar(x_, method = 'secant', x0 = 0.01, x1 = 0.05)
print(r)
root = 0.018120274089339635
print('La raiz para la funcion de la poscionen x es', root)
```

```python
t1 = np.linspace(0,0.1,50)

plt.plot(t1,x_(t1))
plt.grid()

plt.axhline(y=0, color='red')
plt.xlim(-0.035,0.035)
plt.ylim(-0.00003,0.00003)
plt.plot(root,0,'go')
```

```python
#Metodo de la secante para la funcion x_ tiempo de subida

from scipy.optimize import root_scalar

r = root_scalar(x_, method = 'secant', x0 = -0.01, x1 = 0.001)
print(r)
root2 = 7.931502645061374e-15
print('La raiz para la funcion de la poscionen x es', root2)
```

```python
t1 = np.linspace(0,0.1,50)

plt.plot(t1,x_(t1))
plt.grid()

plt.axhline(y=0, color='red')
plt.xlim(-0.035,0.035)
plt.ylim(-0.00003,0.00003)
plt.plot(root2,0,'go')
```

```python
print('El tiempo de subida es:', root2, 'y el tiempo de bajada es:', root,'en unidades de segundos')
```

```python
#Metodo de la secante para la funcion v_

from scipy.optimize import root_scalar

r = root_scalar(v_, method = 'secant', x0 = 0.001, x1=0.003)
print(r)
root1 = 0.004567416120609578
print('La raiz para la funcion de la poscionen x es', root1)
```

```python
t1 = np.linspace(0,0.1,50)

plt.plot(t1,v_(t1))
plt.grid()

plt.axhline(y=0, color='red')
plt.ylim(-0.01,0.01)
plt.xlim(-0.1,0.1)
plt.plot(root1,0,'go')
```

```python
#Para encontrar la rapidez final evaluamos la funcion v_ en el tiempo decaida

v_final = v_(root)

print('La rapidez final de la esfera es:', v_final,'m/s')
```

```python
#Graficamente

t1 = np.linspace(0,0.1,50)

plt.plot(t1,v_(t1))
plt.grid()

plt.axhline(y=0, color='red')
plt.ylim(-0.01,0.01)
plt.xlim(-0.1,0.1)
plt.plot(root,v_final,'go')
```
