## 1. (50%). Movimiento de una esfera de radio r en un fluido.

 Una esfera de radio $R$ que asciende o desciende por un fluido puede ser modelada a través de un modelo de partícula, sometida a tres fuerzas: el peso, el empuje y la fuerza rozamiento. Las  ecuaciones de posición (x) y velocidad (v) para un sistema de refencia úbicado en la base del fluido, pueden ser escritas como:

\begin{equation}
x = \frac{G}{\alpha} \left( 1+\frac{\alpha v_0}{G} \right)(1-e^{-\alpha t}) - \frac{G}{\alpha}t
\end{equation}

\begin{equation}
v= \left ( v_0 +\frac{G}{\alpha} \right )e^{-\alpha t} - \frac{G}{\alpha}
\end{equation}



donde $G=\left( 1-\frac{\rho_f}{\rho} \right)g$, con $\rho_f$, $\rho$ la densidad del fluido y la esfera respectivamente y $g=9.8$ la aceleración de la gravedad y $\alpha$  una constante que depende del radio de la esfera, la viscosidad y la densidad.
$\alpha=\frac{9\eta}{2R^2\rho}$.

Suponga que una esferita de radio $R=2\times10^{-5}m$ es lanzada verticalmente hacia arriba en el agua, con una rapidez de $v_0=0.01$m/s.
Tomando la densidad del agua en  $\rho_f =10000 kg/m^3$,  la densidad de la esfera en $\rho=2670 kg/m^3$ y la viscosidad en $\eta=0.001 kg/ms$.  Determine: 

1. El tiempo total en que la esfera asciende  y desciende. 

2. Determine la rapidez final, con la que llega nuevamente de donde fue lanzada. 



Para ello construya un programa en python que:
- Permita calcular la posición como función del tiempo
- Permita calcular la rapidez como función del tiempo
- Cálcule  el tiempo total de subida y bajada empleando el método de la secante.

 


```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.optimize as opt
```

```python
# primero definimos los valores que nos dan en el examen

R   = 0.00002 #m - radio de la esfera 
v_0 = 0.01     #m/s - velociadd con la que es lanzada 
ρ_f = 1000    #kg/m³ - densidad del agua
ρ   = 2670    #kg/m³ - densidad de la esfera 
n   = 0.001   #kg/ms - viscosidad 
g   = 9.8 
```

```python
G = (1 - (ρ_f/ρ))
α = (9*n)/(2*R**2*ρ)
```


\begin{equation}
x = \frac{G}{\alpha²} \left( 1+\frac{\alpha v_0}{G} \right)(1-e^{-\alpha t}) - \frac{G}{\alpha}t
\end{equation}



```python
#definimos la funcion de la posicion con respecto al tiempo

def pos_x(t):
  R   = 0.00002 #m - radio de la esfera 
  v_0 =0.01     #m/s - velociadd con la que es lanzada 
  ρ_f = 1000    #kg/m³ - densidad del agua
  ρ   = 2670    #kg/m³ - densidad de la esfera 
  n   = 0.001   #kg/ms - viscosidad 
  g   = 9.8
  G = (1 - (ρ_f/ρ))
  α = (9*n)/(2*R**2*ρ)  


  x =((G/α**2) * (1 + ((α * v_0)/G)) * ( 1 - np.exp(-α*t)) ) - (G/α) *t
  return x

```

\begin{equation}
v= \left ( v_0 +\frac{G}{\alpha} \right )e^{-\alpha t} - \frac{G}{\alpha}
\end{equation}


```python
# definimos calculo de la velocidad con respecto al tiempo


def velocidad(t):
  R   = 0.00002 #m - radio de la esfera 
  v_0 =0.01     #m/s - velociadd con la que es lanzada 
  ρ_f = 1000    #kg/m³ - densidad del agua
  ρ   = 2670    #kg/m³ - densidad de la esfera 
  n   = 0.001   #kg/ms - viscosidad 
  g   = 9.8
  G = (1 - (ρ_f/ρ))
  α = (9*n)/(2*R**2*ρ)  


  v = (( v_0 + (G/α)) * np.exp(-α*t)) -  (G/α)
  
  return v
```

```python
import pandas as pd
from scipy import interpolate
```

```python
url= "https://raw.githubusercontent.com/hernansalinas/autogrades/main/Evaluaciones/dataset/Temperature_Bogota_Barcelona.csv"
```

```python
df = pd.read_csv(url)
df
```

```python
def temperatura(F):  #funcion para el cambio de temperatura
  C = (F-32)/1.8 
  return C
```

```python
#Adicionar una nueva columna AvgTempCelcius con la temperatura en grados celcius.  C=(F−32)/1.8
df["AvgTempCelcius"] =df.AvgTemperature.apply(lambda n :temperatura(n) )
df
```

Grafico de la temperatura con respecto a los años en bogota

```python
#sacar los años para bogota 

bogota= df[df["City"] == "Bogota"] #definimos una nueva lista para los datos solamente de bogota
bogota
```

```python
#sacamos los datos del año y de la temperatura en bogota de la nueva lista de bogota para poder graficarlo
x= bogota["Year"]
y= bogota["AvgTempCelcius"]
```

```python
plt.xlabel("tiempo(años)")
plt.ylabel("Temperatura (C)")
plt.title("temperatura vs tiempo(bogota)")
plt.plot(x,y,"ro")
plt.grid()
```

```python
#Aplique el metodos de interpolación de hermite para Bogotá 

#primero definimos la derivada 

dff= np.diff(y)


# calcular la interpolacion de hermite
hermite= interpolate.CubicHermiteSpline(x[1:], y[1:] , dff)
```

```python
plt.xlabel("tiempo(años)")
plt.ylabel("Temperatura (C)")
plt.title("temperatura vs tiempo(bogota)")
plt.plot(x[1:],y[1:],"ro")
plt.plot(x[1:],hermite(x[1:]))
plt.grid()
```

```python
x4=np.linspace(np.min(x[1:]),np.max(x[1:]),500)
plt.xlabel("tiempo(años)")
plt.ylabel("Temperatura (C)")
plt.title("temperatura vs tiempo(bogota)")
plt.plot(x[1:],y[1:],"ro")
plt.plot(x4,hermite(x4))
plt.grid()
```

Grafico de la temperatura con respecto a los años en Barcelona

```python
#sacar una lista solo para la ciudad de barcelona 

barcelona =  df[df["City"] == "Barcelona"] #definimos una nueva lista para los datos solamente de barcelona
barcelona
```

```python
#sacamos los datos del año y de la temperatura en bogota de la nueva lista de barcelona para poder graficarlo
x2= barcelona["Year"]
y2= barcelona["AvgTempCelcius"]
```

```python
#Aplique el metodos de interpolación de hermite para Bogotá 

#primero definimos la derivada 

dff= np.diff(y2)


# calcular la interpolacion de hermite
cubic= interpolate.CubicSpline(x2, y2)

```

```python

plt.xlabel("tiempo(años)")
plt.ylabel("Temperatura (C)")
plt.title("temperatura vs tiempo(barcelona)")
plt.plot(x2,y2,"ro")
plt.plot(x2,cubic(x2))
plt.grid()
```
