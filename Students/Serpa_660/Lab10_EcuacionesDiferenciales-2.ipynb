Después de realizar un análisis detallado del problema, a través de las leyes de Newton ó empleando las ecuaciones de Euler-Lagrange, se puede mostrar que las ecuaciones de movimiento para el sistema vienen dadas por:

\begin{equation}
m_1 \frac{ \mathrm{d}^2 x_1 }{\mathrm{d}t^2} = -k_1 x_1 + k_2 (x_2-x_1)
\end{equation}



\begin{equation}
m_2 \frac{ \mathrm{d}^2 x_2 }{\mathrm{d}t^2} = -k_3 x_2 - k_2 (x_2-x_1)
\end{equation}

De forma compacta,  la ecuacion anterior puede ser escrita como:


\begin{equation}
m_1 \ddot{x_1} = -k_1 x_1 + k_2 (x_2-x_1)
\end{equation}



\begin{equation}
m_2  \ddot{x_2} = -k_3 x_2 - k_2 (x_2-x_1)
\end{equation}

donde $\ddot{x} =\frac{\mathrm{d^2}x}{\mathrm{d}t^2}$.

Las ecuaciones anteriores, son ecuaciones diferenciales lineales de segundo orden y acopladas, que pueden ser escritas de forma matricial como $m\vec{\ddot{x}}= -A\vec{x}$, donde A, es una matriz simétrica:

\begin{equation}
m
\begin{pmatrix}
\ddot{x_1}  \\
\ddot{x_2}
\end{pmatrix}
=-
\begin{pmatrix}
k_1 + k_2 & -k_2  \\
-k_2       &  k_2+k_3  \\
\end{pmatrix}
\begin{pmatrix}
x_1\\
x_2
\end{pmatrix}
\end{equation}


Para resolver el sistema de ecuaciones acopladas, se puede realizar una transformación, que permite desacoplar el sistema de ecuaciones diferenciales. Sea
 $\vec{\eta}=S\vec{x}$, donde $\eta$ es una nueva variable en el sistema transformado y S una matriz que cumple que $SAS^{-1}=A_d$, siendo $A_d$ una matriz diagonal.

 Se cumple que $\vec{\ddot {\eta}}=S\vec{\ddot{x}}$
 y  $m\vec{\ddot{x}}=A\vec{x}$, multiplicando por una matriz S, a ambos lados de la ecuación $m\vec{\ddot{x}}=-A\vec{x}$ obtenemos:

\begin{equation}
mS\vec{\ddot{x}}=- S A\vec{x}
\end{equation}

\begin{equation}
m \vec{\ddot{\eta}} = -S A S^{-1}S\vec{x}
\end{equation}


\begin{equation}
m \vec{\ddot{\eta}} = -(S A S^{-1})\eta
\end{equation}


\begin{equation}
m \vec{\ddot{\eta}} = -A_d\eta
\end{equation}

\begin{equation}
m\begin{pmatrix}
\ddot{\eta_1}\\
\ddot{\eta_2}
\end{pmatrix}
=
-\begin{pmatrix}
\lambda_1 & 0\\
0          &\lambda_2
\end{pmatrix}
\begin{pmatrix}
\eta_1\\
\eta_2
\end{pmatrix},
\end{equation}

donde $\lambda=m\omega^2$ son los autovalores de A. Notese que S es la matriz formada por los autovectores de A. La ecuacion anterior, tiene solución en términos de funciones armónicas.


Basado en lo anterior, y resolviendo  el problema más sencillo en el que $k_1 = k_3 = k$, tenemos que:


\begin{equation}
\begin{pmatrix}
\frac{k + k_2}{m}-\omega^2 & -\frac{k_2}{m}\\
 -\frac{k_2}{m}          &\frac{k_2 + k}{m}-\omega^2
\end{pmatrix}
\begin{pmatrix}
x_1\\
x_2
\end{pmatrix}
=
-\begin{pmatrix}
0\\
0
\end{pmatrix}
\end{equation}


\begin{equation}
det(A-\omega^2 I)= \left(\frac{k + k_2}{m}-\omega^2\right) \left(\frac{k_2+k}{m}-\omega^2\right )-\frac{k_2^2}{m^2}=0
\end{equation}

Resolviendo para $\omega^2$, tenemos que :

\begin{equation}
\omega^2_1=\frac{k+2k_2}{m}
\end{equation}


\begin{equation}
\omega^2_2=\frac{k}{m},
\end{equation}


los autovecres vienen dados por:

\begin{equation}
\vec{v_1}=\frac{1}{\sqrt{2}}
\begin{pmatrix}
1\\
-1
\end{pmatrix}
\end{equation}


\begin{equation}
\vec{v_2}=\frac{1}{\sqrt{2}}
\begin{pmatrix}
1\\
1
\end{pmatrix}
\end{equation}

y la matriz S, viene dada por:


\begin{equation}
S=\frac{1}{\sqrt{2}}
\begin{pmatrix}
1 & -1\\
1 & 1
\end{pmatrix}
\end{equation}

las variables $x_1,x_2$, en terminos de $\eta$ están dadas por:

\begin{equation}
x_1=\frac{1}{\sqrt{2}}(\eta_1+\eta_2)
\end{equation}


\begin{equation}
x_2=\frac{1}{\sqrt{2}}(\eta_2+\eta_1)
\end{equation}

En conclusión el problema puede ser resuelto, si solucionamos la
ecuaciones diferenciables desacopladas:
\begin{equation}
\ddot{\eta_1}+\omega_1\eta_1=0
\end{equation}

\begin{equation}
\ddot{\eta_2}+\omega_2\eta_2=0
\end{equation}

 con:

 \begin{equation}
\omega^2_1=\frac{k+2k_2}{m}
\end{equation}


\begin{equation}
\omega^2_2=\frac{k}{m}
\end{equation}

```python
import numpy as np
from scipy.integrate import odeint

def fun(y, t, params):
    m1 = params[0]
    k1 = params[1]
    k2 = params[2]
    k3 = params[3]

    omega = k1 / m1
    omega2 = k2 / m1
    omega3 = k3 / m1

    dxdt1 = y[1]
    dvdt1 = -omega * y[0] + omega2 * (y[2] - y[0])

    dxdt2 = y[3]
    dvdt2 = -omega3 * y[2] - omega2 * (y[2] - y[0])

    return np.array([dxdt1, dvdt1, dxdt2, dvdt2])

def solve_equation(t, posi, veli, params):
    y0 = np.array([posi, veli, posi, veli])  # Include initial conditions for both oscillators
    sol = odeint(fun, y0, t, args=(params,))
    return sol

posi = 2
veli = 1E-20
m = 1
k1 = 1
k2 = 2
k3 = 3
t = np.linspace(0, 50, 1000)
params = (m, k1, k2, k3)

sol = solve_equation(t, posi, veli, params)
```

```python

def eigenproblem(k1, k2, k3, m1, m2):
    omega1 = np.sqrt((k1 + k3) / m1)
    omega2 = np.sqrt(k2 / m1)
    omega_matrix = np.diag([omega1, omega2])
    return omega_matrix

def eta_analytical(t, eta0, eta_dot0, omega):
    eta = eta0 * np.cos(omega * t) + eta_dot0 / omega * np.sin(omega * t)
    return eta

def x_analytical(t, x1_0, x2_0, x1_dot0, x2_dot0, omega1, omega2):
    x1 = 0.5 * (x1_0 - x2_0) * np.cos(omega1 * t) + 0.5 * (x1_dot0 - x2_dot0) / omega1 * np.sin(omega1 * t)
    x2 = 0.5 * (x2_0 - x1_0) * np.cos(omega1 * t) + 0.5 * (x2_dot0 - x1_dot0) / omega1 * np.sin(omega1 * t)
    x1 += 0.5 * (x1_0 + x2_0) * np.cos(omega2 * t) + 0.5 * (x1_dot0 + x2_dot0) / omega2 * np.sin(omega2 * t)
    x2 += 0.5 * (x2_0 + x1_0) * np.cos(omega2 * t) + 0.5 * (x2_dot0 + x1_dot0) / omega2 * np.sin(omega2 * t)
    return x1, x2

def coupled_oscillators(y, t, omega_matrix):
    eta1, eta1_dot, eta2, eta2_dot = y
    omega1, omega2 = np.diag(omega_matrix)
    dydt = [eta1_dot,
            -omega1**2 * eta1 + omega2**2 * (eta2 - eta1),
            eta2_dot,
            -omega2**2 * eta2 - omega2**2 * (eta2 - eta1)]
    return dydt

def solve_coupled_oscillators(t, eta1_0, eta1_dot0, eta2_0, eta2_dot0, omega_matrix):
    y0 = [eta1_0, eta1_dot0, eta2_0, eta2_dot0]
    sol = odeint(coupled_oscillators, y0, t, args=(omega_matrix,))
    eta1 = sol[:, 0]
    eta2 = sol[:, 2]
    return eta1, eta2

# Parámetros del sistema
k1 = 1.0
k2 = 2.0
k3 = 1.0
m1 = 1.0
m2 = 1.0

# Resolver el problema de autovalores y autovectores
omega_matrix = eigenproblem(k1, k2, k3, m1, m2)
omega1, omega2 = np.diag(omega_matrix)

# Condiciones iniciales para el modo simétrico
x1_0 = 1.0
x2_0 = 1.0
x1_dot0 = 0.0
```

```python
# Tiempo de integración
t = np.linspace(0, 10, 1000)

# Resolver numéricamente las ecuaciones diferenciales
eta1_num, eta2_num = solve_coupled_oscillators(t, x1_0, 0.0, x2_0, 0.0, omega_matrix)

# Calcular las soluciones analíticas
x1_analytical, x2_analytical = x_analytical(t, x1_0, x2_0, 0.0, 0.0, omega1, omega2)

# Graficar las soluciones numéricas y analíticas
plt.plot(t, eta1_num, label=r'$\eta_1$ (numérico)')
plt.plot(t, eta2_num, label=r'$\eta_2$ (numérico)')
plt.plot(t, x1_analytical, label=r'$x_1$ (analítico)')
plt.plot(t, x2_analytical, label=r'$x_2$ (analítico)')
plt.xlabel('Tiempo')
plt.ylabel('Posición')
plt.legend()
plt.title('Modo simétrico: x1(0) = x2(0) = 1, x1_dot(0) = x2_dot(0) = 0')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def lorenz_attractor(x, y, z, sigma, rho, beta):
    dx = sigma * (y - x)
    dy = x * (rho - z) - y
    dz = x * y - beta * z
    return dx, dy, dz

def runge_kutta_step(x, y, z, h, sigma, rho, beta):
    k1x, k1y, k1z = lorenz_attractor(x, y, z, sigma, rho, beta)
    k2x, k2y, k2z = lorenz_attractor(x + 0.5 * h * k1x, y + 0.5 * h * k1y, z + 0.5 * h * k1z, sigma, rho, beta)
    k3x, k3y, k3z = lorenz_attractor(x + 0.5 * h * k2x, y + 0.5 * h * k2y, z + 0.5 * h * k2z, sigma, rho, beta)
    k4x, k4y, k4z = lorenz_attractor(x + h * k3x, y + h * k3y, z + h * k3z, sigma, rho, beta)
    x += (h / 6.0) * (k1x + 2.0 * k2x + 2.0 * k3x + k4x)
    y += (h / 6.0) * (k1y + 2.0 * k2y + 2.0 * k3y + k4y)
    z += (h / 6.0) * (k1z + 2.0 * k2z + 2.0 * k3z + k4z)
    return x, y, z

def runge_kutta_lorenz(x0, y0, z0, sigma, rho, beta, h, num_steps):
    x_values = [x0]
    y_values = [y0]
    z_values = [z0]

    for _ in range(num_steps):
        x, y, z = runge_kutta_step(x_values[-1], y_values[-1], z_values[-1], h, sigma, rho, beta)
        x_values.append(x)
        y_values.append(y)
        z_values.append(z)

    return x_values, y_values, z_values

# Parámetros del atractor de Lorenz
sigma = 10.0
rho = 28.0
beta = 8.0 / 3.0

# Condiciones iniciales
x0, y0, z0 = 0.1, 0.0, 0.0

# Tamaño del paso y número de pasos
h = 0.01
num_steps = 10000

# Ejecutar el método de Runge-Kutta para el atractor de Lorenz
x_values, y_values, z_values = runge_kutta_lorenz(x0, y0, z0, sigma, rho, beta, h, num_steps)

# Graficar el resultado
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(x_values, y_values, z_values, lw=0.5)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Atractor de Lorenz')
plt.show()
```
