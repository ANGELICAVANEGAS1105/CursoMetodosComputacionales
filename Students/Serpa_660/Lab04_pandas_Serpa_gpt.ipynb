```python
url3 = 'https://raw.githubusercontent.com/hernansalinas/autogrades/main/Laboratorios_Taller/dataset/country_vaccinations.csv'

df3 = pd.read_csv(url3)
df3
```

```python
colombia=df3.loc[:,"country"]=="Colombia" #filtrar solo para colombia

colombia.copy().to_csv("vacunacion_colombia.csv")
```

```python
paises = df3[df3.duplicated(['country'])]['country'].unique()
paises
```

```python
vacunados = df3.groupby("country")["total_vaccinations"].sum()

latam= ["Argentina","Bolivia","Brazil","Chile", "Colombia", "Costa Rica","Ecuador","Paraguay","Peru","Uruguay","Venezuela"]
vacunados_pais = vacunados.loc[latam].sort_values(ascending=False)
vacunados_pais
```
