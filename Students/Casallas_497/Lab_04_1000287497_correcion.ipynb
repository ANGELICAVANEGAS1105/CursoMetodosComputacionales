7. Realizar una visualización de los datos anteriores.Emplee la libreria  seaborn con un grafico tipo barplot. Ej.
   ax = sns.barplot(x = "contador", y="Primary Type", data = datos)

```python
plt.figure(figsize=(24,12))
info = pd.DataFrame(df.groupby("Primary Type")["Date"].count())
info = info.reset_index()
info.columns = ["Primary Type", "contador"]


ax = sns.barplot(x="contador", y="Primary Type", data=info)
ax.set(xlabel="Contador de casos", ylabel="Tipo de crimen")
plt.show()
```

3. Realizar una comparacion con los paises latinoamericanos de la cantidad de vacunados.




```python
df_vacunas = df_v.groupby("country")["total_vaccinations"].sum()

latinoamerica= ["Argentina","Bolivia","Brazil","Chile", "Colombia", "Costa Rica","Ecuador","Paraguay","Peru","Uruguay","Venezuela"]
df_latinoamerica = df_vacunas.loc[latinoamerica].sort_values(ascending=False)
df_latinoamerica
```
