El método de Euler se puede representar matemáticamente de la siguiente manera:

\begin{align*}
\frac{dy}{dx} &= f(x,y) \\
y(x_0) &= y_0
\end{align*}

donde $f(x,y)$ es la función que define la derivada de $y$ con respecto a $x$, $y_0$ es el valor inicial de $y$ en $x_0$ y $\frac{dy}{dx}$ es la derivada de $y$ con respecto a $x$.

La solución aproximada del método de Euler se calcula iterando sobre los siguientes pasos:

- Definir el tamaño del paso $h$.
- Inicializar $y(x_0) = y_0$.
- Calcular el valor de la función derivada $f(x,y)$ en el punto $(x,y)$.
- Calcular el siguiente valor de $y$ como $y_{i+1} = y_i + h\cdot f(x_i,y_i)$.
- Incrementar $x_i$ en $h$ y repetir el paso 3 y 4 hasta alcanzar el punto deseado.

Es importante tener en cuenta que el método de Euler es un método de primer orden, lo que significa que el error de aproximación disminuye proporcionalmente al tamaño del paso $h$. Por lo tanto, es necesario ajustar el tamaño del paso para obtener una solución precisa.

Además, el método de Euler puede presentar problemas de estabilidad numérica si el tamaño del paso $h$ es demasiado grande. En estos casos, pueden ser necesarios métodos más sofisticados como el método de Runge-Kutta.


# Actividad 02, Oscilador armónico 
1. Resolver la siguiente ecuacion difirencial, empleando el metodo de Euler.

\begin{equation}   
m \frac{d^2x}{dt^2}+kx + bv=0 
\end{equation}


3. Construir un codigo en python a traves de la IA, realizando preguntas adeacuadoas para para  dar la solución la ecuación diferencial, compare los  resultados.

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

# Definir la ecuación diferencial del oscilador armónico
def oscilador(y, t, k, m):
    x, v = y
    dydt = [v, -k/m*x]
    return dydt

# Definir las condiciones iniciales
x0 = 1.0  # posición inicial
v0 = 0.0  # velocidad inicial
y0 = [x0, v0]

# Definir los parámetros del oscilador
k = 1.0   # constante del resorte
m = 1.0   # masa del objeto

# Definir los tiempos a evaluar
t = np.linspace(0, 10, 101)

# Resolver la ecuación diferencial
sol = odeint(oscilador, y0, t, args=(k, m))

# Graficar la solución
plt.plot(t, sol[:, 0], 'b', label='posición')
plt.plot(t, sol[:, 1], 'g', label='velocidad')
plt.legend(loc='best')
plt.xlabel('Tiempo')
plt.grid()
plt.show()
```
