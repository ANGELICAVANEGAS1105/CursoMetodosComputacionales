# 3. Movimiento de proyectiles en presencia de viento.

Supongamos que un proyectil es lanzado desde el origen con una rapidez $v=vo( \cos \theta , \sin \theta $. Supongamos además que el efecto del viento sobre el proyectil es el de una fuerza proporcional a la velocidad relativa entre el viento y el proyectil. Teniendo presente que si u es la velocidad del viento y $\alpha$ el ángulo que hace dirección de esta velocidad con el eje x, se puede entrontrar las siguientes ecuaciones diferenciales, asociada al estado de movimiento del [proyectil](https://iopscience.iop.org/article/10.1088/0143-0807/36/2/025016/meta):


\begin{equation}
m\frac{dv_x}{dt} = -mb(v_x-u\cos \alpha )
\end{equation}

\begin{equation}
m\frac{dv_y}{dt} = -mg-mb(v_x-u\sin \alpha ),
\end{equation}

donde b es el párametro de rozamiento. La soluciones a las ecuaciones anteriores vienen dadas por:

Posición en x
\begin{equation}
x=(u\cos \alpha) t + \frac{1}{b} (v_0\cos \theta_0 -u \cos\alpha)(1-\mathrm{exp}(-bt))
\end{equation}

Posicion en y 
\begin{equation}
y= \frac{1}{b}\left(\frac{g}{b}+v_0 \sin \theta_0 -u\sin\alpha \right)(1-\mathrm{exp}(-bt))-\left( \frac{g}{b}-u\sin\alpha\right)t
\end{equation}

# **2. Ceros de funciones**

**Rutina para los distintos métodos:**

````python
#MÉTODO DE LA BISECCIÓN
def biseccion(a,b,f,error):
  p = (a+b)/2
  if abs(f(p)) <= error:
    return p
  elif f(p)*f(a) > 0:
    a = p
    return biseccion(a,b,f,error)
  elif f(p)*f(a) < 0:
    b = p
    return biseccion(a,b,f,error)

#MÉTODO DE NEWTON RAPHSON
def newton(x0,f,fprime,error):
  p = x0 - f(x0)/fprime(x0)
  if abs(f(p)) <= error:
    return p
  else:
    x0 = p
    return newton(x0,f,fprime,error)

#MÉTODO DEL PUNTO FIJO con scipy
def pfijo(x0,f,error):  
  g = lambda x0: x0-f(x0)
  return sp.optimize.fixed_point(g,x0,xtol=error) 

#MÉTODO DE LA SECANTE
def secante(x0,x1,f,error):
  p = x1 - ( f(x1)*(x1-x0) )/( f(x1) - f(x0) )
  if abs(f(p)) <= error:
    return p
  else:
    x0 = p
    return secante(x0,x1,f,error)
````

# Bisección, Newton y punto fijo para:
 
  $e^{x}-2 - \cos(e^x-2) = 0$

````python
#DEFINICIÓN DE LA FUNCIÓN
f = lambda x: np.exp(x) - 2 - np.cos(np.exp(x)-2)            #Definimos la función
fprime = lambda x: np.exp(x) + np.exp(x)*np.sin(np.exp(x)-2) #Definimos la derivada (Necesario para Newton-Raphson)
dom = np.linspace(-3,3,100)                                  #Definimos un dominio
error = float(1e-10)                                         #Definimos un error
print(f'Para un error de {error}:\n')

#MÉTODO DE LA BISECCIÓN ---------------------------------------------------------------------------
a = -3
b = 2
bisection_zero = biseccion(a,b,f,error)   #Cero hallado con bisección
print(f'Raíz hallada con el método de la bisección: {bisection_zero}')
#----Graficación----
plt.figure(figsize=(7,3))
plt.title(f'Método de la bisección con error = ${error}$')
plt.hlines(0, dom[0], dom[-1],color='k')
plt.plot(dom,f(dom), label='$f(x) = ex−2-cos(ex−2)$')
plt.plot(bisection_zero,f(bisection_zero),'o',color='r', label=(f'Cero en $f(x) ≈{round(bisection_zero,3)}$'))
plt.xlabel('$x$')
plt.ylabel('$f(x)$')
plt.legend()
plt.grid()

#MÉTODO DE NEWTON-RAPHSON: ---------------------------------------------------------------------------
x0 = 0.5
newton_zero = newton(x0,f,fprime,error)   #Cero hallado con Newton-Raphson
print(f'Raíz hallada con el método de Newton-Raphson: {newton_zero}')
#----Graficación----
#plt.figure(figsize=(7,3))
#plt.title(f'Método de Newton-Raphson con error = ${error}$')
#plt.hlines(0, dom[0], dom[-1],color='k')
#plt.plot(dom,f(dom), label='$f(x) = ex−2-cos(ex−2)$',color='g')
#plt.plot(newton_zero,f(newton_zero),'o',color='r', label=(f'Cero en $f(x) ≈{round(newton_zero,3)}$'))
#plt.xlabel('$x$')
#plt.ylabel('$f(x)$')
#plt.legend()
#plt.grid()

##MÉTODO DE PUNTO FIJO scipy: ---------------------------------------------------------------------------
x0=0.5
pfijo_zero = pfijo(x0,f,error)              #Cero hallado con Punto fijo
print(f'Raíz hallada con el método del punto fijo: {pfijo_zero}')
#----Graficación----
#plt.figure(figsize=(7,3))
#plt.title(f'Método de Punto fijo con error = ${error}$')
#plt.plot(dom,f(dom), label='$f(x) = ex−2-cos(ex−2)$',color='purple')
#plt.plot(pfijo_zero,f(pfijo_zero),'o',color='r',label=f'Cero en {pfijo_zero}')
#plt.hlines(0, dom[0], dom[-1],color='k')
#plt.plot(pfijo_zero,f(pfijo_zero),'o',color='r')
#plt.ylabel('$f(x)$')
#plt.xlabel('$x$')
#plt.grid()
#plt.legend()

##MÉTODO DE LA SECANTE: ---------------------------------------------------------------------------
x0 = 0
x1 = 2
secante_zero = secante(x0,x1,f,error)   #Cero hallado con secante
print(f'Raíz hallada con el método de la secante: {secante_zero}')
#----Graficación----
#plt.figure(figsize=(7,3))
#plt.title(f'Método de la secante con error = ${error}$')
#plt.hlines(0, dom[0], dom[-1],color='k')
#plt.plot(dom,f(dom), label='$f(x) = ex−2-cos(ex−2)$')
#plt.plot(secante_zero,f(secante_zero),'o',color='r', label=(f'Cero en $f(x) ≈{round(secante_zero,3)}$'))
#plt.xlabel('$x$')
#plt.ylabel('$f(x)$')
#plt.legend()
#plt.grid()

print('---------------------------------------------------')
````

# **4. Área**

Hallar el área de un segmento circular cuya area sea igual a una cuarta parte del área del circulo.

````python
import pandas as pd
````

````python
dom = np.linspace(-2*np.pi,2*np.pi,1000) #Definimos un dominio
df = pd.DataFrame({'theta':dom})

circulo = lambda ang,r=1: ang*(r**2)/2          #Definimos el área en función del ángulo para un radio r=1
triangulo = lambda ang,r=1: (r**2)*np.sin(ang/2)*np.cos(ang/2) #Definimos el área del triángulo formado en un segmento circular en función del ángulo

segmento = lambda ang,r=1: circulo(ang) - triangulo(ang)  #Definimos el área del segmento circular como el del circulo menos triangulo
cuarto_de_circulo = circulo(np.pi/2,1)    #Definimos el área que tendría un cuarto de círculo

traslacion = lambda ang,r=1: segmento(ang) - cuarto_de_circulo #Trasladamos el segmento menos el cuarto de círuculo para que el cero quede en donde conviene
df['F_theta'] = traslacion(df['theta'])

error = float(1e-10) #Definimos un error

#HALLAR CEROS

#Bisección
bisection_zero = biseccion(2, 2.5, traslacion, error)

#Newton-Raphson                                                                                    #Derivada de la traslación necesaria para Newton-Raphson
derivada = lambda ang,r=1: (r**2)/2 - ( (r**2)*(1/2)*(np.cos(ang/2)**2) - (r**2)*(1/2)*(np.sin(ang/2)**2) ) #Derivada de circulo menos derivada de triangulo
newton_zero = newton(2,traslacion,derivada,error)

#Punto fijo
pfijo_zero = pfijo(0.5,traslacion,error)

#Secante
secante_zero = secante(2,2.5,traslacion,error)

print(f'Con un error de {error}, Área de un cuarto de círculo hallada en:')
print(f'{bisection_zero} rad, con Bisección')
print(f'{newton_zero} rad, con Newton-Raphson')
print(f'{pfijo_zero} rad, con Punto Fijo')
print(f'{secante_zero} rad, con Secante')

print('-----------------------------------------------------------------------------------')

#Graficación
plt.title('Área de segmento circular en función del ángulo')
plt.xlabel('Ángulo en radianes')
plt.ylabel('Área del segmento de círculo')
plt.plot(df['theta'],df['F_theta'],label='Segmento de circulo')
plt.hlines(0, df.iloc[0]['theta'], df.iloc[-1]['theta'], color='k')
plt.plot(bisection_zero, traslacion(bisection_zero),'o',color='red', label=f'Área de un cuarto de círculo en {round(bisection_zero,3)} rad')
plt.grid()
plt.legend()

df #Mostrar el DataFrame
````
