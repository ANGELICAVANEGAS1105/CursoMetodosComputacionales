El método de Euler es un método numérico para aproximar la solución de una ecuación diferencial ordinaria de primer orden. La idea básica del método es descomponer la ecuación diferencial en pequeños incrementos, y aproximar la solución en cada incremento como una recta tangente a la curva solución en ese punto. De esta forma, se puede calcular la solución en un punto dado a partir de la solución en el punto anterior.

Para aplicar el método de Euler, se requiere conocer el valor de la función y su derivada en un punto inicial. A partir de estos valores, se pueden aproximar los valores de la función en puntos posteriores mediante la siguiente fórmula:

y_(i+1) = y_i + f(x_i,y_i) * h

donde:

- y_i es el valor de la función en el punto i
- y_(i+1) es el valor aproximado de la función en el punto i+1
- f(x_i,y_i) es la derivada de la función en el punto i
- h es el tamaño del paso, es decir, la distancia entre los puntos i e i+1

A continuación, te proporciono un código Markdown para Python que implementa el método de Euler para aproximar la solución de una ecuación diferencial ordinaria de primer orden:

    # f es la función a la que se le aplica el método

    # h es el tamaño del paso
    # n es el número de pasos a tomar

````python
#@title **Decaimiento radioactivo** \begin{equation}\frac{\mathrm{d}N}{\mathrm{d}t}=-λ N\end{equation}

#MÉTODO DE EULER
def euler(dNdt, t0, N0, h, n):
    dt = np.zeros(n+1)
    N = np.zeros(n+1)
    # Asignamos los valores iniciales
    dt[0] = t0
    N[0] = N0
    #Iteración del método de Euler
    for i in range(n):
      N[i+1] = N[i] + dNdt(N[i]) *h #Siguiente valor de N está dado por el valor anterior más el valor de la derivada en ese punto por un dt
      dt[i+1] = dt[i] + h
    return dt, N     # Devolvolvemos los valores del tiempo y de N

#DECAIMIENTO RADIOACTIVO:-----------------------------------------

def dNdt(N): return -λ*N

#Condiciones iniciales y constantes
λ = 2 
t0 = 0
N0 = 1000
h = 0.01
n = 100

t, N = euler(dNdt, t0, N0, h, n)

plt.figure(figsize=(5,3))
plt.title('Decaimiento radiactivo')
plt.plot(t,N,'.-')
plt.xlabel('Tiempo [t]')
plt.ylabel('Valor de N')
plt.grid();
````
