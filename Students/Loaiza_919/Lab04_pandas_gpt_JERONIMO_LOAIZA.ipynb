# **Problema 2:**

**Importar datetime**

````python
from datetime import datetime
````

**Importar PS4_1.csv**

````python
df_excel=pd.read_excel('https://docs.google.com/spreadsheets/d/e/2PACX-1vQEVgASJnYYcOLX5RZiJOc8a0q2siq_8Z2OoDR8K4E-LzW2z3NhYqzVCHIeBtwEWQ/pub?output=xlsx')
df_excel.drop('Unnamed: 1', axis=1, inplace=True) #Eliminar columna unnamed1
````

````python
#Cambiar lo que está en formato str para el date y convertirlo a datetime
for d in df_excel['Date']:
    if(type(d)==str):
        df_excel["Date_"] = datetime.strptime(d, "%d/%m/%Y")
    else:
        df_excel["Date_"]=d
df_excel=df_excel.set_index("Date_")
````

````python
df_excel.drop('Date', axis=1, inplace=True)  #Eliminar Date anterior
df_excel.index.name = 'Date'            #Renombrar el nuevo Date
````

````python
df_excel
````

**Diccionario con información**

````python
num_cols = len(df_excel.columns)
col_names = list(df_excel.columns)

no_NaN = df_excel.count() #Número de registros no NaN de cada columna

memoria = df_excel.memory_usage().sum() #Cantidad total de memoria usada para cargar el dataset

estadistica = df_excel.describe() #Estadística básica para cada columna

# Remplazar NaN por None
df_excel['Brebeuf'] = None
none = list(df_excel['Brebeuf'])


#DICCIONARIO
diccionario = {'Cantidad de columnas': num_cols,
             'Nombre de las columnas': col_names,
             'Registros no NaN': no_NaN,
             'Memoria usada': memoria,
             'Estadísticas básicas': estadistica.to_dict(),
             'Valores remplazados': none}

print(diccionario)

#info_dict
````

**Realizar una comparacion con los paises latinoamericanos de la cantidad de vacunados.**

````python
df_latinoamerica = df[(df['country'] == 'Colombia') | (df['country'] == 'Mexico') |
                      (df['country'] == 'Brazil') | (df['country'] == 'Argentina') |
                      (df['country'] == 'Peru') | (df['country'] == 'Chile') |
                      (df['country'] == 'Ecuador') | (df['country'] == 'Antigua y Barbuda')]
df_latinoamerica
````

**A traves de mascaras determinar el numero de vacunados en el intervalo [1.5E6, 2.0E6]**

````python
mask_vacunados = (df_latinoamerica['total_vaccinations'] >= 1.5E6) & (df_latinoamerica['total_vaccinations'] <= 2.0E6)
num_vacunados = len(df_latinoamerica[mask_vacunados])
num_vacunados
````
