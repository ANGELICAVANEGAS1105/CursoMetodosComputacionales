## 1. (50%). Movimiento de una esfera de radio r en un fluido.

 Una esfera de radio $R$ que asciende o desciende por un fluido puede ser modelada a través de un modelo de partícula, sometida a tres fuerzas: el peso, el empuje y la fuerza rozamiento. Las  ecuaciones de posición (x) y velocidad (v) para un sistema de refencia úbicado en la base del fluido, pueden ser escritas como:

\begin{equation}
x = \frac{G}{\alpha^2} \left( 1+\frac{\alpha v_0}{G} \right)(1-e^{-\alpha t}) - \frac{G}{\alpha}t
\end{equation}

\begin{equation}
v= \left ( v_0 +\frac{G}{\alpha} \right )e^{-\alpha t} - \frac{G}{\alpha}
\end{equation}



donde $G=\left( 1-\frac{\rho_f}{\rho} \right)g$, con $\rho_f$, $\rho$ la densidad del fluido y la esfera respectivamente y $g=9.8$ la aceleración de la gravedad y $\alpha$  una constante que depende del radio de la esfera, la viscosidad y la densidad.
$\alpha=\frac{9\eta}{2R^2\rho}$.

Suponga que una esferita de radio $R=2\times10^{-5}m$ es lanzada verticalmente hacia arriba en el agua, con una rapidez de $v_0=0.01$m/s.
Tomando la densidad del agua en  $\rho_f =1000 kg/m^3$,  la densidad de la esfera en $\rho=2670 kg/m^3$ y la viscosidad en $\eta=0.001 kg/ms$.  Determine: 

1. El tiempo total en que la esfera asciende  y desciende. 

2. Determine la rapidez final, con la que llega nuevamente de donde fue lanzada. 



Para ello construya un programa en python que:
- Permita calcular la posición como función del tiempo
- Permita calcular la rapidez como función del tiempo
- Cálcule  el tiempo total de subida y bajada empleando el método de la secante.

# **Solución punto 1:**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import optimize
```

```python
#Definimos las constantes
g = 9.8
R = float(2e-5)
v0 = 0.01
rhof = 1000
rho = 2670
n = 0.001

alpha = (9*n)/(2*(R**2)*rho)
G = (1-rhof/rho)*g

#Definimos las funciones
x = lambda t: (G/alpha**2)*(1+(alpha*v0)/G)*(1-np.exp(-alpha*t)) - (G/alpha)*t
v = lambda t: (v0+(G/alpha))*np.exp(-alpha*t)-(G/alpha)

#Definimos un intervalo para el tiempo
t = np.linspace(0,0.003,1000)

#Hallamos cuándo se hace cero la posición con método de la secante de scipy
t_final = optimize.newton(x,6)

print(f'El tiempo que tarda en subir y bajar es de {t_final}\n')
print(f'La rapidez (positiva) con la que llega es, evaluando la velocidad v(t_final) = {abs(v(t_final))}')
print('----------------------------------------------------------------------')

#Gráfica
plt.figure(figsize=(5,3))
plt.title('Posición vs tiempo')
plt.xlabel('Tiempo [t]')
plt.ylabel('Posición (altura)')
plt.hlines(0,t[0],t[-1],color='k')
plt.plot(t,x(t),label=('Posición'))
plt.plot(t_final,x(t_final),'r.',label=(f'Cero en {round(t_final,4)}'))
plt.legend()
plt.grid()
```

```python
import pandas as pd
```

**Leer el DataFrame**

```python
#Importamos el DataFrame
df = pd.read_csv('https://raw.githubusercontent.com/hernansalinas/autogrades/main/Evaluaciones/dataset/Temperature_Bogota_Barcelona.csv')
df
```

**Agregar columna con la temperatura en Celsius**

```python
#Definimos una nueva columna
df['AvgTempCelcius'] = 0

#Hacemos ciclo for para poner la temperatura en celcius a la nueva columna
contador = 0
for d in df['AvgTemperature']:
  df.loc[contador,"AvgTempCelcius"] = (d-32)/1.8 #Se utiliza cada temperatura que estaba en Farenheit para pasarla a Celsius
  contador += 1

df
```

**Gráfico de la temperatura de Bogotá vs los años**

```python
#Definimos listas vacías
temp_bogota = []
año_bogota = []

#Ciclo for para rellenar esas listas
contador = 0
for d in df['City']:                                    #Barremos la columna de la ciudad
  if d == 'Bogota':                                     #Nos interesan los datos solo para Bogotá
    temp_bogota.append(df['AvgTempCelcius'][contador])  #Añadimos la temperatura asociada a cuando se encuentra a Bogotá
    año_bogota.append(df['Year'][contador])             #Añadimos el año asociado a esa temperatura
  contador += 1

#Gráfica
plt.figure(figsize=(7,2.5))
plt.title('Temperatura de Bogotá vs años')
plt.xlabel('Año')
plt.ylabel('Temperatura')
plt.plot(año_bogota,temp_bogota,'--')
plt.plot(año_bogota,temp_bogota,'r.',markersize='6')
plt.grid()
```

**Gráfico de la temperatura de Barcelona vs los años**

```python
#Definimos listas vacías
temp_barcelona = []
año_barcelona = []

#Ciclo for para rellenar esas listas
contador = 0
for d in df['City']:                                       #Barremos la columna de la ciudad
  if d == 'Barcelona':                                     #Nos interesan los datos solo para Barcelona
    temp_barcelona.append(df['AvgTempCelcius'][contador])  #Añadimos la temperatura asociada a cuando se encuentra a Barcelona
    año_barcelona.append(df['Year'][contador])             #Añadimos el año asociado a esa temperatura
  contador += 1

#Gráfica
plt.figure(figsize=(7,2.5))
plt.title('Temperatura de Barcelona vs años')
plt.xlabel('Año')
plt.ylabel('Temperatura')
plt.plot(año_barcelona,temp_barcelona,'g--')
plt.plot(año_barcelona,temp_barcelona,'r.',markersize='6')
plt.grid()
```

**Interpolación de hermite para Bogotá**



```python
from scipy.interpolate import CubicHermiteSpline as hermite
```

```python
#Definimos una derivada numérica con Numpy
dydx = np.diff(temp_bogota)

#Definimos la interpolación
interpolacion_hermite = hermite(año_bogota[1:], temp_bogota[1:], dydx) #Esto utiliza solo hasta lña derivada n-1

#Definimos un linspace con los que trabajará la interpolación
x_bogota = np.linspace(np.min(año_bogota),np.max(año_bogota),100)

#Definimos un linspace con los que trabajará la interpolación
plt.figure(figsize=(7,2.5))
plt.title('Interpolación de Hermite para Bogotá')
plt.xlabel('Año')
plt.ylabel('Temperatura')
plt.plot(x_bogota, interpolacion_hermite(x_bogota))
plt.plot(año_bogota[1:],temp_bogota[1:],'r.',markersize='5')
plt.grid()
```

**Interpolación de Splines Cubicos para Barcelona**

```python
from scipy.interpolate import CubicSpline
```

```python
#Definimos la interpolación
interpolacion_cubic = CubicSpline(año_barcelona, temp_barcelona)

#Definimos un linspace con los que trabajará la interpolación
x_barcelona = np.linspace(np.min(año_barcelona),np.max(año_barcelona),100)

#Gráfica
plt.figure(figsize=(7,2.5))
plt.title('Interpolación de Splines Cúbicos para Barcelona')
plt.xlabel('Año')
plt.ylabel('Temperatura')
plt.plot(x_barcelona, interpolacion_cubic(x_barcelona),color='g')
plt.plot(año_barcelona,temp_barcelona,'r.',markersize='5')
plt.grid()
```
