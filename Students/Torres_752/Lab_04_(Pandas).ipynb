```python
import pandas as pd
enlace="https://raw.githubusercontent.com/hernansalinas/autogrades/main/Laboratorios_Taller/dataset/country_vaccinations.csv"
df_4=pd.read_csv(enlace)


# Filtro para datos de Colombia
colombia_filt=df_4.loc[:,"country"]=="Colombia"
df_4_colombia = df_4.loc[colombia_filt]



df_4_colombia.copy().to_csv("vacunacion_colombia.csv")

df_4_colombia
```

```python

latin_america = ['Argentina', 'Bolivia', 'Brazil', 'Chile', 'Colombia', 'Costa Rica', 'Cuba', 'Dominican Republic', 'Ecuador', 'El Salvador', 'Guatemala', 'Haiti', 'Honduras', 'Jamaica', 'Mexico', 'Nicaragua', 'Panama', 'Paraguay', 'Peru', 'Puerto Rico', 'Trinidad and Tobago', 'Uruguay', 'Venezuela']

df_4_0=df_4.copy()

#for i in latin_america:
 # latam_filt= df_4.loc[:,"country"]==i
  #df_4_latam=df_4.loc[latam_filt]


df_4_0
```

```python

# Filtro para paises latinoamericanos:

latam_filt = df_4_0.country.isin(latin_america)
df_4_latam = df_4_0[latam_filt]
df_4_latam
```

```python
#Elaboramos un filtro para obtener las filas correspondientes a un pais especifico
f1  =  (df_4_latam["country"] == "Colombia")

#De la nueva tabla filtrada, sacamos la columna de personas vacunadas:
df_4_latam[f1]["people_vaccinated"]

# Y obtenemos el valor maximo o total de personas vacunadas para ese pais
df_4_latam[f1]["people_vaccinated"].max()
```

```python

latin_america = ['Argentina', 'Bolivia', 'Brazil', 'Chile', 'Colombia', 'Costa Rica', 'Cuba', 'Dominican Republic', 'Ecuador', 'El Salvador', 'Guatemala', 'Haiti', 'Honduras', 'Jamaica', 'Mexico', 'Nicaragua', 'Panama', 'Paraguay', 'Peru', 'Puerto Rico', 'Trinidad and Tobago', 'Uruguay', 'Venezuela']
personas_vacunadas=[]

for i in latin_america:

  filt=(df_4_latam["country"] == i)
  n_i = df_4_latam[filt]["people_vaccinated"].max()
  personas_vacunadas.append(n_i)
  


personas_vacunadas



```

```python
# Finalmente, construioms un diccionario cuya clave son los paises latinoamericanos y cuyo valor es el numero de vacunados por cada pais

dict_from_list={}
for j in range(len(latin_america)):
  dict_from_list[latin_america[j]]=personas_vacunadas[j]

dict_from_list
```
