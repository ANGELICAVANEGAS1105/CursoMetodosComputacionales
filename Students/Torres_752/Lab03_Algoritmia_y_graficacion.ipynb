````python
# Definicion de la función :


import numpy as np

def logistic_map(xi,r,n):
  X_= np.zeros(n)
  X_[0]  = xi

  for i in range(1,n):
    x_f = r*xi*(1-xi)
    xi= x_f

    X_[i] = x_f

  return X_

logistic_map(0.8,3.1,100)



  

````

2.0 El algoritmo de gradiente descendente permite encontrar el minimo de funciones en espacios multidimensionales proponiendo puntos en la direccion en la que el gradiente desciende. Para el caso de una dimensión, podemos encontrar el mínimo realizando iteraciones y multiplicando su derivada en cada nuevo punto por una constante $\alpha$

\begin{equation}
x_{i+1}=x_i - \alpha \frac{\mathrm{d}f}{\mathrm{d}x}(x_i)
\end{equation}




Ejemplo:

Sea $f(\omega) = (\omega-1)^2$ , con $\omega \in \mathbb{R} $. El minimo de esta función esta determinado por
\begin{equation}
\frac{\mathrm{d}f}{\mathrm{d}\omega} = 0
\end{equation}

\begin{equation}
\omega = 1
\end{equation}


Para determinar el minimo,  a través del gradiente descendente puede ser aplicada el siguiente algoritmo:
1. Proponer un número aleatorios inicial $\omega_i$
2. Para descender al mínimo de la función, encontremos un  valor para el cual 
la derivada de la función descenciende, asi:
\begin{equation}
\omega_{i+1} = \omega_{i} - \alpha \frac{\mathrm{d}f(\omega_i)}{\mathrm{d}\omega}
\end{equation}

donde, $\alpha$ es conocido como la tasa de aprendizaje del algoritmo. 

3. Evaluar $f(\omega_{i+1})$

4. Iterar hasta encontrar el minimo de la función 


En el caso de la funcion propuesta, asumiendo un valor de $\alpha=1$, tenemos que:

1. Supongamos que $\omega_{0} = 2$, luego :
2. $\omega_{0+1} = 2 - 1 f'(\omega_{0}) =  = 0$
3. $f(0) = 1$
4. $\omega_{1+1} = 0 - 1 f'(\omega_{1}) = 0-1(2(0-1)) =  2$

El valor del $\omega$  despues de un par de iteraciones, no es el esperado, por que no minimiza la función, se debe proponer un tasa de aprendizaje  **$\alpha$** de tal forma que permita descender suavemente para encontrar el minimo.

3. Analiza como se minimiza la función, gráficando de la funcion en cada punto a iterar,  como funcion del numero de iteraciones. Para dos valores diferentes de la constante alpha (prueba con $\alpha=0.1$, $\alpha=0.01$)


````python
import numpy as np
from numpy import random

import matplotlib.pylab as plt

f=lambda x: x**2          # Definimos la función f(x) = x^2.
g=lambda x: 2*x           # Definimos la derivada de f como : g(x) = 2x.
x=np.linspace(-10,10,100) # Construimos un dominio para las dos funciones f y g.
             




def grad_descendente(x_ini, alpha, precision):
  x_lista =[x_ini]                              # Inicializamos una lista vacía que contendrá los sucesivos valores de entrada obtenidos en cada iteración.
  y_lista = [f(x_ini)]                          # Inicializamos una lista vacía que contendrá las imágenes de los valores de la lista que está justo  arriba.
  x=np.linspace(-3,3,100)                       # Construimos un dominio para las dos funciones f y g.

  while True:
    xn= x_ini - alpha*g(x_ini)
    x_ini = xn
  
    x_lista.append(xn)
    y_lista.append(f(xn))

    if abs( x_lista[-1]  - x_lista[-2] ) <= precision :
      break

  
  print(" El mínimo de f(x) ocurriría en " + str(x_ini))
  print(" Número de iteraciones: "+ str(len(x_lista)))




  plt.figure(figsize = (10,10))

  plt.subplot(2,1,1)
  plt.grid()
  plt.plot(x,f(x), c= "g", label = " $f(x) = x²$")
  plt.scatter(x_lista,y_lista, c= "k")
  plt.plot(x,np.zeros(len(x)), c="k", linewidth = 1)

  plt.xlabel("$x$", fontsize = 14)
  plt.ylabel("$y$", fontsize = 14)

  plt.title("Gradiente descendente para $ f(x) = x^2 $", fontsize= 16 )

  plt.xticks(fontsize = 14)
  plt.yticks(fontsize = 14)

   
  plt.legend(loc = "best") 
  plt.show

  
  #------------------------------------------------------

  plt.subplot(2,1,2)
  
  plt.plot(x,g(x), c="b", label = " $g(x) = f´(x) = 2x$")
  plt.plot(x,np.zeros(len(x)), c="k", linewidth = 1)

  plt.xlabel("$x$", fontsize = 14)
  plt.ylabel("$y$", fontsize = 14)

  plt.xticks(fontsize = 14)
  plt.yticks(fontsize = 14)


  plt.grid()
  plt.legend(loc = "best") 
  plt.show




  

````

````python


def resultados_iter(alpha):

  x=np.linspace(-10,10,100)
  x0=1         # Escogemos un valor arbitrario de x, en el intervalo definido.
  x_iter=[]    # Inicializamos una lista vacía que contendrá los sucesivos valores de entrada obtenidos en cada iteración.
              # Inicializamos una lista vacía que contendrá las imágenes de los valores de la lista que está justo  arriba.

  for i in range(1,len(x)):
    xn= x0 - alpha*g(x0)
    x_iter.append(xn)
    x0=xn

  x_iter=np.array(x_iter)
  return x_iter
````

````python

plt.figure(figsize = (12,5))


plt.subplot(1,2,1)
  
plt.scatter(range(1,len(x)), resultados_iter(0.1), color="purple", label = " $ alpha = 0.1$")


plt.xlabel("iteraciones , $ n$ ", fontsize = 14)
plt.ylabel("$ x_n$", fontsize = 14)

plt.xticks(fontsize = 14)
plt.yticks(fontsize = 14)


plt.grid()
plt.legend(loc = "best") 
plt.show

#-----------------------------------------------------------------------------

plt.subplot(1,2,2)
  
plt.scatter(range(1,len(x)), resultados_iter(0.01), c="g", label = " $ alpha = 0.01$")


plt.xlabel("iteraciones , $ n$ ", fontsize = 14)
plt.ylabel("$ x_n$", fontsize = 14)

plt.xticks(fontsize = 14)
plt.yticks(fontsize = 14)


plt.grid()
plt.legend(loc = "best") 
plt.show
````
