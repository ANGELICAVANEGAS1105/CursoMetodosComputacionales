**Método de Euler**

Es un método numérico para resolver ecuaciones diferenciales ordinarias de primer orden. Su objetivo es aproximar la solución de una ecuación diferencial en pequeños intervalos de tiempo, utilizando la derivada de la solución en un punto dado.

**El método de Euler se puede dividir en los siguientes pasos:**

* Seleccionar el valor inicial de la variable dependiente y el valor 
inicial de la variable independiente.
* Calcular la derivada de la función en el punto inicial.
* Utilizar la derivada calculada para aproximar el valor de la función en el siguiente punto de la variable independiente.
* Utilizar el valor calculado en el paso anterior como valor inicial para el siguiente paso, y repetir los pasos 2 y 3 para el siguiente punto de la variable independiente.

* Continuar repitiendo los pasos 2, 3 y 4 hasta alcanzar el valor final deseado de la variable independiente.



**Deducción Matemática**

Supongamos que queremos resolver la ecuación diferencial ordinaria de primer orden:

\begin{equation}
\frac{dy}{dx} = f(x,y)
\end{equation}

con condición inicial $y(x_0) = y_0$. Para aplicar el método de Euler, dividimos el intervalo de integración $[x_0, x_f]$ en $N$ subintervalos iguales de tamaño $h$, tal que $x_i = x_0 + i \cdot h$ y $y_i \approx y(x_i)$. Entonces, podemos aproximar la solución de la ecuación diferencial en el punto $x_{i+1} = x_i + h$ utilizando la derivada en el punto $x_i$:

\begin{equation}
y_{i+1} = y_i + f(x_i, y_i) \cdot h + O(h^2)
\end{equation}

donde $O(h^2)$  representa el error de aproximación en la solución debido a la discretización del intervalo de integración.

La aproximación anterior se puede obtener utilizando una expansión de Taylor de primer orden de la solución alrededor del punto $x_i$:

\begin{equation}
y(x_{i+1}) = y(x_i + h) = y(x_i) + h \cdot \frac{dy}{dx} \Big|_{x=x_i} + O(h^2)
\end{equation}

Sustituyendo la ecuación diferencial en la expresión anterior, tenemos:

\begin{equation}
y(x_{i+1}) = y(x_i) + h \cdot f(x_i, y_i) + O(h^2)
\end{equation}

Aproximando $y(x_{i+1})$ por $y_{i+1}$ y $y(x_i)$ por $y_i$, obtenemos la ecuación del método de Euler:

\begin{equation}
y_{i+1} = y_i + f(x_i, y_i) \cdot h + O(h^2)
\end{equation}

La ecuación del método de Euler muestra que el valor de la función $y$ en el punto $x_{i+1}$ se puede aproximar utilizando el valor de la función $y$ en el punto $x_i$ y la derivada de la función en el punto $x_i$. El término del error de truncamiento indica que el error de aproximación disminuye con el tamaño del intervalo de integración $h$.

```python
#y_i es la condición inicial
#delta es el ancho de cada subintervalo 
#N es la cantidad de veces q se repite 

def EulerM(y_i,delta,N,λ):
  y_i1=y_i
  aRR=np.zeros(N)
  for i in range(N):
     y_i1=y_i-λ*y_i*delta
     aRR[i]=y_i1
     y_i=y_i1
  return aRR
```

```python
N=EulerM(100,1,100,0.1)
```

```python
#@title Librerías 
import numpy as np
import matplotlib.pylab as plt 
```

```python
def funcion(y0,v0,delta,w,N,b):
  yi=np.zeros(N)
  vi=np.zeros(N)
  for i in range(N):
     y_ii=y0+v0*delta
     a = -w*y_ii- b*v0
     v_ii = v0 + a*delta
     yi[i]=y_ii
     vi[i]=v_ii
     y0=y_ii
     v0=v_ii
  Array=np.array([yi,vi])
  return Array
```

```python
#Oscilador amortiguado 
def euler_oscillator(m, b, k, x0, v0, t0, tf, delta_t):
    # Definir las condiciones iniciales
    x = x0
    v = v0
    
    # Definir los arreglos para almacenar la posición y velocidad en cada iteración
    position = [x]
    velocity = [v]
    
    # Definir el tiempo inicial y agregarlo al arreglo t
    t = t0
    t_values = [t]
    
    # Iterar usando el método de Euler
    while t < tf:
        # Calcular la aceleración
        a = (-b*v - k*x)/m
        
        # Calcular la nueva posición y velocidad usando el método de Euler
        x = x + v*delta_t
        v = v + a*delta_t
        
        # Agregar la posición y velocidad a los arreglos
        position.append(x)
        velocity.append(v)
        
        # Actualizar el tiempo y agregarlo al arreglo t
        t += delta_t
        t_values.append(t)
        
    return np.array([t_values, position, velocity])

# Parámetros del oscilador
m = 1
b = 0.1
k = 1
x0 = 1
v0 = 0

# Intervalo de tiempo y paso
t0 = 0
tf = 100
delta_t = 0.01

# Resolver la ecuación diferencial
solution = euler_oscillator(m, b, k, x0, v0, t0, tf, delta_t)

# Extraer los resultados de la posición y velocidad
t = solution[0]
position = solution[1]

# Graficar la posición vs tiempo
plt.figure(figsize=(5,5))
plt.plot(t, position,color='green')
plt.title('Posición vs tiempo del oscilador amortiguado')
plt.xlabel('Tiempo')
plt.ylabel('Posición')
plt.grid()
plt.show()
```

Las dos gráficas evidencian que tanto la funcion de la IA como la mía muestran el comportamiento esperado en un oscilador  amortiguado por una fuerza de fricción.

A pesar de que los dos códigos realizan la solución por el método de Euler ,hay algunas diferencias en el paso a paso ,por ejemplo la IA usa un while en lugar de for y se itera hasta que llega al tiempo final(tf),los argumentos también son diferentes pues solicita el tiempo inicial,el final y la masa.
