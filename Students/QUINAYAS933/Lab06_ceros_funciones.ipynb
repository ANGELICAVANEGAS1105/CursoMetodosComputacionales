De lo contrario, si la distancia entre a y b es menor que la tolerancia T, entonces se considera que se ha encontrado una buena aproximaci√≥n de la ra√≠z, y el programa tambi√©n sale del bucle.
7. Si el signo de $f(c)$ es igual al signo de $f(a)$, entonces la ra√≠z se encuentra en el intervalo $(a, c)$, por lo que se redefine $a=c$. De lo contrario, la ra√≠z se encuentra en el intervalo $(c, b)$, por lo que se redefine $b=c$.

**b)Deducci√≥n del m√©todo de Newton-Rapshon**

Primero es importante recalcar que el m√©todo d**e Newton-Raphson** converge r√°pidamente a la raiz de la funci√≥n si se cumplen ciertas condiciones ,como que la funci√≥n sea **continuamente diferenciable** y que la derivada no se anule en la vecindad de la raiz.

El m√©todo de Newton-Raphson es un algoritmo **iterativo** utilizado para encontrar  las ra√≠ces de una funci√≥n con errores relativamente peque√±os .
Supongamos que tenemos una funci√≥n  $f(x)$  que es diferenciable en un intervalo  [ùëé,ùëè] . Queremos encontrar una aproximaci√≥n de la ra√≠z de la funci√≥n en este intervalo, es decir, un valor  $x_0$  tal que  $f(x_0)=0$ .

Empezamos con una aproximaci√≥n inicial  $ùë•_1$  y trazamos la tangente a la curva de la funci√≥n  $f(x)$  en el punto  $(ùë•_1,ùëì(ùë•_1))$ . La ecuaci√≥n de esta tangente es:

**$$ùë¶‚àíùëì(ùë•1)=ùëì‚Ä≤(ùë•1)(ùë•‚àíùë•1)$$**

Donde  $f‚Ä≤(x_1)$  es la derivada de la funci√≥n  $f(x)$  evaluada en  $x_1$ .

Ahora, se debe encontrar el punto en el eje  $x$  donde esta tangente cruza el eje  x . Es decir, buscamos el valor de  x_2  tal que  y=0  en la ecuaci√≥n anterior:

$$0‚àíf(x_1)=f‚Ä≤(x_1)(x_2‚àíx_1)$$

Despejando  x_2 , obtenemos:

$$x_2=x_1‚àíf(x_1)f‚Ä≤(x_1)$$

Este es el paso iterativo del m√©todo de Newton-Raphson. A partir de la aproximaci√≥n inicial  x_1 , podemos obtener la siguiente aproximaci√≥n  x_2 mediante la f√≥rmula anterior. Repitiendo este proceso con  x_2  , podemos obtener una mejor aproximaci√≥n  x_3 , y as√≠ sucesivamente.
En general, la f√≥rmula de iteraci√≥n del m√©todo de Newton-Raphson es:
$$x_{n+1}=x_n‚àíf(x_n)f‚Ä≤(x_n)$$ 
Donde  x_n es la aproximaci√≥n  n-√©sima de la ra√≠z de la funci√≥n.






**c)¬øCu√°l es la diferencia del metodo de Newton-Rapson con respecto al de la secante?**

De manera general la principal diferencia entre estos dos m√©todos es que el primero utiliza la informaci√≥n de la derivada de la funcion para llegar a la raiz y el segundo utiliza s√≥lo la informaci√≥n de dos aproximaciones dadas.

por tanto el m√©todo de Newton-Raphson es m√°s preciso que el m√©todo de la secante, ya que utiliza una mejor aproximaci√≥n de la ra√≠z en cada iteraci√≥n. Sin embargo, la precisi√≥n del m√©todo de la secante puede mejorarse aumentando el n√∫mero de iteraciones.



**d. Escriba en pseudoc√≥digo los pasos para aplicar el m√©todo Punto Fijo.**

Dado una funci√≥n $f(x)$ y una aproximaci√≥n inicial $x_0$:
1. Establecer una tolerancia $e$ y un m√°ximo n√∫mero de iteraciones $N$.

2. Establecer $i=0$
3. Mientras $i<N$ entonces :
4. Calcular la aproximaci√≥n $$x_{i+1} = g(x_i)$$

donde $g(x)$ es la funci√≥n de punto fijo 
$$g(x) = x - f(x)$$

5. Si $|x_{i+1} - x_i| < e$, detener el proceso y retornar $x_{i+1}$ como la soluci√≥n.

6. si no establecer $$i = i + 1$$

7. Si el n√∫mero m√°ximo de iteraciones se alcanza sin converger, mostrar un mensaje de error y retornar el valor m√°s reciente de $x_i$ como la soluci√≥n aproximada.

# 2. Ceros de funciones
*Encontrar* los ceros de $e^{x}-2= \cos(e^x-2)$ por diferentes metodos, bisection, newton-rapshon, punto fijo.


 

````python
import numpy as np 
import scipy.optimize as opt
import scipy as sc
import matplotlib.pylab as plt 
opt.bisect
````

````python
def func(x):
  return  np.exp(x)-2-np.cos(np.exp(x)-2)
````

````python
#Hallando el cero por medio del m√©todo de Bisecci√≥n
root=sc.optimize.bisect(func,-1,2)
````

````python
#Hallando el cero por medio del m√©todo Newthon-Rapshon
import sympy as sp
x=sp.symbols("x")
def f1(x):
   return  sp.exp(x)-2-sp.cos(sp.exp(x)-2)
````

````python
#calculando la derivada con sympy
d=sp.diff(f1(x),x)
funcion1 =sp.lambdify([x],f1(x),"numpy")
fderivate=sp.lambdify([x],d,"numpy")
#finalmente ya se puede aplicar el opt.newton
root1= opt.newton(funcion1,1.5, fderivate)
````

````python
#Punto fijo
def g(x):
  g=x-funcion1(x)
  return(g)
````

````python
sol = opt.fixed_point(g, 0.5);sol
````

````python
#Gr√°fica de la funci√≥n con su ra√≠z
x=np.linspace(-1,4,100000)
plt.figure(figsize=(10,10))
plt.plot(x,func(x),color='r')
plt.plot(root,func(root),marker="o",color='blue',label="Ra√≠z")
plt.xlabel("$x$",size=14)
plt.ylabel("$f(x)$",size=14)
plt.legend(loc=2) 
plt.title("Gr√°fica de $e^x - \cos(e^x - 2)-2 $",size=15)
plt.grid(True)
````

````python
print(f'Las raices encontradas por los diferentes m√©todos fueron {root},{root1},{sol }')
````

Supongamos que un proyectil es lanzado desde el origen con una rapidez $v=vo( \cos \theta , \sin \theta $. Supongamos adem√°s que el efecto del viento sobre el proyectil es el de una fuerza proporcional a la velocidad relativa entre el viento y el proyectil. \begin{equation}
m\frac{dv_x}{dt} = -mb(v_x-u\cos \alpha )

\begin{equation}
y= \frac{1}{b}\left(\frac{g}{b}+v_0 \sin \theta_0 -u\sin\alpha \right)(1-\mathrm{exp}(-bt))-\left( \frac{g}{b}-u\sin\alpha\right)t

````python
# Parametros iniciales
import math 
import scipy.optimize as opt
v0 = 60 #Velocidad en m/s
alpha = np.pi #Angulo del viento
theta = np.deg2rad(45) #Angulo de lanzamiento en radianes
u = v0/2             #rapidez del viento
b = 0.05             #Parametro de friccion 
g = 9.8              #Aceleracion de la gravedad m/s2

params = (alpha, v0, b, u,g)  #Parametros en tupla.
````

````python
#Se define dos funciones, una para x y otra para y.
# Posicion en x
def pos_x(t, theta, alpha, v0, b, u,g):
  
  t1=u*np.cos(alpha)*t
  t2=(1/b)*(v0*np.cos(theta)-u*np.cos(alpha))*(1-np.exp(-b*t))
  x = t1+t2
  return x

# Posicion en y
def pos_y(t, theta, alpha, v0, b, u,g):
  
  t1=(1/b)*(g/b+v0*np.sin(theta)-u*np.sin(alpha))*(1-np.exp(-b*t))
  t2=-(g/b-u*np.sin(alpha))*t
  y = t1+t2
  return y 
````

````python
def time(t, theta, alpha, v0, b, u, g=9.8):
    h = t - pos_y(t, theta, alpha, v0, b, u, g)
    return h
````

````python
#se usa m√©toodo del punto fijo para encontrar el tiempo cuando y=0
t_vuelo = opt.fixed_point(time, x0=9.0, args=(theta, alpha, v0, b, u))
t = np.linspace(0, t_vuelo, 1000)
x = pos_x(t,theta, alpha, v0, b, u, g=9.8)
y = pos_y(t,theta, alpha, v0, b, u, g=9.8)
````

````python
import numpy as np
import matplotlib.pylab as plt
from PIL import Image

Npx_X = 80 #Numero de pixeles en x
Npx_Y = 80 #Numero de pixeles en y

image = Image.new("RGB", (Npx_X, Npx_Y)) #Creacion de la imagen inicial
#Image.new crea una imagen con el modo y tama√±o dados

npx = np.arange(0, Npx_X)
npy = np.arange(0, Npx_Y)

xmin = -1
xmax = 1
ymin = -1
ymax = 1
# Transformaci√≥n al plano complejo en un rectangulo (xmin, xmax), (ymin, ymax)
zx = npx * (xmax - xmin) / (Npx_X - 1)   + xmin
zy = npy * (ymax-ymin) / (Npx_Y - 1) + ymin

# Definicion  la funcion y las derivada 
def f(z): 
  return z**3 - 1

def fp(z): 
  return 3*z**2

h = 1E-4
def fpn(z, h): 
  return (f(z+h)-f(z))/h


a = 0.2
Niteraciones = 120
eps = 1E-8

x = - 1
for zx_ in zx:
  x = x + 1 
  y = -1
  for zy_ in zy: 
      y = y + 1
      Z_0 = complex(zx_, zy_)

      for i in range(0, Niteraciones): 
        Z = Z_0 - a * f(Z_0)/fp(Z_0)
        if( abs(Z-Z_0)<= eps):
          break           
        Z_0 = Z
      #print(i)
      #print(i)
      r, g, b = i % 4 *64 , i % 8 *16, i% 16* 16      
      image.putpixel((x,y), (r,g,b) )

plt.imshow(image)
#plt.xlim(400, 600)
#plt.ylim(400, 523)
````
