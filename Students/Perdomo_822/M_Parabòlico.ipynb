```python
# Condiciones iniciales:

def mov(v_0,a):
  tv= (2*v_0*np.sin(np.deg2rad(a)))/9.8

  t = np.linspace(0, tv, 100)
  x =v_0* np.cos(a)* t
  y = (v_0 * np.sin(np.deg2rad(a))*t) - (9.8*t**2/2)

  return x , y

```

```python
x,y = mov(10,45)

plt.plot(x,y)
```
