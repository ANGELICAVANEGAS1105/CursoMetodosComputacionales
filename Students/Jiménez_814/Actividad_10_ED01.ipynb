### Método Euler


1.   Dividir en intervalos el dominio (desde $t_0 → t_f$).
    
    Con una amplitud dada por:
  $$\Delta t=\frac{t_f-t_0}n$$

2.   $$f(t_i)=\frac{dN}{dt}(t_i)$$

Hacer separación de variables
$$\begin{align}
\frac {dN}N =& -λ dt\\
\ln(N)=& -\lambda t \\
N=&~ e^{-\lambda t}
\end{align}$$



$$ N(0)=N_0\\
t→∞\\
N(t→∞)=0$$
_______________________________________
$$N(t)=y\\
g(t,y) = 1\\
f(t,y) = -λ * N $$

$$
y_{i+1} = y_i + \Delta t f(t,y_0)$$

````python
def dec_euler(f,n,a,b,y0):
  rango = np.linspace(a,b,n)
  h = rango[1]-rango[0]
  Y = np.zeros(n)
  Y[0] = y0
  for i in range(n-1):
    Y[i+1] = Y[i] + h*f(Y[i])
  return np.array([rango, Y])
````

````python
f = lambda x: -lamda*x
a = 0
b = 10
y0 = 1
funcion = dec_euler(f,1000,a,b,y0)
plt.figure()
plt.plot(funcion[0], funcion[1],c = "pink")
plt.xlabel("t (s)")
plt.ylabel("N")
plt.title("Decaimiento radioactivo euler función")
plt.grid(alpha = 0.2)
plt.show()
````

$$
\frac{dy}{dt} = \frac{y_{i+1} - y_i}{\Delta t} = v_i\\
y_{i+1} = y_i + v_i \Delta t\\
\frac{dv}{dt}= \frac{d^2y}{dt^2}= \frac{v_{i+1}-v_i}{\Delta t} = -(\gamma v_i+\omega ^2y_i)\\
v_{i+1} = -\Delta t ( \gamma v_i+ \omega ^2 y_i)+v_i$$

````python
n=10000
````

````python
#----------------------------------------------------------------------------------------------------------------
# Condiciones Iniciales
#----------------------------------------------------------------------------------------------------------------
gamma = 0.5
omega = 10 # rad/s
params = (gamma,omega)

y0 = 10
v0 = 20
y_ci = np.array([y0,v0])

T = np.linspace(0,10,n)
h = T[1]-T[0]  #delta t
y = np.zeros(n)
v = np.zeros(n)

y[0] = y_ci[0]
v[0] = y_ci[1]
````

````python
for i in range(1,n):
  y[i] = y[i-1] + v[i-1]*h
  v[i] = - h * (params[0]*v[i-1] + (params[1]**2)*y[i-1]) + v[i-1] #+ y[i-1]
````

````python
plt.figure()
plt.plot(T,y, c="tomato")
plt.plot(T,v)
plt.xlabel(r"t (s)")
plt.ylabel("y (m)")
plt.title("Oscilador armónico euler")
plt.grid(alpha = 0.3)
plt.show()
````

````python
def arm_euler(params,ci,n,a,b):
  rango = np.linspace(a,b,n)
  h = rango[1]-rango[0]
  Y = np.zeros(n)
  V = np.zeros(n)
  Y[0] = ci[0]
  V[0] = ci[1]

  for i in range(n-1):
    Y[i+1] = Y[i] + V[i]*h
    V[i+1] = -h * (params[0]*v[i-1] + (params[1]**2)*y[i-1]) + v[i-1]
  dicc = {"Altura" : Y, "Velocidad" : V}
  return dicc
````

````python
arm_euler(params,y_ci,n,0,10)
````
