1. **Modelo de población de una especie de pajaros** 
Se estudiará un modelo de crecimiento de la poblacion para una especie de pájaros. En esta población se supone que el número de pajaros hembras es igual al número de machos.  Sea :
- $p_{j,n-1}$ la poblacion juvenil (inmadura) de hembras en el año (n-1) 
- $p_{a,n-1}$ el numero de hembras adultas en el mismo año.

Algunos de los pájaros jóvenes morirán  durante el año. Se suponen que cierta proporción de pájaros $\alpha$ de los pájaros jovenes sobrevivirán para llegar a adultos en la primavera del año $n$. Cada hembra que sobrevive prdocuce huevos en la primavera, los incupan y producen, en promedio, $k$ pájaros hembras jóvenes en al siguiente primavera. Los adultos también mueren y la proporcion de adultos que sobrevive de una primavera a la siguiente es $\beta$.

Un modelo para el crecimiento poblacional de pajaros 
\begin{equation}
p_{j,n} = kp_{a, n-1}
\end{equation}


\begin{equation}
p_{a,n} =\alpha p_{j,n-1} + \beta p_{a, n-1}
\end{equation}

Que puede ser escrita como ${\bf p_n}=A{\bf p_{n-1}}$, Siendo A la matrix definida en las ecuaciones anteriores, y ${\bf p_n}$ un vector. Para la evolución temporal se cumple que: 

- ${\bf p_1}=A{\bf p_{0}}$, ${\bf p_2}=A{\bf p_{1}}= A(A{\bf p_0})=A^2{\bf p_{0}}$ $...$ y asi sucesivamente. Entonces:

\begin{equation}
{\bf p_n} = A^n {\bf p_{0}}
\end{equation}


Supongamos que tenemos 10 hembras  (y 10 machos) adultos y no hay jovenes, la supongmaos que la matriz A viene dada por:

\begin{bmatrix} 
0 & 2 \\
0.3 & 0.5 
\end{bmatrix}

para la primera poblacion $p_{j,1}=20$, $p_{a,1}=5$, el total de hembras es 25 y la razón jovenes adultas es de 4 a 1. Muestre que la razon $p_{j,n}/p_{a,n}$ converge $1.88$,(como no se puede tener un número racional de pájaros,. aproxime siempre el resultados al menor entero).

- Para la  solución general del modelo del problema de valores y vetores propios
se tiene que:
\begin{equation}
p_n=\lambda_1^n\left [ a_1 \vec{v}_1 +\left( \frac{\lambda_2}{\lambda_1} \right)^n \right]a_2 \vec{v}_2
\end{equation}

Asumiendo que $\lambda_2/ \lambda_1 < 1$, entonces $(\lambda_2/ \lambda_1)^n << 1$ cuando n crece.  Entonce spara n grande se cumple que:

donde $p_n=a_1\lambda_1^n\vec{v}_1$


donde:

${\bf p_0} = a_1 \vec{v}_1 + a_2 \vec{v}_2$, con $v_1$ y $v_2$ vectores propios de A

````python
#Vamos a modelar la matriz
import numpy as np
A=np.array([[0,2],[0.3,0.5]])
P0=np.array([0,10])
````

````python

def F(x):
  r=[]
  for i in x:
    i_=int(i)
    P= (np.linalg.matrix_power(A,i_))@P0
    P1= int(P[0])
    P2=int(P[1])
    s= P1/P2
    r.append(s)
  return r


````

````python
x= np.linspace(0,100,100)
print(F(x))
````

````python
import matplotlib.pyplot as plt
plt.axhline(y=1.88,color='y')
plt.plot(x,F(x), color='b')
plt.grid()
````

````python
#Hagamoslo sin ciclos utilizando la opcion vectorize de numpy

def F(x):
  P= np.linalg.matrix_power(A,x)@P0
  P1= int(P[0])
  P2=int(P[1])
  return P1/P2
Fv= np.vectorize(F)
````

````python
import numpy as np

# Definir la matriz de código
codigo = np.array([[1, 2, -3, 4, 5],
                   [-2, -5, 8, -8, -9],
                   [1, 2, -2, 7, 9],
                   [1, 1, 0, 6, 12],
                   [2, 4, -6, 8, 11]])

# Calcular la matriz de código inversa
codigo_inversa = np.linalg.inv(codigo)

# Definir el mensaje encriptado
mensaje_encriptado = np.array([47, 49, -19, 257, 487, 10, -9, 63, 137, 236, 79, 142, -184, 372, 536, 59, 70, -40, 332, 588])

# Obtener el tamaño de la matriz de código
tamano_codigo = codigo.shape[1]

# Dividir el mensaje encriptado en filas según el tamaño de la matriz de código
filas_mensaje = [mensaje_encriptado[i:i+tamano_codigo] for i in range(0, len(mensaje_encriptado), tamano_codigo)]

# Decodificar cada fila del mensaje encriptado utilizando la matriz de código inversa
mensaje_decodificado = ""
for fila in filas_mensaje:
    fila_decodificada = np.dot(fila, codigo_inversa).round().astype(int)
    caracteres_decodificados = [chr(numero + 64) if numero >= 1 and numero <= 26 else ' ' for numero in fila_decodificada]
    mensaje_decodificado += ''.join(caracteres_decodificados)

# Imprimir el mensaje decodificado
print("El mensaje decodificado es:")
print(mensaje_decodificado)
````

````python
import numpy as np

# Definir la matriz A
A = np.array([[2, 2, -1],
              [-3, 1, 3],
              [0, 0, 3]])

# Calcular los eigenvalores y eigenvectores
eigenvalues, eigenvectors = np.linalg.eig(A)

# Imprimir los eigenvalores y eigenvectores
print("Eigenvalores:")
print(eigenvalues)
print("Eigenvectores:")
print(eigenvectors)
````

````python
from scipy.linalg import expm

# Definir el intervalo de tiempo [0, 2]
t_initial = 0
t_final = 2

# Definir la condición inicial r(0)
r_0 = np.array([1, 1, 2])

# Calcular la matriz exponencial de A
A_exp = expm(A * t_final)

# Calcular la solución r(t)
r_t = np.dot(A_exp, r_0)

# Imprimir la solución
print("Solución en el intervalo de tiempo [0, 2]:")
print(r_t)
````

````python
import matplotlib.pyplot as plt

# Crear un arreglo de valores de tiempo en el intervalo [0, 2]
time = np.linspace(t_initial, t_final, 100)

# Calcular las soluciones x(t), y(t), z(t) para cada valor de tiempo
x_t = r_t[0] * np.exp(2 * time) + r_0[0] * (np.exp(2 * time) - 1 - 2 * time)
y_t = r_t[1] * np.exp(2 * time) + r_0[1] * (2 * time * np.exp(2 * time) - 2 * time - 1)
z_t = r_t[2] * np.exp(2 * time) + r_0[2] * np.exp(2 * time) * (1 + np.exp(2 * time))

# Graficar las soluciones
plt.plot(time, x_t, label='x(t)')
plt.plot(time, y_t, label='y(t)')
plt.plot(time, z_t, label='z(t)')
plt.xlabel('Tiempo')
plt.ylabel('Valor')
plt.title('Soluciones del sistema de ecuaciones diferenciales')
plt.legend()
plt.grid()
````

````python
#Grafiquemos la solucion analitica para comparar este resultado con la solucion numerico que obtuvimos. 

# Definir el intervalo de tiempo [0, 2]
t_initial = 0
t_final = 2

# Crear un arreglo de valores de tiempo en el intervalo [0, 2]
time = np.linspace(t_initial, t_final, 100)

# Solución analítica
x_t_analytical = r_0[0] * (np.exp(2 * time) * (1 + np.exp(2 * time) - 2 * time))
y_t_analytical = r_0[1] * (-2 * time * np.exp(2 * time))
z_t_analytical = r_0[2] * (np.exp(2 * time) * (-1 + np.exp(2 * time)))

# Graficar solución analítica
plt.plot(time, x_t_analytical, label='x(t)')
plt.plot(time, y_t_analytical, label='y(t)')
plt.plot(time, z_t_analytical, label='z(t)')
plt.xlabel('Tiempo')
plt.ylabel('Valor')
plt.title('Solución analítica del sistema de ecuaciones diferenciales')
plt.legend()
plt.grid()
````
