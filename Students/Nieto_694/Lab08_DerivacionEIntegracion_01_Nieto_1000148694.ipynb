```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt

#Creamos el dataframe con los datos de tiempo y corriente
t= np.array([1.00,1.01,1.02,1.03,1.04])
i= np.array([3.10,3.12,3.14,3.18,3.24])

pd= pd.DataFrame({'Tiempo':t,'Corriente':i}) #Creacion del dataframe
pd
```

```python
from scipy.interpolate import lagrange
#Realicemos la interpolacion empelando el metodo de lagranje para obtener la i(t) que derivaremos
t_= np.linspace(t.min(),t.max(),100)
f=lagrange(t,i)
i_=f(t_)
print(i_)

#Grafiquemos i vs t
plt.plot(t,i,'o')
plt.plot(t_,i_)
plt.grid()
plt.legend(['Datos', 'Interpolación'])
plt.xlabel('t')
plt.ylabel('i(t)')   
plt.title('i vs t')  
```

```python
#Derivemos la ecuacion que acabamos de encontrar interpolando
from scipy import misc
f_= np.vectorize(f)
di_dt =misc.derivative(f_,t_,dx=1e-8)
print(di_dt)
#Grafiquemos di/dt vs t para observar su comportamiento
plt.plot(di_dt,t_)
plt.grid()
plt.title('di/dt vs t')
plt.xlabel('t')
plt.ylabel('di/dt')   
```

```python
#Grafiquemos tambien di/dt vs i 
plt.plot(di_dt,i_)
plt.grid()
plt.title('di/dt vs i')
plt.xlabel('i')
plt.ylabel('di/dt')   
```

\begin{equation}
\varepsilon(t) = L\frac{\mathrm{d}i}{\mathrm{d}t} + Ri
\end{equation}

```python
# Ahora programemos la funcion de la potencia para ver su comportamiento
def E(di,i):
  L=0.98
  R=0.142
  return L*di+R*i
```

```python
E= E(di_dt,i_) #Calculo de E
E
```

```python
#Grafiquemos la relacion entre E y t.
plt.title('E vs t')
plt.plot(E,t_)
plt.grid()
plt.xlabel('t')
plt.ylabel('E')
```

```python
#Grafiquemos la relacion entre E y di/dt. Comprobamos que sea una relacion lineal
plt.title('E vs di/dt')
plt.plot(E,di_dt)
plt.grid()
plt.xlabel('di/dt')
plt.ylabel('E')
```

```python
#Grafiquemos la relacion entre E y i.
plt.title('E vs i')
plt.plot(E,i_)
plt.grid()
plt.xlabel('i')
plt.ylabel('E')
```

```python
import numpy as np
from scipy import integrate

# Expresión para calcular una integral con el método del trapecio
def integral_trapecio(func, a, b, num_puntos):
    x = np.linspace(a, b, num_puntos)
    y = func(x)
    return integrate.trapz(y, x)

# Expresión para calcular una integral con el método de Simpson
def integral_simpson(func, a, b, num_puntos):
    x = np.linspace(a, b, num_puntos)
    y = func(x)
    return integrate.simps(y, x)

# Expresión para calcular una integral con el método de cuadratura
def integral_cuadratura(func, a, b):
    return integrate.quad(func, a, b)[0]

# Calcular aproximaciones de las integrales con diferentes métodos
epsilon = 1E-5

# ∫π/201tanx−−−−√dx
func1 = lambda x: np.tan(x)**(-0.5)
a1 = 0
b1 = np.pi / 20
approx1_trapz = integral_trapecio(func1, a1, b1, 1000)
approx1_simpson = integral_simpson(func1, a1, b1, 1000)
approx1_cuadratura = integral_cuadratura(func1, a1, b1)
print("Aproximación con método del trapecio:", approx1_trapz)
print("Aproximación con método de Simpson:", approx1_simpson)
print("Aproximación con método de cuadratura:", approx1_cuadratura)

# ∫∞−∞11+x2dx
func2 = lambda x: 1 / (1 + x**2)
a2 = -np.inf
b2 = np.inf
approx2_trapz = integral_trapecio(func2, a2, b2, 1000)
approx2_simpson = integral_simpson(func2, a2, b2, 1000)
approx2_cuadratura = integral_cuadratura(func2, a2, b2)
print("Aproximación con método del trapecio:", approx2_trapz)
print("Aproximación con método de Simpson:", approx2_simpson)
print("Aproximación con método de cuadratura:", approx2_cuadratura)

# ∫π/40∫cosxsinx(2ysinx+cos2x)dxdy
func3 = lambda x, y: np.cos(x) * np.sin(x) * (2 * y * np.sin(x) + np.cos(2 * x))
a3 = 0
b3 = np.pi / 4
c3 = 0
d3 = 1
approx3_trapz = integrate.dblquad(func3, a3, b3, lambda x: c3, lambda x: d3)
approx3_simpson = integrate.dblquad(func3, a3, b3, lambda x: c3, lambda x: d3)
approx3_cuadratura = integrate.dblquad(func3, a3, b3, lambda x: c3, lambda x: d3)
print("Aproximación con método del trapecio:", approx3_trapz[0])
print("Aproximación con método de Simpson:", approx3_simpson[0])
print("Aproximación con método de cuadratura:", approx3_cuadratura[0])
```
